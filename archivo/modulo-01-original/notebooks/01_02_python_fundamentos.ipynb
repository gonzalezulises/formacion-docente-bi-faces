## 📦 Sección 1: Variables y tipos de datos

Una **variable** es como una etiqueta que le pones a un valor para poder usarlo después. Es como cuando en Excel nombras una celda.

### Crear variables

En Python, crear una variable es tan simple como:

```python
nombre_variable = valor
```

### Tipos de datos básicos

| Tipo | Nombre en Python | Ejemplo | Uso común |
|------|-----------------|---------|----------|
| Número entero | `int` | `15`, `2024`, `-3` | Cantidades, años, conteos |

## ➕ Sección 2: Operaciones

### Operaciones aritméticas

| Operador | Operación | Ejemplo | Resultado |
|----------|-----------|---------|----------|
| `+` | Suma | `5 + 3` | `8` |
| `-` | Resta | `20 - 7` | `13` |
| `*` | Multiplicación | `4 * 5` | `20` |
| `/` | División | `15 / 4` | `3.75` |
| `//` | División entera | `15 // 4` | `3` |
| `%` | Módulo (residuo) | `15 % 4` | `3` |
| `**` | Potencia | `2 ** 3` | `8` |

````python
# Calcular el promedio de calificaciones de un estudiante
# En Venezuela, la escala es de 0 a 20

nota_estadistica = 15
nota_microeconomia = 12
nota_contabilidad = 18
nota_matematica = 14
nota_ingles = 16

# Calcular el promedio
suma_notas = nota_estadistica + nota_microeconomia + nota_contabilidad + nota_matematica + nota_ingles
total_materias = 5
promedio = suma_notas / total_materias

print(f"Suma de notas: {suma_notas}")
print(f"Promedio: {promedio}")
print(f"Promedio redondeado: {round(promedio, 2)}")
````

````python
# Calcular tasa de aprobación
inscritos = 45
aprobados = 32
reprobados = 8
retirados = 5

tasa_aprobacion = (aprobados / inscritos) * 100
tasa_reprobacion = (reprobados / inscritos) * 100
tasa_retiro = (retirados / inscritos) * 100

print(f"Tasa de aprobación:  {tasa_aprobacion:.1f}%")
print(f"Tasa de reprobación: {tasa_reprobacion:.1f}%")
print(f"Tasa de retiro:      {tasa_retiro:.1f}%")
print(f"Total:               {tasa_aprobacion + tasa_reprobacion + tasa_retiro:.1f}%")
````

### Operaciones de comparación

Las comparaciones devuelven `True` o `False`:

| Operador | Significado | Ejemplo |
|----------|------------|---------|  
| `==` | Igual a | `nota == 15` |
| `!=` | Diferente de | `nota != 0` |
| `>` | Mayor que | `nota > 10` |
| `<` | Menor que | `nota < 10` |
| `>=` | Mayor o igual | `nota >= 10` (aprobado) |
| `<=` | Menor o igual | `nota <= 20` |

````python
# Comparaciones con notas
nota = 14

print(f"La nota es {nota}")
print(f"¿Aprobado (>= 10)?    {nota >= 10}")
print(f"¿Sobresaliente (>= 18)? {nota >= 18}")
print(f"¿Reprobado (< 10)?    {nota < 10}")
print(f"¿Nota perfecta (== 20)? {nota == 20}")
````

### Operaciones lógicas

Permiten combinar condiciones:

| Operador | Significado | Ejemplo |
|----------|------------|---------|  
| `and` | Y (ambas deben ser verdaderas) | `nota >= 10 and asistencia >= 75` |
| `or` | O (al menos una debe ser verdadera) | `beca == True or trabaja == True` |
| `not` | Negación (invierte el valor) | `not es_retirado` |

````python
# Combinando condiciones
nota = 14
asistencia = 82

# ¿El estudiante aprueba? (nota >= 10 Y asistencia >= 75%)
aprueba = nota >= 10 and asistencia >= 75
print(f"Nota: {nota}, Asistencia: {asistencia}%")
print(f"¿Aprueba la materia? {aprueba}")

# ¿Tiene algún beneficio?
tiene_beca = False
es_trabajador = True
tiene_beneficio = tiene_beca or es_trabajador
print(f"¿Tiene algún beneficio? {tiene_beneficio}")
````

## 📝 Sección 3: Strings (cadenas de texto)

Los strings son fundamentales para trabajar con nombres, carreras, códigos de materias y cualquier dato textual. Se crean con comillas simples o dobles:

```python
nombre = "María García"    # comillas dobles
carrera = 'Economía'        # comillas simples (equivalente)
```

````python
# Concatenación: unir textos con +
nombre = "Carlos"
apellido = "Rodríguez"
titulo = "Prof."

nombre_completo = titulo + " " + nombre + " " + apellido
print(nombre_completo)
````

````python
# f-strings: la forma más elegante de combinar texto y variables
# Se escribe una f antes de las comillas y se ponen las variables entre {}

nombre_profesor = "Ana Méndez"
departamento = "Administración"
estudiantes = 38
promedio = 13.5

# Con f-string
mensaje = f"La profesora {nombre_profesor} del Dpto. de {departamento} tiene {estudiantes} estudiantes con promedio {promedio:.1f}"
print(mensaje)
````

````python
# Métodos útiles de strings
carrera = "  licenciatura en economía  "

print(f"Original:      '{carrera}'")
print(f".strip():      '{carrera.strip()}'"          )  # Quita espacios al inicio y final
print(f".upper():      '{carrera.strip().upper()}'"   )  # Todo en mayúsculas
print(f".lower():      '{carrera.strip().lower()}'"   )  # Todo en minúsculas
print(f".title():      '{carrera.strip().title()}'"   )  # Primera letra de cada palabra en mayúscula
print(f".replace():    '{carrera.strip().replace('economía', 'administración')}'")
````

````python
# .split() divide un texto en partes
codigo_materia = "ECO-301-MICRO"
partes = codigo_materia.split("-")

print(f"Código completo: {codigo_materia}")
print(f"Partes: {partes}")
print(f"Departamento: {partes[0]}")
print(f"Número: {partes[1]}")
print(f"Nombre corto: {partes[2]}")

# len() cuenta la longitud de un texto
print(f"\nLongitud del código: {len(codigo_materia)} caracteres")
````

````python
# Ejemplo práctico: formatear nombres de profesores
nombres_crudos = ["maria garcia", "CARLOS RODRIGUEZ", "  Ana Méndez  ", "pedro LOPEZ"]

print("Nombres formateados:")
for nombre in nombres_crudos:
    nombre_limpio = nombre.strip().title()
    print(f"  '{nombre}' -> '{nombre_limpio}'")
````

## 📋 Sección 4: Listas

Una **lista** es una colección ordenada de elementos. Es como una columna de Excel donde puedes guardar múltiples valores.

Se crean con corchetes `[]`:

```python
mi_lista = [elemento1, elemento2, elemento3]
```

````python
# Crear listas con datos de FACES
carreras = ["Economía", "Administración", "Contaduría", "Relaciones Industriales"]
notas_seccion = [15, 12, 18, 9, 14, 7, 16, 11, 13, 10, 17, 8, 15, 14, 12]
periodos = ["2023-I", "2023-II", "2024-I", "2024-II", "2025-I"]

print(f"Carreras de FACES: {carreras}")
print(f"Cantidad de carreras: {len(carreras)}")
print(f"Notas de la sección: {notas_seccion}")
print(f"Cantidad de notas: {len(notas_seccion)}")
````

````python
# Acceder a elementos por índice (empieza en 0)
print(f"Primera carrera:  {carreras[0]}")
print(f"Segunda carrera:  {carreras[1]}")
print(f"Última carrera:    {carreras[-1]}")
print(f"Penúltima carrera: {carreras[-2]}")

# Slicing: obtener una porción de la lista
print(f"\nPrimeras 3 notas:  {notas_seccion[:3]}")
print(f"Últimas 3 notas:   {notas_seccion[-3:]}")
print(f"Del índice 2 al 5: {notas_seccion[2:6]}")
````

````python
# Modificar listas
sedes = ["Bárbula", "La Morita"]
print(f"Sedes iniciales: {sedes}")

# Agregar un elemento al final
sedes.append("Valencia Centro")
print(f"Después de append: {sedes}")

# Eliminar un elemento
sedes.remove("Valencia Centro")
print(f"Después de remove: {sedes}")

# Ordenar una lista
notas_desordenadas = [15, 8, 12, 19, 10, 7, 14]
notas_desordenadas.sort()
print(f"\nNotas ordenadas (asc): {notas_desordenadas}")

notas_desordenadas.sort(reverse=True)
print(f"Notas ordenadas (desc): {notas_desordenadas}")
````

````python
# Funciones útiles con listas numéricas
notas = [15, 12, 18, 9, 14, 7, 16, 11, 13, 10]

print(f"Notas: {notas}")
print(f"Cantidad:  {len(notas)}")
print(f"Suma:      {sum(notas)}")
print(f"Promedio:  {sum(notas) / len(notas):.1f}")
print(f"Máxima:    {max(notas)}")
print(f"Mínima:    {min(notas)}")
````

````python
# List comprehensions: crear listas de forma compacta
# (No te preocupes si esto parece complejo, es un atajo útil)

notas = [15, 12, 18, 9, 14, 7, 16, 11, 13, 10]

# Filtrar solo las notas aprobatorias (>= 10)
aprobatorias = [nota for nota in notas if nota >= 10]
print(f"Notas aprobatorias: {aprobatorias}")

# Convertir notas a porcentaje (escala 0-100)
porcentajes = [nota * 5 for nota in notas]
print(f"En porcentaje:      {porcentajes}")
````

## 📖 Sección 5: Diccionarios

Un **diccionario** guarda información como pares de **clave: valor**. Es ideal para representar un registro o ficha, como la información de un estudiante.

Se crean con llaves `{}`:

```python
mi_diccionario = {
    "clave1": valor1,
    "clave2": valor2
}
```

````python
# Diccionario: ficha de un estudiante
estudiante = {
    "nombre": "Luis Martínez",
    "cedula": "V-28.456.789",
    "carrera": "Economía",
    "semestre": 6,
    "promedio": 14.8,
    "activo": True,
    "sede": "Bárbula"
}

# Acceder a valores por clave
print(f"Nombre:   {estudiante['nombre']}")
print(f"Carrera:  {estudiante['carrera']}")
print(f"Promedio: {estudiante['promedio']}")
print(f"Activo:   {estudiante['activo']}")
````

````python
# Métodos útiles de diccionarios
print("Claves:", list(estudiante.keys()))
print("Valores:", list(estudiante.values()))

# Recorrer pares clave-valor
print("\nFicha completa del estudiante:")
print("-" * 35)
for clave, valor in estudiante.items():
    print(f"  {clave:12s}: {valor}")
````

````python
# Agregar y modificar elementos
estudiante["email"] = "luis.martinez@uc.edu.ve"     # Agregar nueva clave
estudiante["semestre"] = 7                           # Modificar valor existente

print("Ficha actualizada:")
for clave, valor in estudiante.items():
    print(f"  {clave:12s}: {valor}")
````

## 🚦 Sección 6: Condicionales (if / elif / else)

Los condicionales permiten que tu código tome **decisiones** basadas en condiciones. Es como decir: "Si la nota es mayor a 18, es sobresaliente; si es mayor a 14, es buena..."

### Estructura básica

```python
if condición:
    # código si la condición es verdadera
elif otra_condición:
    # código si esta otra condición es verdadera
else:
    # código si ninguna condición anterior es verdadera
```

**Nota importante:** La indentación (los espacios al inicio de la línea) es obligatoria en Python. Es lo que le dice a Python qué código pertenece a cada bloque.

````python
# Clasificar la nota de un estudiante
nota = 16

if nota >= 18:
    clasificacion = "Sobresaliente"
elif nota >= 15:
    clasificacion = "Distinguido"
elif nota >= 12:
    clasificacion = "Bueno"
elif nota >= 10:
    clasificacion = "Regular"
else:
    clasificacion = "Insuficiente (Reprobado)"

print(f"Nota: {nota} -> Clasificación: {clasificacion}")
````

````python
# Ejemplo práctico: verificar estatus de un estudiante
semestre = 8
promedio = 13.5
materias_pendientes = 4

print(f"Semestre: {semestre}, Promedio: {promedio}, Materias pendientes: {materias_pendientes}")

# ¿Puede optar a beca?
if promedio >= 15 and semestre >= 3:
    print("Puede optar a beca académica")
else:
    print("No cumple requisitos para beca académica")

# ¿Está próximo a graduarse?
if materias_pendientes <= 3 and semestre >= 8:
    print("Está en condición de tesista")
elif materias_pendientes <= 6:
    print("Está próximo a culminar")
else:
    print("Aún tiene un recorrido por delante")
````

````python
# Clasificar múltiples estudiantes
estudiantes_notas = {
    "Ana Pérez": 18,
    "Carlos Gómez": 12,
    "Laura Torres": 8,
    "Miguel López": 15,
    "Sofía Herrera": 20
}

print("Clasificación de estudiantes:")
print("=" * 45)
for nombre, nota in estudiantes_notas.items():
    if nota >= 18:
        clasificacion = "Sobresaliente"
    elif nota >= 15:
        clasificacion = "Distinguido"
    elif nota >= 12:
        clasificacion = "Bueno"
    elif nota >= 10:
        clasificacion = "Regular"
    else:
        clasificacion = "Reprobado"
    
    print(f"  {nombre:20s} | Nota: {nota:2d} | {clasificacion}")
````

## 🔁 Sección 7: Bucles (for y while)

Los **bucles** permiten repetir una acción múltiples veces. Son extremadamente útiles cuando necesitas procesar muchos datos (estudiantes, notas, períodos).

### Bucle for

Repite una acción **para cada elemento** de una colección:

```python
for elemento in coleccion:
    # hacer algo con elemento
```

````python
# Recorrer una lista de carreras
carreras = ["Economía", "Administración", "Contaduría", "Relaciones Industriales"]

print("Carreras de FACES:")
for carrera in carreras:
    print(f"  - {carrera}")
````

````python
# range(): generar secuencias de números
# range(inicio, fin) - el fin NO se incluye

print("Semestres del 1 al 10:")
for semestre in range(1, 11):
    print(f"  Semestre {semestre}")
````

````python
# enumerate(): recorrer con índice y valor
materias = ["Estadística I", "Microeconomía", "Contabilidad I", "Matemática I", "Inglés I"]

print("Plan de estudios - Semestre 1:")
for i, materia in enumerate(materias, start=1):
    print(f"  {i}. {materia}")
````

````python
# Ejemplo práctico: calcular estadísticas de notas
notas = [15, 12, 18, 9, 14, 7, 16, 11, 13, 10, 17, 8, 15, 14, 12]

# Inicializar contadores
total = 0
aprobados = 0
reprobados = 0
sobresalientes = 0

# Procesar cada nota
for nota in notas:
    total += nota  # Sumar para calcular promedio
    
    if nota >= 18:
        sobresalientes += 1
    
    if nota >= 10:
        aprobados += 1
    else:
        reprobados += 1

promedio = total / len(notas)

print("RESUMEN DE LA SECCIÓN")
print("=" * 35)
print(f"Total de estudiantes: {len(notas)}")
print(f"Promedio:             {promedio:.1f}")
print(f"Aprobados:            {aprobados} ({aprobados/len(notas)*100:.1f}%)")
print(f"Reprobados:           {reprobados} ({reprobados/len(notas)*100:.1f}%)")
print(f"Sobresalientes:       {sobresalientes}")
````

````python
# Bucle while: repite mientras una condición sea verdadera
# Útil cuando no sabes cuántas veces necesitas repetir

# Simulación: ¿cuántos semestres necesita un estudiante para
# acumular 180 créditos si aprueba ~30 por semestre?

creditos_acumulados = 0
creditos_por_semestre = 30
creditos_requeridos = 180
semestre = 0

while creditos_acumulados < creditos_requeridos:
    semestre += 1
    creditos_acumulados += creditos_por_semestre
    print(f"  Semestre {semestre}: {creditos_acumulados} créditos acumulados")

print(f"\nEl estudiante necesita {semestre} semestres para completar {creditos_requeridos} créditos.")
````

````python
# Contar estudiantes por categoría de nota
notas = [15, 12, 18, 9, 14, 7, 16, 11, 13, 10, 17, 8, 15, 14, 12, 19, 6, 13, 16, 11]

# Crear contadores por categoría
categorias = {
    "Sobresaliente (18-20)": 0,
    "Distinguido (15-17)": 0,
    "Bueno (12-14)": 0,
    "Regular (10-11)": 0,
    "Reprobado (0-9)": 0
}

for nota in notas:
    if nota >= 18:
        categorias["Sobresaliente (18-20)"] += 1
    elif nota >= 15:
        categorias["Distinguido (15-17)"] += 1
    elif nota >= 12:
        categorias["Bueno (12-14)"] += 1
    elif nota >= 10:
        categorias["Regular (10-11)"] += 1
    else:
        categorias["Reprobado (0-9)"] += 1

print("DISTRIBUCIÓN DE NOTAS POR CATEGORÍA")
print("=" * 45)
for categoria, conteo in categorias.items():
    porcentaje = conteo / len(notas) * 100
    barra = "█" * conteo
    print(f"  {categoria:25s} | {conteo:2d} ({porcentaje:5.1f}%) {barra}")
````

## ⚙️ Sección 8: Funciones

Una **función** es un bloque de código reutilizable. En lugar de escribir el mismo código una y otra vez, lo empaquetas en una función y la llamas cuando la necesites.

### Estructura básica

```python
def nombre_funcion(parametro1, parametro2):
    """Descripción de lo que hace la función (docstring)."""
    # código de la función
    resultado = parametro1 + parametro2
    return resultado
```

- `def` indica que estamos definiendo una función
- Los **parámetros** son los datos de entrada
- El **docstring** (entre `"""`) describe lo que hace
- `return` devuelve el resultado

````python
# Función para clasificar una nota
def clasificar_nota(nota):
    """Clasifica una nota en la escala venezolana (0-20)."""
    if nota >= 18:
        return "Sobresaliente"
    elif nota >= 15:
        return "Distinguido"
    elif nota >= 12:
        return "Bueno"
    elif nota >= 10:
        return "Regular"
    else:
        return "Reprobado"

# Usar la función
print(clasificar_nota(18))  # Sobresaliente
print(clasificar_nota(14))  # Bueno
print(clasificar_nota(8))   # Reprobado

# Usarla en un bucle
notas = [15, 8, 18, 12, 10]
for nota in notas:
    print(f"  Nota {nota:2d} -> {clasificar_nota(nota)}")
````

````python
# Función para calcular estadísticas de una sección
def estadisticas_seccion(notas):
    """Calcula estadísticas básicas de una lista de notas."""
    n = len(notas)
    promedio = sum(notas) / n
    aprobados = sum(1 for nota in notas if nota >= 10)
    tasa_aprobacion = aprobados / n * 100
    
    return {
        "total_estudiantes": n,
        "promedio": round(promedio, 2),
        "nota_maxima": max(notas),
        "nota_minima": min(notas),
        "aprobados": aprobados,
        "reprobados": n - aprobados,
        "tasa_aprobacion": round(tasa_aprobacion, 1)
    }

# Usar la función con diferentes secciones
seccion_a = [15, 12, 18, 9, 14, 7, 16, 11, 13, 10]
seccion_b = [17, 14, 13, 16, 15, 12, 18, 11, 14, 15]

for nombre, notas in [("Sección A", seccion_a), ("Sección B", seccion_b)]:
    stats = estadisticas_seccion(notas)
    print(f"\n{nombre}:")
    for clave, valor in stats.items():
        print(f"  {clave:20s}: {valor}")
````

````python
# Función para contar estudiantes por estatus
def contar_por_estatus(lista_estudiantes):
    """Cuenta estudiantes por su estatus académico."""
    conteo = {}
    for est in lista_estudiantes:
        estatus = est["estatus"]
        if estatus in conteo:
            conteo[estatus] += 1
        else:
            conteo[estatus] = 1
    return conteo

# Datos de ejemplo
estudiantes = [
    {"nombre": "Ana", "carrera": "Economía", "estatus": "Activo", "promedio": 15.2},
    {"nombre": "Carlos", "carrera": "Administración", "estatus": "Activo", "promedio": 12.1},
    {"nombre": "Laura", "carrera": "Contaduría", "estatus": "Retirado", "promedio": 8.5},
    {"nombre": "Miguel", "carrera": "Economía", "estatus": "Egresado", "promedio": 16.8},
    {"nombre": "Sofía", "carrera": "RRII", "estatus": "Activo", "promedio": 14.5},
    {"nombre": "Pedro", "carrera": "Contaduría", "estatus": "Activo", "promedio": 11.2},
    {"nombre": "María", "carrera": "Administración", "estatus": "Retirado", "promedio": 7.8},
]

resultado = contar_por_estatus(estudiantes)
print("Conteo por estatus:")
for estatus, cantidad in resultado.items():
    print(f"  {estatus}: {cantidad}")
````
