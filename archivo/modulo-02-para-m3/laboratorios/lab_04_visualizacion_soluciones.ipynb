```python
# ============================================================
# Configuración inicial
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 12
sns.set_palette('Set2')
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)

# Cargar datasets
df_bcv = pd.read_csv('../../datasets/economia/indicadores_bcv.csv')
df_latam = pd.read_csv('../../datasets/economia/inflacion_latam.csv')

df_bcv['fecha'] = pd.to_datetime(df_bcv['fecha'])

print('✅ Bibliotecas y datasets cargados')
```

---

## Ejercicio 1: Inflación mensual de Venezuela

```python
# ============================================================
# SOLUCIÓN Ejercicio 1: Línea de inflación mensual
# ============================================================

fig, ax = plt.subplots(figsize=(14, 6))

ax.plot(df_bcv['fecha'], df_bcv['inflacion_mensual'], 
        color='steelblue', linewidth=1.5)

ax.set_xlabel('Fecha')
ax.set_ylabel('Inflación mensual (%)')
ax.set_title('Inflación Mensual de Venezuela (BCV)', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f'\n📌 Rango de inflación mensual: {df_bcv["inflacion_mensual"].min():.1f}% a {df_bcv["inflacion_mensual"].max():.1f}%')
print(f'   Inflación promedio mensual: {df_bcv["inflacion_mensual"].mean():.1f}%')
```

---

## Ejercicio 2: Comparar inflación por país

```python
# ============================================================
# SOLUCIÓN Ejercicio 2: Multi-línea inflación
# ============================================================

paises = ['Venezuela', 'Colombia', 'Argentina']
colores = ['#E53935', '#1565C0', '#2E7D32']

fig, ax = plt.subplots(figsize=(14, 6))

for pais, color in zip(paises, colores):
    datos = df_latam[df_latam['pais'] == pais].sort_values('anio')
    ax.plot(datos['anio'], datos['inflacion_anual'], 
            color=color, linewidth=2, marker='o', markersize=4, label=pais)

ax.set_xlabel('Año')
ax.set_ylabel('Inflación anual (%)')
ax.set_title('Comparación de Inflación Anual: Venezuela, Colombia y Argentina',
             fontsize=14, fontweight='bold')
ax.legend(title='País', fontsize=11)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print('\n📌 Nota: Venezuela tiene inflación mucho más alta que los demás,')
print('   lo que puede comprimir visualmente las líneas de Colombia y Argentina.')
print('   Considera usar escala logarítmica (ax.set_yscale("log")) para ver mejor las diferencias.')
```

---

## Ejercicio 3: Top 5 países con mayor inflación promedio

```python
# ============================================================
# SOLUCIÓN Ejercicio 3: Top 5 barras horizontales
# ============================================================

# Calcular inflación promedio por país
inflacion_promedio = df_latam.groupby('pais')['inflacion_anual'].mean().sort_values(ascending=True)
top5 = inflacion_promedio.tail(5)

# Colores: gradiente de intensidad
colores = ['#FFCDD2', '#EF9A9A', '#E57373', '#EF5350', '#D32F2F']

fig, ax = plt.subplots(figsize=(10, 5))

barras = ax.barh(range(len(top5)), top5.values, color=colores, edgecolor='white')
ax.set_yticks(range(len(top5)))
ax.set_yticklabels(top5.index)
ax.set_xlabel('Inflación anual promedio (%)')
ax.set_title('Top 5 Países con Mayor Inflación Promedio en América Latina',
             fontsize=13, fontweight='bold')

# Añadir valores al final de cada barra
for i, v in enumerate(top5.values):
    ax.text(v + max(top5.values) * 0.01, i, f'{v:.1f}%', 
            va='center', fontweight='bold', fontsize=11)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

---

## Ejercicio 4: Scatter PIB vs desempleo

```python
# ============================================================
# SOLUCIÓN Ejercicio 4: Scatter PIB vs desempleo
# ============================================================

fig, ax = plt.subplots(figsize=(12, 7))

# Filtrar datos válidos (sin NaN en las variables de interés)
datos_scatter = df_latam.dropna(subset=['pib_per_capita_usd', 'desempleo_pct'])

sns.scatterplot(data=datos_scatter, x='pib_per_capita_usd', y='desempleo_pct',
                hue='pais', style='pais', s=60, alpha=0.7, ax=ax)

ax.set_xlabel('PIB per cápita (USD)')
ax.set_ylabel('Desempleo (%)')
ax.set_title('Relación entre PIB per cápita y Desempleo en América Latina',
             fontsize=14, fontweight='bold')
ax.legend(title='País', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=9)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Correlación
corr = datos_scatter['pib_per_capita_usd'].corr(datos_scatter['desempleo_pct'])
print(f'\n📌 Correlación PIB per cápita vs Desempleo: {corr:.3f}')
print('   Cada punto es un país-año (ej: Venezuela 2015, Chile 2020, etc.)')
```

---

## Ejercicio 5: Boxplot de inflación por país

```python
# ============================================================
# SOLUCIÓN Ejercicio 5: Boxplot inflación por país
# ============================================================

# Orden por mediana de inflación (mayor a menor)
orden = df_latam.groupby('pais')['inflacion_anual'].median().sort_values(ascending=False).index

fig, ax = plt.subplots(figsize=(14, 6))
sns.boxplot(data=df_latam, x='pais', y='inflacion_anual', 
            order=orden, palette='YlOrRd_r', ax=ax)

ax.set_xlabel('País')
ax.set_ylabel('Inflación anual (%)')
ax.set_title('Distribución de Inflación Anual por País (ordenado por mediana)',
             fontsize=13, fontweight='bold')
ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

print('\n📌 Interpretación:')
for pais in orden[:3]:
    datos = df_latam[df_latam['pais'] == pais]['inflacion_anual']
    print(f'  {pais}: mediana={datos.median():.1f}%, std={datos.std():.1f}%')
print(f'\n  País con mayor variabilidad: {orden[0]} (caja más ancha)')
```

---

## Ejercicio 6: Panel de indicadores BCV

```python
# ============================================================
# SOLUCIÓN Ejercicio 6: Panel de 4 indicadores BCV
# ============================================================

fig, axes = plt.subplots(2, 2, figsize=(16, 10))

indicadores = [
    ('inflacion_mensual', 'Inflación Mensual (%)', '#E53935'),
    ('tipo_cambio_indice', 'Índice de Tipo de Cambio', '#1565C0'),
    ('liquidez_monetaria_indice', 'Índice de Liquidez Monetaria', '#2E7D32'),
    ('reservas_internacionales_mmusd', 'Reservas Internacionales (MM USD)', '#FF6F00')
]

for i, (col, titulo, color) in enumerate(indicadores):
    ax = axes[i // 2, i % 2]
    
    # Filtrar datos válidos para esta columna
    datos = df_bcv[['fecha', col]].dropna()
    
    ax.plot(datos['fecha'], datos[col], color=color, linewidth=1.5)
    ax.set_title(titulo, fontsize=12, fontweight='bold')
    ax.set_xlabel('Fecha')
    ax.set_ylabel(titulo.split('(')[-1].replace(')', '') if '(' in titulo else '')
    ax.grid(True, alpha=0.3)
    ax.tick_params(axis='x', rotation=30)

plt.suptitle('Indicadores Económicos de Venezuela (BCV)', 
             fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

---

## Ejercicio 7: Figura pulida para presentación

```python
# ============================================================
# SOLUCIÓN Ejercicio 7: Figura pulida
# ============================================================

# Preparar datos: reservas internacionales
reservas = df_bcv[['fecha', 'reservas_internacionales_mmusd']].dropna()

fig, ax = plt.subplots(figsize=(14, 6))

# Gráfico de área para las reservas
ax.fill_between(reservas['fecha'], reservas['reservas_internacionales_mmusd'],
                alpha=0.3, color='#1565C0')
ax.plot(reservas['fecha'], reservas['reservas_internacionales_mmusd'],
        color='#1565C0', linewidth=2)

# Anotar máximo y mínimo
idx_max = reservas['reservas_internacionales_mmusd'].idxmax()
idx_min = reservas['reservas_internacionales_mmusd'].idxmin()

max_fecha = reservas.loc[idx_max, 'fecha']
max_valor = reservas.loc[idx_max, 'reservas_internacionales_mmusd']
min_fecha = reservas.loc[idx_min, 'fecha']
min_valor = reservas.loc[idx_min, 'reservas_internacionales_mmusd']

ax.annotate(f'Máximo: ${max_valor:,.0f} MM\n({max_fecha.strftime("%b %Y")})',
            xy=(max_fecha, max_valor), xytext=(30, 20),
            textcoords='offset points', fontsize=10,
            arrowprops=dict(arrowstyle='->', color='gray'),
            bbox=dict(boxstyle='round,pad=0.3', facecolor='lightyellow', alpha=0.8))

ax.annotate(f'Mínimo: ${min_valor:,.0f} MM\n({min_fecha.strftime("%b %Y")})',
            xy=(min_fecha, min_valor), xytext=(30, -30),
            textcoords='offset points', fontsize=10,
            arrowprops=dict(arrowstyle='->', color='gray'),
            bbox=dict(boxstyle='round,pad=0.3', facecolor='lightyellow', alpha=0.8))

# Formato
ax.set_xlabel('Fecha', fontsize=12)
ax.set_ylabel('Miles de Millones de USD', fontsize=12)
ax.set_title('Reservas Internacionales de Venezuela',
             fontsize=16, fontweight='bold', pad=15)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.grid(axis='y', alpha=0.3)

# Nota al pie
ax.text(0.5, -0.12,
        'Fuente: Banco Central de Venezuela (BCV). Datos simulados para fines académicos.',
        transform=ax.transAxes, fontsize=9, color='gray', ha='center')

plt.tight_layout()
# plt.savefig('reservas_venezuela.png', dpi=300, bbox_inches='tight')
plt.show()
```

---

## Ejercicio 8 (Opcional): Heatmap de inflación país x año

```python
# ============================================================
# SOLUCIÓN Ejercicio 8: Heatmap inflación país x año
# ============================================================

# Crear tabla pivote
pivot = df_latam.pivot_table(
    values='inflacion_anual', 
    index='pais', 
    columns='anio'
)

# Ordenar países por inflación promedio (mayor arriba)
orden = pivot.mean(axis=1).sort_values(ascending=False).index
pivot = pivot.loc[orden]

fig, ax = plt.subplots(figsize=(18, 7))

sns.heatmap(pivot, annot=True, fmt='.0f', cmap='YlOrRd',
            linewidths=0.5, ax=ax,
            cbar_kws={'label': 'Inflación anual (%)'})

ax.set_title('Inflación Anual por País y Año — América Latina',
             fontsize=14, fontweight='bold')
ax.set_xlabel('Año')
ax.set_ylabel('País')
ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

print('\n📌 Interpretación:')
print('  - Las celdas más oscuras (rojas) indican inflación más alta')
print('  - Venezuela y Argentina concentran los valores más altos')
print('  - Países como Chile y Perú mantienen inflación baja y estable')
```
