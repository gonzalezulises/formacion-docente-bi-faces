## Caso 4: ¿El dato miente? — Un caso para descubrir

### El estudio

Un informe interno de la Dirección Académica de FACES presenta los siguientes datos:

> **"El rendimiento académico en FACES mejoró significativamente entre 2020 y 2024"**

| Indicador | 2020 | 2024 | Cambio |
|-----------|------|------|--------|
| Promedio general de la facultad | 12.8 | 14.3 | +1.5 puntos |
| Tasa de aprobación | 68% | 79% | +11 pp |
| Estudiantes con honores | 45 | 112 | +149% |
| Índice de satisfacción docente | 3.2/5 | 3.9/5 | +22% |

El informe concluye: *"Las reformas curriculares implementadas en 2021 han sido exitosas. Se recomienda extender el modelo a otras facultades."*

### Análisis: Lo que el informe no dice

Este caso tiene **al menos 5 problemas** que el informe no menciona:

**1. Sesgo de supervivencia**
- En 2020, FACES tenía 3,800 estudiantes. En 2024, tiene 2,900 (caída de 24% por emigración y crisis económica).
- Los estudiantes que desertaron o emigraron tenían, en promedio, peores notas.
- El promedio subió porque los estudiantes con peor rendimiento **ya no están**, no porque los que quedan mejoraron.

**2. Cambio en la escala de evaluación**
- En 2021, la reforma incluyó cambiar el peso de los exámenes finales (de 60% a 40%) y aumentar la evaluación continua.
- Es más fácil obtener buenas notas con evaluación continua — el instrumento cambió, no necesariamente el aprendizaje.

**3. Falacia ecológica**
- El promedio general subió, pero ¿subió en todas las carreras? Podría ser que Administración subió mucho (por tener más estudiantes) mientras Economía bajó.

**4. Causalidad no demostrada**
- Atribuir la mejora a las "reformas de 2021" ignora otros cambios simultáneos: retorno a presencialidad post-COVID, nuevos docentes, cambios en el perfil socioeconómico de los estudiantes.

**5. Cherry-picking temporal**
- 2020 fue un año atípico (pandemia, virtualización forzada). Comparar contra un año de crisis maximiza artificialmente la "mejora". ¿Qué pasa si comparamos contra 2019 o 2018?

### Lección clave
> **Los datos nunca "mienten", pero pueden contar historias incompletas.** Un buen analista siempre pregunta: ¿qué datos faltan? ¿qué cambió además de lo que estoy midiendo? ¿a quién no estoy viendo?

```python
# Visualización: El dato que cambia la narrativa
# Ejecuta esta celda para ver cómo la "mejora" se explica por la composición

import matplotlib.pyplot as plt

años = [2018, 2019, 2020, 2021, 2022, 2023, 2024]
promedio = [13.1, 13.0, 12.8, 13.2, 13.6, 14.0, 14.3]
matricula = [4200, 4100, 3800, 3500, 3300, 3100, 2900]

fig, ax1 = plt.subplots(figsize=(10, 5))

color1 = '#40b4e5'
ax1.set_xlabel('Año', fontsize=12)
ax1.set_ylabel('Promedio general', color=color1, fontsize=12)
ax1.plot(años, promedio, 'o-', color=color1, linewidth=2.5, markersize=8, label='Promedio')
ax1.tick_params(axis='y', labelcolor=color1)
ax1.set_ylim(12, 15)

ax2 = ax1.twinx()
color2 = '#e74c3c'
ax2.set_ylabel('Matrícula total', color=color2, fontsize=12)
ax2.bar(años, matricula, alpha=0.3, color=color2, label='Matrícula')
ax2.tick_params(axis='y', labelcolor=color2)
ax2.set_ylim(0, 5000)

plt.title('¿Mejora real o sesgo de supervivencia?', fontsize=14, fontweight='bold')
fig.tight_layout()
plt.show()

print("📊 Observa: el promedio sube MIENTRAS la matrícula baja.")
print("   Cuando los estudiantes con peores notas desertan,")
print("   el promedio de los que quedan sube automáticamente.")
print("\n   Esto NO significa que la calidad educativa mejoró.")
```
