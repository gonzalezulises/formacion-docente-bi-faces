## Caso 1: Determinantes de la deserción universitaria en América Latina

### Contexto

Diversos estudios del Banco Mundial y la CEPAL han documentado que América Latina tiene una de las tasas de deserción universitaria más altas del mundo: aproximadamente **50% de los estudiantes que ingresan a la educación superior no completan su carrera**.

Un estudio representativo (Ferreyra et al., 2017, Banco Mundial) analizó datos de 12 países latinoamericanos para identificar los factores asociados con la deserción.

### Hallazgos principales

| Factor | Efecto sobre deserción |
|--------|----------------------|
| Nivel socioeconómico bajo | Aumenta la probabilidad de deserción en 2.3x |
| Trabajar >20 horas semanales | Aumenta en 1.8x |
| Primera generación universitaria | Aumenta en 1.5x |
| Preparación académica previa débil | Aumenta en 2.1x |
| Tener beca o apoyo financiero | Reduce en 0.6x |
| Participar en tutorías o mentorías | Reduce en 0.7x |

### Metodología
- Datos de encuestas de hogares y registros universitarios de 12 países
- Modelos de regresión logística con variables de control
- Período: 2000-2015

## Caso 2: Impacto de la inflación en la distribución del ingreso

### Contexto

Venezuela experimentó uno de los procesos inflacionarios más severos de la historia moderna. Diversos análisis del BCV, CEPAL y académicos independientes han estudiado cómo la inflación afecta de manera diferenciada a distintos estratos de la población.

### Datos y hallazgos clave

**Fuentes utilizadas en estudios típicos:**
- Series de precios del BCV (Banco Central de Venezuela)
- Encuesta de Condiciones de Vida (ENCOVI) — UCAB
- Datos de la CEPAL sobre distribución del ingreso en la región

**Hallazgos recurrentes en la literatura:**

| Indicador | Observación |
|-----------|------------|
| Coeficiente de Gini | Aumentó significativamente durante períodos de hiperinflación |
| Pobreza de ingreso | Los quintiles más bajos perdieron poder adquisitivo más rápido |
| Canasta alimentaria | Su costo creció más rápido que los salarios formales |
| Remesas | Se convirtieron en principal fuente de ingreso para 30%+ de hogares |
| Sector informal | Creció como mecanismo de adaptación (dolarización de facto) |

### Debate metodológico

La medición de inflación y pobreza en contextos de hiperinflación presenta desafíos únicos:
- El BCV dejó de publicar datos oficiales durante varios años
- Las encuestas de hogares tienen tasas de respuesta decrecientes
- La economía informal dificulta la medición del ingreso real
- Las remesas y la dolarización parcial crean economías paralelas

## Caso 3: Rotación de personal y clima organizacional

### Contexto

Una empresa de servicios financieros con 800 empleados en Caracas registró una tasa de rotación voluntaria del 28% anual, significativamente superior al promedio del sector (18%). La dirección de RRHH decidió investigar la relación entre clima organizacional y rotación.

### Diseño del estudio

- **Encuesta de clima**: aplicada a 620 empleados (78% de tasa de respuesta)
- **Dimensiones medidas**: liderazgo, compensación, desarrollo profesional, ambiente físico, comunicación, balance vida-trabajo
- **Escala**: Likert 1-5 por cada dimensión
- **Datos cruzados con**: registros de RRHH sobre rotación en los 12 meses siguientes a la encuesta

### Hallazgos

| Dimensión | Puntuación media | Correlación con rotación |
|-----------|:---------------:|:------------------------:|
| Liderazgo | 3.1 | -0.42 |
| Compensación | 2.4 | -0.38 |
| Desarrollo profesional | 2.8 | -0.51 |
| Ambiente físico | 3.8 | -0.12 |
| Comunicación | 3.0 | -0.29 |
| Balance vida-trabajo | 2.6 | -0.45 |

**Conclusión del informe**: "La falta de oportunidades de desarrollo profesional es el principal determinante de la rotación voluntaria."

### La empresa decidió
Invertir $200,000 en un programa de desarrollo de carrera: mentorías internas, subsidio de posgrados, plan de carrera formalizado.
