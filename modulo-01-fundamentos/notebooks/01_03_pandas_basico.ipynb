## 📖 Sección 1: ¿Qué es Pandas?

**Pandas** es una librería de Python especializada en análisis de datos tabulares (datos en filas y columnas, como en Excel).

Tiene dos estructuras principales:

| Estructura | Analogía en Excel | Descripción |
|-----------|-------------------|-------------|
| **Series** | Una columna | Una secuencia de valores con un índice |
| **DataFrame** | Una hoja completa | Una tabla con filas y columnas |

### Series: una sola columna de datos

```python
# Una Series es como una columna de Excel
notas = pd.Series([15, 12, 18, 9, 14, 7, 16])
print("Series de notas:")
print(notas)
print(f"\nTipo: {type(notas)}")
print(f"Promedio: {notas.mean():.1f}")
print(f"Máximo: {notas.max()}")
```

### DataFrame: una tabla completa de datos

```python
# Un DataFrame es como una hoja de cálculo
# Se puede crear desde un diccionario donde cada clave es una columna

datos_estudiantes = {
    "nombre": ["Ana Pérez", "Carlos Gómez", "Laura Torres", "Miguel López", "Sofía Herrera"],
    "carrera": ["Economía", "Administración", "Contaduría", "Economía", "RRII"],
    "semestre": [4, 6, 3, 8, 5],
    "promedio": [15.2, 12.8, 17.1, 14.5, 13.9],
    "activo": [True, True, True, False, True]
}

df = pd.DataFrame(datos_estudiantes)
print("DataFrame de estudiantes:")
df
```

Observa que:
- Cada **columna** tiene un nombre (nombre, carrera, semestre, etc.)
- Cada **fila** tiene un índice numérico automático (0, 1, 2, 3, 4)
- Los datos pueden ser de diferentes tipos (texto, números, booleanos)

```python
# También se puede crear desde una lista de diccionarios
# (cada diccionario es una fila)

registros = [
    {"carrera": "Economía", "inscritos": 120, "periodo": "2024-I"},
    {"carrera": "Administración", "inscritos": 180, "periodo": "2024-I"},
    {"carrera": "Contaduría", "inscritos": 150, "periodo": "2024-I"},
    {"carrera": "RRII", "inscritos": 95, "periodo": "2024-I"},
]

df_inscripcion = pd.DataFrame(registros)
df_inscripcion
```

Vamos a cargar el archivo de matrícula de FACES, que contiene datos reales de inscripción.

## 🔍 Sección 3: Exploración básica

Una vez cargados los datos, lo primero que hacemos siempre es **explorarlos** para entender qué contienen.

```python
# .describe() genera estadísticas descriptivas de las columnas numéricas
print("Estadísticas descriptivas:")
matricula.describe()
```

```python
# .value_counts() cuenta las ocurrencias de cada valor en una columna
# Es como hacer un conteo en Excel con CONTAR.SI

print("Estudiantes por CARRERA:")
print(matricula['carrera'].value_counts())
print(f"\nTotal: {matricula['carrera'].value_counts().sum():,}")
```

```python
# Conteo por estatus
print("Estudiantes por ESTATUS:")
print(matricula['estatus'].value_counts())
```

```python
# Conteo por género
print("Estudiantes por GÉNERO:")
print(matricula['genero'].value_counts())
print()

# value_counts con porcentajes (normalize=True)
print("Porcentaje por género:")
print(matricula['genero'].value_counts(normalize=True).apply(lambda x: f"{x:.1%}"))
```

```python
# .unique() y .nunique(): valores únicos
print(f"Carreras disponibles: {matricula['carrera'].unique()}")
print(f"Número de carreras:   {matricula['carrera'].nunique()}")
print()
print(f"Sedes disponibles:   {matricula['sede'].unique()}")
print(f"Número de sedes:     {matricula['sede'].nunique()}")
print()
print(f"Períodos disponibles: {sorted(matricula['periodo'].unique())}")
print(f"Número de períodos:  {matricula['periodo'].nunique()}")
```

## 🎯 Sección 4: Selección de columnas y filas

### Seleccionar filas con .loc[] y .iloc[]

| Método | Tipo de selección | Ejemplo |
|--------|-------------------|---------|  
| `.loc[]` | Por etiqueta/nombre | `df.loc[0:5, 'carrera']` |
| `.iloc[]` | Por posición numérica | `df.iloc[0:5, 0:3]` |

```python
# .iloc[] selecciona por posición (como en Excel: fila 1, columna A)
# Primeras 5 filas, primeras 3 columnas
print("Primeras 5 filas, primeras 3 columnas:")
matricula.iloc[:5, :3]
```

```python
# .loc[] selecciona por nombre de columna
# Primeras 5 filas, columnas específicas
print("Primeras 5 filas, columnas seleccionadas:")
matricula.loc[:4, ['carrera', 'genero', 'sede', 'estatus']]
```

## 📊 Sección 5: Filtrado de datos

El filtrado es una de las operaciones más importantes. Permite seleccionar solo las filas que cumplen ciertas condiciones, como "mostrar solo estudiantes de Economía" o "filtrar solo los retirados".

### Filtro simple: una condición

```python
# Filtrar estudiantes de Economía
economia = matricula[matricula['carrera'] == 'Economia']

print(f"Estudiantes de Economía: {len(economia):,}")
print(f"De un total de:          {len(matricula):,}")
print(f"Porcentaje:              {len(economia)/len(matricula)*100:.1f}%")
print("\nPrimeras filas:")
economia.head()
```

```python
# Filtrar estudiantes retirados
retirados = matricula[matricula['estatus'] == 'Retirado']

print(f"Estudiantes retirados: {len(retirados):,}")
print(f"\nDistribución por carrera:")
print(retirados['carrera'].value_counts())
```

```python
# Filtrar por edad (mayores de 22 años)
mayores_22 = matricula[matricula['edad'] > 22]

print(f"Estudiantes mayores de 22 años: {len(mayores_22):,}")
print(f"Porcentaje del total:           {len(mayores_22)/len(matricula)*100:.1f}%")
```

### Filtro múltiple: varias condiciones

Para combinar condiciones:
- Usar `&` para **Y** (ambas condiciones deben cumplirse)
- Usar `|` para **O** (al menos una debe cumplirse)
- Cada condición debe ir entre paréntesis `()`

```python
# Estudiantes de Economía que están activos
eco_activos = matricula[
    (matricula['carrera'] == 'Economia') & 
    (matricula['estatus'] == 'Activo')
]

print(f"Estudiantes de Economía activos: {len(eco_activos):,}")
print(f"\nDistribución por género:")
print(eco_activos['genero'].value_counts())
```

```python
# Estudiantes mujeres de La Morita que son egresadas
filtro = matricula[
    (matricula['genero'] == 'F') & 
    (matricula['sede'] == 'La Morita') & 
    (matricula['estatus'] == 'Egresado')
]

print(f"Mujeres egresadas de La Morita: {len(filtro):,}")
print(f"\nDistribución por carrera:")
print(filtro['carrera'].value_counts())
```

```python
# Estudiantes de Economía O Administración (usando |)
eco_o_admin = matricula[
    (matricula['carrera'] == 'Economia') | 
    (matricula['carrera'] == 'Administracion')
]

print(f"Estudiantes de Economía o Administración: {len(eco_o_admin):,}")

# Alternativa más elegante con .isin()
eco_o_admin_v2 = matricula[
    matricula['carrera'].isin(['Economia', 'Administracion'])
]
print(f"Con .isin(): {len(eco_o_admin_v2):,} (mismo resultado)")
```

## 📈 Sección 6: Operaciones básicas

### Ordenar datos

```python
# Ordenar por edad (descendente)
print("Estudiantes de mayor a menor edad:")
matricula.sort_values('edad', ascending=False).head(10)
```

```python
# Ordenar por múltiples columnas
print("Ordenado por carrera y luego por edad:")
matricula.sort_values(['carrera', 'edad']).head(10)
```

### Agrupar y resumir con .groupby()

`.groupby()` es una de las funciones más poderosas de Pandas. Permite agrupar datos por una o más columnas y calcular estadísticas para cada grupo. Es como hacer una **tabla dinámica** en Excel.

```python
# Contar estudiantes por carrera
print("¿Cuántos estudiantes hay por carrera?")
print("=" * 40)
conteo_carreras = matricula.groupby('carrera').size()
print(conteo_carreras)
print(f"\nTotal: {conteo_carreras.sum():,}")
```

```python
# Edad promedio por carrera
print("Edad promedio por carrera:")
print("=" * 35)
print(matricula.groupby('carrera')['edad'].mean().round(1))
```

```python
# Múltiples estadísticas con .agg()
print("Estadísticas de edad por carrera:")
print("=" * 50)
stats_carrera = matricula.groupby('carrera')['edad'].agg(['count', 'mean', 'min', 'max'])
stats_carrera.columns = ['Cantidad', 'Edad Media', 'Edad Mínima', 'Edad Máxima']
stats_carrera = stats_carrera.round(1)
print(stats_carrera)
```

```python
# Agrupar por período y contar para ver la tendencia de matrícula
print("Evolución de la matrícula por período:")
print("=" * 40)
tendencia = matricula.groupby('periodo').size()
print(tendencia)
print(f"\nPeríodo con más inscripciones:  {tendencia.idxmax()} ({tendencia.max():,})")
print(f"Período con menos inscripciones: {tendencia.idxmin()} ({tendencia.min():,})")
```

```python
# Agrupar por DOS columnas: período y carrera
print("Matrícula por período y carrera (primeros períodos):")
print("=" * 55)
cruce = matricula.groupby(['periodo', 'carrera']).size().reset_index(name='cantidad')
print(cruce.head(12))
```

```python
# Tabla cruzada: estatus por carrera
print("Tabla cruzada: estatus por carrera")
print("=" * 55)
tabla_cruzada = pd.crosstab(matricula['carrera'], matricula['estatus'])
print(tabla_cruzada)
print(f"\nTotal por estatus:")
print(tabla_cruzada.sum())
```

## 📋 Resumen

En esta sesión aprendimos a usar Pandas para análisis de datos tabulares:

| Operación | Función | Ejemplo |
|-----------|---------|---------|  
| **Cargar CSV** | `pd.read_csv()` | `df = pd.read_csv('archivo.csv')` |
| **Primeras filas** | `.head()` | `df.head(10)` |
| **Dimensiones** | `.shape` | `df.shape` -> (filas, columnas) |
| **Info general** | `.info()` | `df.info()` |
| **Estadísticas** | `.describe()` | `df.describe()` |
| **Conteo de valores** | `.value_counts()` | `df['carrera'].value_counts()` |
| **Valores únicos** | `.unique()`, `.nunique()` | `df['carrera'].unique()` |
| **Seleccionar columna** | `df['col']` | `df['carrera']` |
| **Seleccionar varias** | `df[['c1','c2']]` | `df[['carrera','genero']]` |
| **Filtrar filas** | `df[condición]` | `df[df['carrera'] == 'Economia']` |
| **Filtro múltiple** | `&` y `\|` | `df[(cond1) & (cond2)]` |
| **Ordenar** | `.sort_values()` | `df.sort_values('edad')` |
| **Agrupar** | `.groupby()` | `df.groupby('carrera').size()` |
| **Guardar CSV** | `.to_csv()` | `df.to_csv('salida.csv', index=False)` |
