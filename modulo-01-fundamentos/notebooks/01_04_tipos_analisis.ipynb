## 1. La escalera analítica

No todos los análisis de datos son iguales. Existe una progresión natural desde lo más simple hasta lo más complejo, donde cada nivel **requiere** los anteriores:

```
                                    ┌─────────────────┐
                                    │  PRESCRIPTIVO   │  ¿Qué debemos hacer?
                               ┌────┴─────────────────┴────┐
                               │       PREDICTIVO          │  ¿Qué podría pasar?
                          ┌────┴───────────────────────────┴────┐
                          │          INFERENCIAL                │  ¿Se puede generalizar?
                     ┌────┴────────────────────────────────────┴────┐
                     │             EXPLORATORIO                     │  ¿Qué patrones hay?
                ┌────┴─────────────────────────────────────────────┴────┐
                │                DESCRIPTIVO                            │  ¿Qué pasó?
                └──────────────────────────────────────────────────────┘
```

Cada nivel agrega complejidad, incertidumbre y valor.

## 2. Análisis descriptivo: ¿Qué pasó?

### Definición
Resume y describe las características principales de un conjunto de datos **sin intentar explicar por qué**.

### Herramientas típicas
- Medidas de tendencia central (media, mediana, moda)
- Medidas de dispersión (rango, desviación estándar)
- Tablas de frecuencia y tablas cruzadas
- Gráficos básicos (barras, líneas, histogramas)

### Ejemplo: Estadísticas de matrícula en FACES

| Indicador | Valor |
|-----------|-------|
| Total matriculados 2025-I | 3,450 |
| % Mujeres | 58% |
| Carrera con más inscritos | Administración (32%) |
| Promedio general | 13.8/20 |
| Tasa de deserción semestral | 12% |

### Pregunta que responde
"¿Cuántos estudiantes tenemos, de dónde vienen y cómo les va?"

### Lo que NO puede responder
Por qué la deserción es del 12% ni qué hacer al respecto.

## 3. Análisis exploratorio (EDA): ¿Qué patrones hay?

### Definición
Busca **patrones, anomalías y relaciones** en los datos sin hipótesis previas. Es el detective que investiga antes de que haya sospechosos.

### Herramientas típicas
- Gráficos de dispersión y correlación
- Boxplots para detectar valores atípicos
- Mapas de calor de correlaciones
- Análisis de distribuciones

### Ejemplo: Exploración de rendimiento académico

Al explorar los datos de rendimiento en FACES, podrían surgir patrones como:
- Los estudiantes que trabajan más de 30 horas tienen promedios significativamente menores
- Hay una caída pronunciada de promedio entre 3er y 4to semestre
- Los estudiantes de turnos nocturnos tienen mayor dispersión en sus notas

### Pregunta que responde
"¿Hay algo interesante en estos datos que valga la pena investigar más?"

### Lo que NO puede responder
Si esos patrones son generalizables o si son relaciones causales.

## 4. Análisis inferencial: ¿Se puede generalizar?

### Definición
Usa **muestras** para hacer conclusiones sobre **poblaciones**, cuantificando la incertidumbre.

### Herramientas típicas
- Intervalos de confianza
- Pruebas de hipótesis (t-test, chi-cuadrado, ANOVA)
- Análisis de varianza
- Regresión con significancia estadística

### Ejemplo: Encuesta a muestra de estudiantes

> Se encuestó a 200 estudiantes de FACES (de 3,450 total). El 73% reportó que la carga académica es "excesiva" o "muy excesiva". Con un intervalo de confianza del 95%, la proporción real está entre 67% y 79%.

### Pregunta que responde
"Si encontré este patrón en mi muestra, ¿puedo afirmar que existe en toda la población?"

### Conceptos clave
- **Margen de error**: qué tan imprecisa puede ser mi estimación
- **Significancia estadística**: ¿el resultado es distinto de lo que esperaría por azar?
- **Tamaño de muestra**: determina qué tan precisa es mi inferencia

### Advertencia
"Estadísticamente significativo" no significa "importante" ni "grande". Con muestras muy grandes, diferencias triviales pueden ser significativas.

## 5. Análisis predictivo: ¿Qué podría pasar?

### Definición
Usa datos históricos para **estimar qué ocurrirá en el futuro** o en casos no observados.

### Herramientas típicas
- Regresión lineal y logística
- Árboles de decisión
- Series temporales
- Modelos de clasificación

### Ejemplo: Modelo de predicción de deserción

> Usando datos históricos de 5 años, se construyó un modelo que predice la probabilidad de deserción de cada estudiante al inicio del semestre, basándose en: promedio acumulado, asistencia, horas de trabajo y nivel socioeconómico. El modelo identifica correctamente al 78% de los estudiantes que efectivamente desertan.

### Pregunta que responde
"Basándome en lo que ya pasó, ¿qué es probable que ocurra?"

### Lo que NO puede responder
**Por qué** ocurrirá. Un modelo puede predecir deserción sin explicar sus causas.

## 6. Análisis prescriptivo: ¿Qué debemos hacer?

### Definición
Recomienda **acciones específicas** basadas en los resultados de los análisis anteriores.

### Herramientas típicas
- Optimización matemática
- Simulación de escenarios (what-if)
- Análisis costo-beneficio
- Árboles de decisión con utilidad esperada

### Ejemplo: Optimización de recursos académicos

> Dado que el modelo predictivo identifica 280 estudiantes en riesgo de deserción, y el presupuesto alcanza para intervenir a 100, el análisis prescriptivo recomienda:
> - Priorizar a los 100 con mayor probabilidad de deserción Y mayor probabilidad de responder a la intervención
> - Asignar tutorías personalizadas a los 50 con problemas académicos
> - Asignar apoyo socioeconómico a los 50 con problemas financieros
> - Resultado esperado: reducción de deserción en 23% vs. intervención aleatoria

### Pregunta que responde
"De todas las opciones posibles, ¿cuál nos da el mejor resultado dado nuestros recursos y restricciones?"

### Nota importante
El análisis prescriptivo es el más sofisticado y el que más valor genera, pero **depende completamente** de la calidad de los niveles anteriores.

## 7. Resumen comparativo

| Tipo | Pregunta | Complejidad | Ejemplo FACES |
|------|----------|:-----------:|---------------|
| Descriptivo | ¿Qué pasó? | Baja | Reporte de matrícula semestral |
| Exploratorio | ¿Qué patrones hay? | Media-baja | EDA de calificaciones por carrera |
| Inferencial | ¿Se puede generalizar? | Media | Encuesta de satisfacción con margen de error |
| Predictivo | ¿Qué podría pasar? | Media-alta | Modelo de deserción temprana |
| Prescriptivo | ¿Qué debemos hacer? | Alta | Asignación óptima de becas y tutorías |

## 7b. Un mismo problema, cinco niveles: el caso de la deserción en FACES

Para ver la progresión completa, apliquemos los cinco tipos de análisis al **mismo problema**: la deserción estudiantil.

| Nivel | Pregunta | Qué harías | Resultado |
|-------|----------|-----------|----------|
| **Descriptivo** | ¿Cuál es la tasa de deserción? | Contar desertores / total matriculados por semestre | "La deserción fue 12% en 2025-I, 15% en 2025-II" |
| **Exploratorio** | ¿Hay patrones? | Cruzar deserción con carrera, turno, semestre, empleo | "Los estudiantes de turno nocturno en 3er semestre tienen 3x más deserción" |
| **Inferencial** | ¿Es generalizable? | Encuestar 200 desertores, calcular IC | "Con 95% de confianza, entre 60% y 72% de los desertores citan razones económicas" |
| **Predictivo** | ¿Quién desertará? | Modelo logístico con datos históricos | "Estos 85 estudiantes tienen >70% de probabilidad de desertar este semestre" |
| **Prescriptivo** | ¿Qué hacemos? | Optimizar asignación de 50 becas | "Asignar becas a estos 50 estudiantes reduce la deserción esperada en 28%" |

**Observa la progresión:**
- Cada nivel usa los resultados del anterior
- Cada nivel agrega más valor pero también más incertidumbre
- No siempre necesitas llegar al nivel prescriptivo — a veces un buen análisis descriptivo es suficiente para actuar

````python
# Esquema mental rápido para clasificar tu análisis:

def tipo_de_analisis(pregunta: str) -> str:
    """
    Guía simplificada (no es un clasificador real,
    es una herramienta de pensamiento).
    """
    if "cuántos" in pregunta or "qué porcentaje" in pregunta:
        return "Descriptivo: resume lo que ya pasó"
    elif "hay relación" in pregunta or "qué patrones" in pregunta:
        return "Exploratorio: busca patrones sin hipótesis previa"
    elif "se puede generalizar" in pregunta or "es significativo" in pregunta:
        return "Inferencial: extrapola de la muestra a la población"
    elif "qué pasará" in pregunta or "cuál es la probabilidad" in pregunta:
        return "Predictivo: estima el futuro basado en el pasado"
    elif "qué debemos hacer" in pregunta or "cuál es la mejor opción" in pregunta:
        return "Prescriptivo: recomienda acciones óptimas"
    else:
        return "Reflexiona: ¿qué tipo de respuesta necesitas?"
````

## 9. Reflexión de cierre

- **No siempre necesitas el análisis más sofisticado.** A veces un buen análisis descriptivo es más útil que un modelo predictivo mal planteado.
- **Cada nivel tiene sus requisitos.** No se puede predecir sin describir primero. No se puede prescribir sin predecir.
- **La pregunta determina el análisis**, no al revés. Primero define qué necesitas saber, luego elige la herramienta.
