## 1. ¿Por qué importan los sesgos?

Un sesgo en el análisis de datos no es un error técnico: es una **distorsión sistemática** que hace que nuestras conclusiones no reflejen la realidad. En ciencias sociales, los sesgos son especialmente peligrosos porque:

1. Los datos tratan sobre **personas y sociedades**, donde las consecuencias de decisiones equivocadas son directas.
2. El investigador es parte del mismo sistema que estudia — sus **marcos mentales** influyen en qué pregunta, qué mide y cómo interpreta.
3. Los datos sociales rara vez son experimentales: trabajamos con datos **observacionales** donde las relaciones causales son difíciles de establecer.

## 2. Sesgo de selección: muestras no representativas

### ¿Qué es?
Ocurre cuando la muestra analizada **no representa** a la población que queremos estudiar.

### Ejemplos en ciencias sociales

**Encuesta de satisfacción laboral por correo electrónico**
- Problema: solo responden quienes tienen acceso a correo y tiempo para responder
- Los más insatisfechos (sobrecargados) probablemente no responden
- Resultado: satisfacción inflada artificialmente

**Estudio sobre ingresos usando declaraciones de impuestos**
- Problema: excluye a la economía informal (que en Venezuela puede superar el 50%)
- Resultado: subestimación de la desigualdad real

**Evaluación docente al final del semestre**
- Problema: los estudiantes que desertaron de la materia no evalúan
- Resultado: el docente recibe evaluación solo de quienes sobrevivieron el curso

### Cómo mitigarlo
- Analizar la **tasa de respuesta** y el perfil de los no-respondientes
- Usar muestreo estratificado cuando sea posible
- Reportar siempre las limitaciones de la muestra

## 3. Confusión correlación vs. causalidad

### La falacia más peligrosa en análisis de datos

**Correlación**: dos variables se mueven juntas (positiva o negativamente).  
**Causalidad**: una variable **produce** el cambio en la otra.

### Ejemplos

**En economía:**
> "Los países con más inversión extranjera tienen mayor PIB per cápita."

¿La inversión extranjera causa mayor PIB? ¿O los países con mayor PIB atraen más inversión? ¿O ambos son resultado de instituciones sólidas? Este es un problema clásico de **endogeneidad**.

**En educación:**
> "Los estudiantes que asisten a tutorías tienen mejores notas."

¿Las tutorías causan mejores notas? ¿O los estudiantes más motivados (que ya tendrían buenas notas) son los que asisten a tutorías? **Autoselección**.

**En gestión:**
> "Las empresas que invierten en capacitación tienen menor rotación."

¿La capacitación reduce la rotación? ¿O las empresas con recursos para capacitar también pagan mejor, y es el salario lo que retiene? **Variable omitida**.

### Las tres trampas de la causalidad espuria

1. **Causalidad inversa**: B causa A, no A causa B
2. **Variable confusora**: C causa tanto A como B
3. **Coincidencia estadística**: con suficientes variables, siempre habrá correlaciones espurias

## 4. Sobreajuste ideológico

### ¿Qué es?
Interpretar datos exclusivamente desde un marco teórico preconcebido, **ignorando evidencia que lo contradice** o **forzando los datos** para que confirmen la hipótesis.

### Manifestaciones comunes

- **Sesgo de confirmación**: buscar solo datos que apoyen la tesis propia
- **Cherry-picking**: seleccionar el período temporal o la submuestra que conviene
- **Mover los postes**: cambiar la hipótesis después de ver los resultados

### Ejemplo en economía

> Un investigador convencido de que el libre comercio siempre es beneficioso analiza datos de apertura comercial en América Latina. Encuentra que 3 de 7 países mostraron crecimiento post-apertura. Reporta solo esos 3 como "evidencia" de su tesis.

### Ejemplo en relaciones industriales

> Un consultor que vende programas de bienestar laboral presenta un estudio donde las empresas con programas de bienestar tienen mayor productividad. No menciona que esas empresas también tienen mayores salarios, mejor tecnología y más antigüedad en el mercado.

### Cómo mitigarlo
- Buscar **activamente** evidencia contra tu hipótesis
- Pre-registrar hipótesis antes de analizar los datos
- Pedir a un colega con perspectiva diferente que revise tu análisis

## 5. Sesgo de supervivencia

### ¿Qué es?
Analizar solo los casos **que sobrevivieron** a un proceso, ignorando los que no.

### Ejemplos en datos organizacionales

**"Las empresas exitosas tienen X característica"**
- Problema: no analizamos las empresas que también tenían X pero fracasaron
- El libro "Built to Last" de Jim Collins seleccionó empresas "visionarias" exitosas; muchas dejaron de serlo después

**"Los egresados de nuestra facultad tienen buenos salarios"**
- Problema: solo encuestamos a quienes se graduaron. ¿Qué pasó con los que desertaron?
- Si 50% deserta, estamos analizando solo la mitad más exitosa

**"Nuestro programa de emprendimiento tiene 80% de éxito"**
- Problema: ¿cómo medimos "éxito"? ¿A los 6 meses, al año, a los 3 años?
- Los emprendimientos que cerraron en el mes 3 no aparecen en la evaluación del año 1

### Cómo mitigarlo
- Siempre preguntar: **¿quién falta en estos datos?**
- Incluir datos de fracasos, no solo de éxitos
- Usar análisis de cohorte completa cuando sea posible

## 6. Otros sesgos frecuentes

| Sesgo | Descripción | Ejemplo |
|-------|------------|----------|
| **Anclaje** | La primera información recibida influye desproporcionadamente | "El PIB cayó 15%" condiciona toda la discusión posterior |
| **Disponibilidad** | Sobreestimar lo que recordamos fácilmente | Creer que la criminalidad aumentó porque hay más noticias al respecto |
| **Efecto manada** | Aceptar conclusiones porque "todos" las aceptan | "Todo el mundo sabe que la inflación es solo monetaria" |
| **Falacia ecológica** | Atribuir al individuo lo que es cierto del grupo | "El promedio del país subió, entonces todos mejoraron" |
| **Paradoja de Simpson** | Una tendencia que aparece en grupos se invierte al combinarlos | Un tratamiento que funciona en hombres y mujeres por separado puede parecer ineficaz al combinar los datos |

## 7b. Respuestas sugeridas

### Afirmación 1: Gasto educativo → mejores resultados PISA

**Sesgo principal:** Confusión correlación vs. causalidad + variable omitida

**Por qué la conclusión podría ser incorrecta:**
- Los países con mayor gasto también suelen tener instituciones más sólidas, menor desigualdad, docentes mejor formados y familias con mayor capital cultural. Cualquiera de estos factores (o todos juntos) podría explicar los mejores resultados — no el gasto por sí solo.
- Además, hay ejemplos que contradicen la relación: EE.UU. gasta más por alumno que Finlandia pero tiene peores resultados en PISA. El gasto es necesario pero no suficiente.

---

### Afirmación 2: Emprendedores exitosos trabajan >12 horas

**Sesgo principal:** Sesgo de supervivencia

**Por qué la conclusión podría ser incorrecta:**
- Solo entrevistaron a emprendedores **exitosos**. No sabemos cuántos emprendedores que también trabajan >12 horas fracasaron. Si el 90% de los que trabajan 12+ horas también fracasa, entonces las horas no predicen el éxito.
- Es como preguntar a los ganadores de la lotería qué números eligieron: cualquier respuesta parece "la estrategia ganadora" porque no entrevistamos a los que perdieron.

---

### Afirmación 3: Encuesta online de satisfacción con modalidad virtual

**Sesgo principal:** Sesgo de selección (múltiple)

**Por qué la conclusión podría ser incorrecta:**
- **Autoselección**: responden quienes siguen la página de la facultad (los más conectados y probablemente más cómodos con lo virtual).
- **Exclusión**: los estudiantes sin acceso a internet estable — precisamente los más afectados por la virtualidad — probablemente no siguen la página ni responden encuestas online.
- **Sesgo de canal**: encuestar sobre virtualidad *por medios virtuales* excluye sistemáticamente a quienes la virtualidad perjudica.

## 8. Reflexión de cierre

- Los sesgos no son señal de incompetencia: son **parte de la condición humana**. Lo importante es reconocerlos y mitigarlos.
- Pensar con datos no significa confiar ciegamente en los números. Significa **cuestionar** qué hay detrás de cada dato.
- La mejor herramienta contra el sesgo es la **pregunta incómoda**: ¿qué tendría que ser cierto para que mi conclusión sea incorrecta?
