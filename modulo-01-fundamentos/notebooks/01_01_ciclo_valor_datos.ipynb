## 1. De la pregunta a la decisión

Todo proceso de análisis de datos sigue un ciclo que transforma **preguntas** en **acciones**:

```
Pregunta → Datos → Análisis → Hallazgo → Acción
(Question)  (Data)  (Analysis) (Insight)  (Action)
```

### ¿Por qué un ciclo y no un proceso lineal?

Porque cada acción genera nuevas preguntas. El análisis de datos no termina con un informe: termina cuando la decisión tomada produce resultados que podemos volver a medir.

### Las cinco etapas

| Etapa | Descripción | Ejemplo |
|-------|------------|----------|
| **Pregunta** | Definir qué necesitamos saber | ¿Por qué la deserción aumentó 15% este semestre? |
| **Datos** | Identificar y recopilar información relevante | Registros académicos, encuestas a estudiantes |
| **Análisis** | Procesar, explorar y modelar los datos | Cruzar calificaciones con factores socioeconómicos |
| **Hallazgo** | Extraer conclusiones interpretables | Los estudiantes que trabajan >30h/semana desertan 3x más |
| **Acción** | Tomar decisiones basadas en evidencia | Implementar horarios flexibles y tutorías focalizadas |

## 1b. Del ciclo conceptual al flujo profesional

El ciclo de 5 pasos que acabamos de ver es una simplificación útil para pensar. Pero en la práctica profesional, los proyectos de datos siguen un flujo más detallado y **iterativo**.

El modelo más utilizado en la industria es el **flujo de trabajo de ciencia de datos** (Wickham & Grolemund, *R for Data Science*), adoptado universalmente tanto en Python como en R:

```
                    ┌─────────────────────────────┐
                    │        Loop iterativo        │
                    │                              │
Import → Tidy → Transform ⇄ Visualize → Communicate
                         ⇅        ⇅
                        Model ←──┘
```

### ¿Qué significa cada paso?

| Paso | ¿Qué hace? | Ejemplo en FACES |
|------|-----------|------------------|
| **Import** | Cargar datos desde archivos, bases de datos o APIs | Leer la planilla de notas en Excel o descargar datos del BCV |
| **Tidy** | Organizar los datos: cada variable en una columna, cada observación en una fila | Reestructurar una tabla donde los semestres están en columnas |
| **Transform** | Filtrar, crear nuevas variables, agrupar, resumir | Calcular la tasa de deserción por carrera y semestre |
| **Visualize** | Crear gráficos para descubrir patrones | Histograma de promedios, dispersión notas vs. horas de trabajo |
| **Model** | Ajustar modelos estadísticos o de ML | Regresión para estimar factores de deserción |
| **Communicate** | Presentar hallazgos a quienes toman decisiones | Dashboard para el Consejo de Facultad, informe ejecutivo |

### ¿Cómo se conecta con nuestro ciclo simple?

| Ciclo simple | Flujo profesional |
|:---:|:---:|
| Pregunta | *Fase previa*: definir el problema antes de tocar datos |
| Datos | **Import** + **Tidy** |
| Análisis | **Transform** + **Visualize** + **Model** (el loop iterativo) |
| Hallazgo | Resultado del modelado y la exploración |
| Acción | **Communicate** + toma de decisiones |

> **Clave:** El ciclo simple es *lineal*. El flujo profesional tiene un **loop** — transform, visualize y model se repiten muchas veces hasta encontrar la respuesta. Esto refleja la realidad: el análisis de datos es un proceso de descubrimiento, no una receta paso a paso.

## 1c. Las tres fases de un proyecto de datos real

En el ámbito empresarial y organizacional, los proyectos de ciencia de datos tienen **tres grandes fases**:

```
┌──────────────────────┐    ┌──────────────────────┐    ┌──────────────────────┐
│  1. BUSINESS         │    │  2. MACHINE          │    │  3. PRODUCTIVI-      │
│     ANALYTICS        │ →  │     LEARNING         │ →  │     ZACIÓN           │
│                      │    │                      │    │                      │
│  Datos → Análisis    │    │  Históricos →        │    │  Modelos →           │
│  → Insights          │    │  Modelos → Predicción│    │  Dashboards → Apps   │
└──────────────────────┘    └──────────────────────┘    └──────────────────────┘
```

| Fase | ¿Qué hace? | Ejemplo FACES |
|------|-----------|---------------|
| **Business Analytics** | Separar señal de ruido en los datos. Encontrar insights accionables | Analizar patrones de deserción, identificar los factores clave |
| **Machine Learning** | Usar datos históricos para predecir nuevos datos | Modelo que predice qué estudiantes desertarán el próximo semestre |
| **Productivización** | Poner los modelos en producción: dashboards, aplicaciones, automatización | Dashboard en tiempo real para la Dirección Académica |

### La metodología de Business Analytics

La fase de BA — donde vivirá la mayor parte de nuestro trabajo — sigue una metodología estructurada:

```
Objetivo,        Entidades      Preguntas      Análisis e      Comuni-       Business
palancas    →    y datos    →   semilla    →   insights    →   cación    →   Case
y KPIs
```

| Paso | ¿Qué haces? | Ejemplo: Deserción en FACES |
|------|------------|----------------------------|
| **Objetivo y KPIs** | Definir qué quieres lograr y cómo lo medirás | Reducir deserción del 15% al 10%. KPI: tasa semestral por carrera |
| **Entidades y datos** | Identificar qué entidades están involucradas y qué datos existen | Estudiantes, materias, docentes. Datos: notas, asistencia, socioeconómicos |
| **Preguntas semilla** | Formular las primeras preguntas a resolver | ¿Desertan más los que trabajan? ¿Influye el turno? ¿La carrera? |
| **Análisis e insights** | Analizar "tirando del hilo" hasta encontrar insights | Los que trabajan >30h y están en 3er semestre desertan 3x más |
| **Comunicación** | Presentar hallazgos de forma simple y gráfica | Informe visual para el Consejo de Facultad |
| **Business Case** | (Opcional) Estimar el impacto esperado de actuar sobre los insights | Un programa de tutorías focalizadas costaría $X y reduciría la deserción en Y% |

### ¿Dónde encaja nuestro programa en estas fases?

| Fase del proyecto | Módulos del programa | Nivel |
|:-:|:-:|:---|
| **Business Analytics** | M01 + M02 + M03 | Donde vivirá el 80% de su trabajo como docentes/investigadores |
| **Machine Learning** | M04 | Extensión para proyectos que requieran predicción |
| **Productivización** | Más allá del programa | Para quienes quieran desplegar soluciones automatizadas |

> **Para docentes de ciencias sociales:** La fase de Business Analytics es la más relevante. No todos los proyectos necesitan Machine Learning ni productivización. Un buen análisis exploratorio con visualizaciones claras puede transformar la toma de decisiones en su institución.

## 1d. El checklist del analista profesional

Así como un piloto usa una lista de verificación antes de despegar, un analista de datos profesional sigue un **checklist** en cada proyecto. No importa si el proyecto es grande o pequeño — estas verificaciones evitan errores costosos.

A continuación, las 6 verificaciones esenciales que todo analista tiene en mente:

### 1. Formular la pregunta correctamente

Antes de tocar un solo dato, verifica:

- [ ] ¿Especificaste el **tipo de pregunta**? (exploración, asociación, causalidad, predicción)
- [ ] ¿Definiste la **métrica de éxito** antes de comenzar?
- [ ] ¿Entendiste el **contexto** del problema y su aplicación?
- [ ] ¿Consideraste si la pregunta **puede responderse** con los datos disponibles?

### 2. Verificar y organizar los datos

Antes de analizar, verifica:

- [ ] ¿Graficaste resúmenes de los datos? (distribuciones, valores atípicos)
- [ ] ¿Identificaste los **datos faltantes** y entendiste por qué faltan?
- [ ] ¿Cada **variable** es una columna y cada **observación** es una fila? (*tidy data*)
- [ ] ¿Registraste los pasos para pasar de datos crudos a datos organizados?

### 3. Explorar antes de modelar

- [ ] ¿Realizaste gráficos univariados? (histogramas, boxplots)
- [ ] ¿Exploraste correlaciones entre variables? (gráficos de dispersión)
- [ ] ¿Verificaste las **unidades** de todos los datos?
- [ ] ¿Buscaste **errores** o codificación incorrecta?

### 4. Comunicar con rigor

- [ ] ¿Cada figura comunica una pieza **importante** de información?
- [ ] ¿Todas las figuras incluyen **etiquetas de ejes** comprensibles?
- [ ] ¿Explicaste el contexto y los datos **antes** de mostrar resultados?
- [ ] ¿Reportaste medidas de **incertidumbre** junto a cada estimación?

### 5. Sobre causalidad

- [ ] ¿Identificaste si el estudio fue **aleatorio** o observacional?
- [ ] ¿Consideraste posibles **factores de confusión**?
- [ ] Si no hay evidencia causal, ¿evitaste lenguaje que implique **causa y efecto**?

### 6. Reproducibilidad

- [ ] ¿Evitaste hacer cálculos **manualmente**?
- [ ] ¿Creaste un **script** que reproduzca todos los análisis?
- [ ] ¿Guardaste las versiones **crudas y procesadas** de los datos?
- [ ] ¿Alguien más podría ejecutar tu análisis y obtener los **mismos resultados**?

> **No necesitas saber hacer todo esto todavía.** En los módulos siguientes aprenderás las herramientas. Pero desde ahora, este checklist te da el **estándar profesional** contra el cual evaluar cualquier análisis de datos — propio o ajeno.

## 2. El ciclo en las carreras de FACES

El ciclo de valor de los datos se manifiesta de manera distinta en cada disciplina. Veamos ejemplos concretos:

### 2.1 Economía — Política pública

| Etapa | Aplicación |
|-------|------------|
| **Pregunta** | ¿Cuál es el impacto del salario mínimo en el empleo juvenil? |
| **Datos** | Series históricas de empleo (INE), encuestas de hogares, registros del IVSS |
| **Análisis** | Comparación antes/después de ajustes salariales, regresión con variables de control |
| **Hallazgo** | Ajustes superiores al 40% se asocian con reducción del empleo formal juvenil en 8% |
| **Acción** | Recomendar ajustes graduales con subsidios temporales al primer empleo |

### 2.2 Relaciones Industriales — Gestión de talento

| Etapa | Aplicación |
|-------|------------|
| **Pregunta** | ¿Qué factores predicen la rotación voluntaria en los primeros 6 meses? |
| **Datos** | Encuestas de clima, evaluaciones de desempeño, entrevistas de salida |
| **Análisis** | Análisis de frecuencia de motivos, correlación entre variables de satisfacción |
| **Hallazgo** | La falta de retroalimentación del supervisor inmediato es el predictor más fuerte |
| **Acción** | Programa de formación en feedback para gerentes de primera línea |

### 2.3 Administración — Análisis de mercado

| Etapa | Aplicación |
|-------|------------|
| **Pregunta** | ¿Cuál es el segmento más rentable para un nuevo producto? |
| **Datos** | Encuestas de intención de compra, datos de ventas históricas, análisis de competencia |
| **Análisis** | Segmentación por variables demográficas y comportamentales |
| **Hallazgo** | Profesionales 25-35 años en ciudades intermedias muestran mayor disposición a pagar |
| **Acción** | Focalizar lanzamiento en Barquisimeto, Valencia y Mérida |

### 2.4 Contaduría — Auditoría basada en datos

| Etapa | Aplicación |
|-------|------------|
| **Pregunta** | ¿Existen patrones anómalos en los gastos de representación de la empresa? |
| **Datos** | Registros contables, facturas digitalizadas, límites de póliza |
| **Análisis** | Distribución de montos por categoría, detección de valores atípicos (Ley de Benford) |
| **Hallazgo** | 12% de las facturas de viáticos se concentran justo debajo del límite de aprobación automática |
| **Acción** | Auditoría focalizada en ese segmento + revisión de controles internos |

## 3. La pregunta correcta es la mitad del análisis

Una mala pregunta produce datos irrelevantes, análisis costosos y decisiones equivocadas.

### Criterios para una buena pregunta analítica

- **Específica**: no "¿cómo está la economía?" sino "¿cómo varió el PIB per cápita en los últimos 5 años?"
- **Medible**: debe existir (o poder generarse) data que la responda
- **Relevante**: debe conectar con una decisión concreta
- **Acotada**: en tiempo, espacio y población

### Preguntas débiles vs. fuertes

| Débil ❌ | Fuerte ✅ |
|----------|----------|
| ¿Los estudiantes están satisfechos? | ¿Cuáles son los 3 factores de mayor insatisfacción entre estudiantes de 3er año en FACES? |
| ¿La inflación afecta a las empresas? | ¿Cómo cambió el margen operativo de PYMEs manufactureras ante inflación >50% anual? |
| ¿El clima laboral es bueno? | ¿Qué dimensiones del clima laboral se correlacionan con rotación en el sector bancario? |

## 4b. Cuando el ciclo falla: un anti-ejemplo

No seguir el ciclo completo produce decisiones costosas. Veamos un caso real:

### El caso del programa de laptops en FACES

| Etapa | Lo que ocurrió | El problema |
|-------|---------------|-------------|
| **Pregunta** | "¿Cómo mejorar el rendimiento académico?" | Pregunta demasiado vaga |
| **Datos** | Se saltó esta etapa | No se recopiló información sobre las causas reales del bajo rendimiento |
| **Análisis** | Se saltó esta etapa | Sin datos, no hubo análisis |
| **Hallazgo** | Se asumió sin evidencia: "los estudiantes necesitan tecnología" | Supuesto no validado |
| **Acción** | Se invirtieron $150,000 en laptops para estudiantes | Inversión basada en intuición |

**Resultado:** El rendimiento no mejoró. Un análisis posterior reveló que los factores principales eran la carga laboral (72% trabaja >20h) y la falta de tutorías, no la falta de tecnología.

### Lecciones
- **Saltar etapas es costoso**: la inversión se habría orientado mejor con un análisis previo
- **La intuición no sustituye a los datos**: "más tecnología = mejor rendimiento" suena razonable pero no era cierto en este contexto
- **El ciclo protege contra decisiones precipitadas**: cada etapa es un checkpoint de calidad

## 4c. Respuestas sugeridas al ejercicio

> **Nota:** No hay una única respuesta correcta. Estas son orientaciones para autoevaluación.

| Etapa | Ejemplo de respuesta adecuada |
|-------|------------------------------|
| **Pregunta** | ¿Qué factores distinguen a los becarios que mantienen el promedio de los que no lo mantienen? |
| **Datos** | Registros académicos (promedios por semestre), datos socioeconómicos de la solicitud de beca, registro de asistencia a tutorías, horas de trabajo |
| **Análisis** | Comparar el perfil de becarios que mantienen el promedio vs. los que no (análisis exploratorio + inferencial) |
| **Hallazgo** | Por ejemplo: los becarios que trabajan >25h/semana y no asisten a tutorías tienen 4x más probabilidad de perder la beca |
| **Acción** | Implementar tutorías obligatorias para becarios + evaluar si los criterios de renovación consideran la carga laboral |

**Criterios de una buena respuesta:**
- La pregunta es específica (no "¿por qué reprueban?")
- Los datos son obtenibles (no pide información imposible)
- El análisis corresponde a la pregunta
- El hallazgo es falsificable (podría resultar diferente)
- La acción se conecta directamente con el hallazgo
