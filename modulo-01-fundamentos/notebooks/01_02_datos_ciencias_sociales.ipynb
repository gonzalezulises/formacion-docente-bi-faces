## 1. Fuentes de datos en ciencias sociales

A diferencia de las ciencias naturales, donde los datos provienen de instrumentos de medición controlados, en ciencias sociales las fuentes son diversas y cada una tiene sus particularidades.

### 1.1 Datos estructurados

Organizados en filas y columnas, con formato predecible:

- **Encuestas y cuestionarios**: la fuente más clásica en ciencias sociales. Escalas Likert, preguntas cerradas, datos demográficos.
- **Registros administrativos**: matrícula universitaria, nóminas, declaraciones fiscales, registros del IVSS.
- **Bases de datos públicas**: Banco Mundial, CEPAL, INE, BCV — series históricas de indicadores.
- **Datos transaccionales**: ventas, movimientos bancarios, registros contables.

### 1.2 Datos no estructurados

Sin formato tabular predefinido, requieren procesamiento previo:

- **Entrevistas y grupos focales**: transcripciones de texto que requieren codificación.
- **Redes sociales**: publicaciones, comentarios, reacciones — texto libre con metadatos.
- **Documentos institucionales**: actas, informes, normativas — texto con estructura variable.
- **Medios de comunicación**: artículos de prensa, discursos — para análisis de contenido.

### 1.3 Datos mixtos

La realidad de la investigación social suele combinar ambos:

| Proyecto | Datos estructurados | Datos no estructurados |
|----------|--------------------|-----------------------|
| Estudio de clima laboral | Encuesta con escala Likert | Entrevistas a profundidad |
| Análisis de política fiscal | Series del BCV | Discursos del ministro |
| Evaluación docente | Calificaciones numéricas | Comentarios abiertos de estudiantes |

## 2. Tipos de variables

### 2.1 Variables cualitativas (categóricas)

Representan categorías o grupos, no cantidades.

#### Nominales
- No tienen orden inherente
- Ejemplos: carrera (Economía, Administración, Contaduría, RRII), estado civil, género, sector económico
- Operaciones válidas: frecuencia, moda, chi-cuadrado

#### Ordinales
- Tienen un orden natural pero las distancias entre categorías no son iguales
- Ejemplos: nivel educativo (primaria < secundaria < universitario), escala Likert (muy insatisfecho → muy satisfecho), nivel de riesgo (bajo, medio, alto)
- Operaciones válidas: mediana, percentiles, correlación de Spearman

### 2.2 Variables cuantitativas (numéricas)

Representan cantidades medibles.

#### De intervalo
- Distancias iguales entre valores pero **sin cero absoluto**
- Ejemplos: temperatura en °C, puntuación en un test estandarizado, año calendario
- Operaciones válidas: media, desviación estándar, correlación de Pearson
- **No se puede decir** "el doble" (20°C no es el doble de calor que 10°C)

#### De razón
- Distancias iguales **con cero absoluto**
- Ejemplos: salario, edad, matrícula estudiantil, PIB, tasa de desempleo
- Operaciones válidas: todas las anteriores + proporciones y tasas
- **Sí se puede decir** "el doble" (un salario de $2000 es el doble de $1000)

## 3. Niveles de medición y sus implicaciones

El nivel de medición de una variable **determina qué análisis son válidos**. Usar una técnica inadecuada produce resultados sin sentido.

| Nivel | Ejemplo | Media | Mediana | Moda | Comparar proporciones | Regresión |
|-------|---------|:-----:|:-------:|:----:|:--------------------:|:---------:|
| Nominal | Carrera | ❌ | ❌ | ✅ | ✅ | Solo como dummy |
| Ordinal | Satisfacción | ⚠️ | ✅ | ✅ | ✅ | Con precaución |
| Intervalo | Puntaje test | ✅ | ✅ | ✅ | ✅ | ✅ |
| Razón | Salario | ✅ | ✅ | ✅ | ✅ | ✅ |

### El debate de las escalas Likert

¿Es ordinal o de intervalo? Este es uno de los debates más frecuentes en ciencias sociales:

- **Postura conservadora**: es ordinal, la distancia entre "de acuerdo" y "muy de acuerdo" no es igual a la distancia entre "neutral" e "indiferente".
- **Postura pragmática**: con 5+ puntos y distribución razonable, se puede tratar como intervalo para efectos prácticos.
- **Recomendación**: reportar ambos análisis (paramétricos y no paramétricos) cuando sea posible.
