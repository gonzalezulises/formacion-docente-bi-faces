## 1. ¿Qué es la ciencia de datos?

La ciencia de datos es un campo interdisciplinario que combina **conocimiento del dominio**, **habilidades de programación** y **estadística/matemáticas** para extraer conocimiento útil de los datos.

```
                    Conocimiento
                    del dominio
                   (su experticia)
                        ╱╲
                       ╱  ╲
                      ╱    ╲
                     ╱ DATA ╲
                    ╱SCIENCE ╲
                   ╱          ╲
        Programación ────────── Estadística
        & tecnología              & matemáticas
```

### ¿Por qué ahora?

Tres fuerzas convergieron en la última década:

| Factor | Antes | Ahora |
|--------|-------|-------|
| **Datos** | Escasos, costosos de recolectar | Abundantes: sensores, redes sociales, registros digitales |
| **Computación** | Mainframes caros, acceso limitado | Laptops potentes, cloud computing, Google Colab gratis |
| **Algoritmos** | Estadística clásica | Machine Learning, Deep Learning, IA generativa |

> **Lo que NO cambió:** la necesidad de pensar críticamente, formular buenas preguntas y entender el contexto. Eso es exactamente lo que ustedes aportan como científicos sociales.

## 3. Las diferencias que importan

### Estadística vs. Ciencia de Datos

| Dimensión | Estadística | Ciencia de Datos |
|-----------|------------|------------------|
| **Origen** | Matemáticas, siglo XVIII | Interdisciplinario, siglo XXI |
| **Enfoque** | Inferencia: ¿qué puedo concluir de esta muestra? | Predicción + acción: ¿qué va a pasar y qué hago? |
| **Datos típicos** | Muestras pequeñas, diseño experimental | Datos masivos, observacionales, no estructurados |
| **Herramientas** | R, SAS, SPSS | Python, SQL, herramientas de visualización |
| **Producto** | Paper académico, informe técnico | Dashboard, modelo en producción, decisión de negocio |

> **No son rivales, son complementarias.** La estadística provee el rigor teórico; la ciencia de datos agrega escala, automatización y orientación a la acción.

### Business Analytics vs. Business Intelligence

| Dimensión | Business Analytics (BA) | Business Intelligence (BI) |
|-----------|------------------------|---------------------------|
| **Pregunta** | ¿Por qué pasó? ¿Qué pasará? | ¿Qué está pasando ahora? |
| **Enfoque** | Análisis exploratorio, modelado, insights | Monitoreo, reportes, alertas |
| **Temporalidad** | Proyectos puntuales (análisis ad-hoc) | Continuo (dashboards actualizados) |
| **Perfil** | Analista que investiga y descubre | Desarrollador BI que construye pipelines y visualizaciones |
| **Producto** | Informe con recomendaciones | Dashboard interactivo, reporte automatizado |

### Business Analytics vs. Ciencia de Datos

| Dimensión | Business Analytics | Ciencia de Datos |
|-----------|-------------------|------------------|
| **Alcance** | Análisis para decisiones de negocio | Incluye BA + ML + productivización |
| **Complejidad técnica** | Media: SQL, Excel avanzado, visualización | Alta: programación, algoritmos, infraestructura |
| **Modelos** | Descriptivos e inferenciales | Predictivos y prescriptivos |
| **Automatización** | Baja: análisis humano | Alta: modelos en producción, pipelines |

> **En la práctica:** BA es el punto de partida de todo proyecto de ciencia de datos. No se puede construir un modelo predictivo sin antes entender los datos (BA). Muchos problemas se resuelven con BA sin necesidad de ML.

## 4. ¿Dónde se ubican ustedes?

Como docentes e investigadores de ciencias sociales, su perfil tiene una **ventaja enorme** y una **brecha que cerrar**:

### Lo que ya tienen (y es difícil de enseñar)

- **Conocimiento profundo del dominio**: entienden la economía, las organizaciones, la contabilidad, las relaciones laborales
- **Pensamiento crítico**: saben cuestionar supuestos, identificar sesgos, evaluar evidencia
- **Capacidad de formular preguntas**: años de investigación los entrenaron para esto
- **Contexto institucional**: conocen FACES, sus estudiantes, sus desafíos

### Lo que este programa les dará (y es más fácil de enseñar)

- **Herramientas técnicas**: Python, SQL, Pandas, visualización
- **Flujo de trabajo profesional**: Import → Tidy → Transform → Visualize → Model → Communicate
- **Capacidad de implementar**: pasar de "tengo una pregunta" a "tengo un análisis reproducible"

### El mapa de roles en datos

| Rol | ¿Qué hace? | ¿Les aplica? |
|-----|-----------|:---:|
| **Analista de datos** | Limpia, explora y visualiza datos para responder preguntas | Sí — este es el perfil más cercano |
| **Analista de negocio (BA)** | Traduce problemas del dominio a preguntas de datos | Sí — su conocimiento del dominio es clave |
| **Desarrollador BI** | Construye dashboards y pipelines de datos | Parcialmente — Módulo 3 |
| **Científico de datos** | Construye modelos predictivos y de ML | Introducción — Módulo 4 |
| **Ingeniero de datos** | Diseña infraestructura de datos a escala | No — fuera del alcance |
| **Ingeniero de ML** | Despliega modelos en producción | No — fuera del alcance |

> **El objetivo de este programa no es convertirlos en científicos de datos.** Es darles las herramientas para ser **docentes e investigadores que trabajan con datos de forma profesional** — y que pueden enseñar a sus estudiantes a hacerlo.

## 6. Ciencia de datos en ciencias sociales: no es nuevo, es una evolución

La idea de usar datos para entender fenómenos sociales no nació con el Big Data. Tiene una historia larga:

| Época | Hito | Disciplina |
|-------|------|------------|
| 1662 | John Graunt publica *Natural and Political Observations* — primer análisis demográfico con datos de mortalidad | Demografía |
| 1800s | Quetelet aplica estadística a fenómenos sociales ("física social") | Sociología |
| 1900s | Encuestas de hogares, censos modernos, econometría | Economía, sociología |
| 1960s | SPSS democratiza el análisis estadístico en ciencias sociales | Todas |
| 2000s | Datos administrativos digitales, registros de internet | Economía, ciencia política |
| 2010s | Big Data, ML aplicado a problemas sociales, text mining | Ciencia de datos social |
| 2020s | IA generativa, datos satelitales para economía, NLP para análisis de políticas | Ciencia de datos + IA |

### Lo que cambia con la ciencia de datos moderna

| Antes (investigación social tradicional) | Ahora (ciencia de datos social) |
|:---|:---|
| Diseño una encuesta y recolecto datos | También uso datos que ya existen (registros, web, satélites) |
| Analizo con SPSS o Stata | También uso Python, R, SQL |
| Publico un paper | También creo dashboards, reportes interactivos |
| Muestra de 200 personas | También trabajo con registros completos de miles o millones |
| Análisis inferencial (p-values) | También uso predicción, clasificación, clustering |
| Trabajo solo o con 2-3 colegas | También colaboro en equipos interdisciplinarios con ingenieros y diseñadores |

> **No se trata de reemplazar lo que saben, sino de expandirlo.** Su formación en investigación social es la base. La ciencia de datos agrega herramientas y escala.
