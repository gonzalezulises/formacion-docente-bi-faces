## 1. El Problema del Dato Ausente

> *"Lo que no está en los datos puede ser tan importante como lo que está."*

### 1.1 ¿Por qué faltan datos?

Los valores faltantes son inevitables en datos del mundo real. Ocurren por diversas razones:

| Causa | Ejemplo | Implicación |
|-------|---------|-------------|
| **No respuesta** | Encuestado no quiere revelar ingreso | Puede ser sistemático |
| **No aplica** | Estado civil de un menor | Es información válida |
| **Error de registro** | Campo dejado en blanco por accidente | Generalmente aleatorio |
| **Pérdida de datos** | Archivo corrupto | Puede afectar períodos específicos |
| **Diseño del estudio** | Variable agregada después de iniciar | Faltantes estructurales |

### 1.2 El peligro de ignorarlos

Ignorar valores faltantes puede:

1. **Sesgar resultados**: Si los faltantes no son aleatorios
2. **Reducir poder estadístico**: Menos observaciones = menos precisión
3. **Invalidar conclusiones**: Análisis incompleto puede ser engañoso
4. **Ocultar problemas de calidad**: Los faltantes son síntoma de procesos rotos

## 2. Taxonomía de Valores Faltantes

Donald Rubin (1976) clasificó los valores faltantes en tres categorías fundamentales:

### 2.1 MCAR: Missing Completely At Random

**Definición**: La probabilidad de que un dato falte es completamente aleatoria, independiente de cualquier variable.

**Ejemplo**: Un asistente de investigación derrama café sobre algunas encuestas al azar.

**Características**:
- Los datos faltantes son una muestra aleatoria de todos los datos
- No hay sesgo al eliminar observaciones incompletas
- El análisis con datos completos es válido (pero menos eficiente)

**Prueba informal**: ¿Los casos con datos faltantes son similares a los casos completos en todas las demás variables?

### 2.2 MAR: Missing At Random

**Definición**: La probabilidad de que un dato falte depende de otras variables observadas, pero NO del valor faltante mismo.

**Ejemplo**: Los estudiantes que trabajan tienen mayor probabilidad de no reportar su ingreso, pero entre los que trabajan, la probabilidad es igual independientemente de cuánto ganan.

**Características**:
- El patrón de faltantes es predecible por otras variables
- Eliminar observaciones incompletas puede sesgar
- La imputación condicionada en otras variables puede funcionar

### 2.3 MNAR: Missing Not At Random

**Definición**: La probabilidad de que un dato falte depende del valor faltante mismo.

**Ejemplo**: Personas con ingresos muy altos o muy bajos prefieren no reportar su ingreso.

**Características**:
- El más problemático: no podemos corregir sin supuestos fuertes
- Ninguna técnica de imputación estándar funciona bien
- Requiere modelar el mecanismo de faltantes
- Frecuente en datos sensibles (ingreso, salud, comportamiento)

## 4. Estrategias de Manejo

### 4.1 Estrategias Comunes (y sus problemas)

| Estrategia | Descripción | Cuándo usar | Problemas |
|------------|-------------|-------------|----------|
| **Eliminar filas** | Usar solo casos completos | MCAR, pocos faltantes | Pérdida de información, posible sesgo |
| **Eliminar columna** | Excluir variable con muchos faltantes | Variable no crítica | Pérdida de información relevante |
| **Imputar con media** | Reemplazar con promedio | Variable numérica, MCAR | Reduce varianza, sesga relaciones |
| **Imputar con mediana** | Reemplazar con mediana | Distribución sesgada | Similar a media |
| **Imputar con moda** | Reemplazar con valor más frecuente | Variables categóricas | Puede sesgar proporciones |
| **Imputación múltiple** | Crear varios datasets imputados | Análisis riguroso | Complejidad técnica |
| **Indicador de faltante** | Crear variable dummy "faltante" | El faltante es informativo | No resuelve el valor |

### 4.2 La Regla de Oro

> **No existe una "solución correcta" universal. Cada decisión tiene implicaciones que deben documentarse y justificarse.**

### 4.3 Preguntas Guía para Decidir

1. **¿Cuántos faltantes hay?**
   - < 5%: Generalmente manejable
   - 5-20%: Requiere cuidado
   - > 20%: Cuestionar si la variable es usable

2. **¿El patrón es aleatorio?**
   - Comparar casos completos vs incompletos en otras variables

3. **¿Cuál es el costo de cada error?**
   - Eliminar: Perder información
   - Imputar incorrectamente: Introducir sesgo

4. **¿Qué análisis haré después?**
   - Algunos métodos (ej: Random Forest) manejan faltantes
   - Otros (ej: regresión lineal) no los toleran

## 8. Resumen y Puntos Clave

### Conceptos Fundamentales

1. **Los valores faltantes NO son neutrales**: Su patrón contiene información

2. **Tres tipos** (Rubin, 1976):
   - MCAR: Completamente aleatorio → Menos problemático
   - MAR: Depende de otras variables → Manejable
   - MNAR: Depende del valor faltante → Problemático

3. **No hay solución universal**: Cada estrategia tiene trade-offs

4. **Transparencia es obligatoria**: Documentar decisiones y justificarlas
