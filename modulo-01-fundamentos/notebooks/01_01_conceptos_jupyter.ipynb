## 📖 Sección 1: ¿Qué es un Jupyter Notebook?

Un **Jupyter Notebook** es un documento interactivo que combina:

1. **Texto formateado** (como el que estás leyendo ahora)
2. **Código ejecutable** (Python, R, u otros lenguajes)
3. **Resultados** (números, tablas, gráficos)

Todo convive en un mismo archivo con extensión `.ipynb`.

### ¿Por qué se llama "Jupyter"?

El nombre viene de tres lenguajes de programación: **Ju**lia, **Py**thon y **R**. Sin embargo, en este programa usaremos exclusivamente Python.

### ¿Por qué son revolucionarios?

| Herramienta tradicional | Limitación | Ventaja de Jupyter |
|------------------------|-----------|--------------------|
| Word / Google Docs | No ejecuta código | Código y texto en un solo lugar |
| Excel | Difícil documentar el proceso | Cada paso queda explicado y es reproducible |
| Scripts .py | Solo código, sin explicaciones ricas | Combina narrativa con código |
| PowerPoint | Estático, no interactivo | Los resultados se actualizan al cambiar los datos |

### ¿Dónde se usan?

- **Investigación académica**: El artículo que detectó las ondas gravitacionales fue publicado con un Jupyter Notebook
- **Ciencia de datos**: Es la herramienta estándar en la industria
- **Docencia universitaria**: Universidades como MIT, Stanford y Harvard los usan en sus cursos
- **Análisis institucional**: Reportes de matrícula, rendimiento académico, presupuesto

## 📝 Sección 2: Tipos de celdas

Un notebook está compuesto por **celdas**. Existen dos tipos principales:

### Celdas de Markdown (texto)
- Contienen texto con formato: títulos, listas, tablas, fórmulas
- Se usan para explicar, documentar, dar instrucciones
- **Esta celda que estás leyendo es una celda de Markdown**
- Para editarla: haz doble clic sobre ella
- Para renderizarla: presiona `Shift + Enter`

### Celdas de Código
- Contienen código Python que se puede ejecutar
- Tienen un indicador `[ ]` a la izquierda
- Al ejecutarlas, el resultado aparece justo debajo

Veamos un ejemplo de cada tipo. La celda de abajo es una **celda de código**:

````python
# Esto es una celda de código
# Las líneas que comienzan con # son comentarios (no se ejecutan)

print("Hola FACES! Bienvenidos al mundo de Jupyter Notebooks")
````

````python
# Python puede hacer cálculos directamente
2 + 2
````

````python
# También operaciones más complejas
promedio = (15 + 12 + 18 + 14 + 16) / 5
print(f"El promedio de calificaciones es: {promedio}")
````

### ¿Cómo saber qué tipo de celda es?

- Las celdas de **código** tienen un borde y un indicador `[ ]` o `[número]` a la izquierda
- Las celdas de **Markdown** se ven como texto normal después de ejecutarlas
- En la barra de herramientas, un menú desplegable muestra el tipo de la celda seleccionada

### Cambiar el tipo de celda

- Selecciona la celda (clic una vez, sin editar)
- Presiona **M** para convertirla en Markdown
- Presiona **Y** para convertirla en Código

## ▶️ Sección 3: Ejecutar celdas

### Formas de ejecutar una celda

| Acción | Atajo | Efecto |
|--------|-------|--------|
| Ejecutar y avanzar | `Shift + Enter` | Ejecuta la celda actual y pasa a la siguiente |
| Ejecutar sin avanzar | `Ctrl + Enter` | Ejecuta la celda actual y se queda en ella |
| Ejecutar e insertar | `Alt + Enter` | Ejecuta la celda actual e inserta una nueva debajo |

### El indicador de ejecución

- `[ ]` — La celda no se ha ejecutado aún
- `[*]` — La celda se está ejecutando ahora mismo
- `[1]` — La celda fue ejecutada (el número indica el orden)

### El concepto de Kernel

El **kernel** es el "motor" que ejecuta tu código Python. Piensa en él como un asistente que:

- **Recuerda** todas las variables que has creado
- **Ejecuta** las celdas cuando tú se lo pides
- **Se puede reiniciar** si algo sale mal

### ¡Cuidado con el orden de ejecución!

El kernel ejecuta las celdas en el orden en que **tú** las ejecutas, no necesariamente de arriba hacia abajo. Esto puede causar confusión. Veamos un ejemplo:

````python
# Ejecuta PRIMERO esta celda
mensaje = "Hola desde FACES"
print(mensaje)
````

````python
# Ahora ejecuta esta celda
# Funciona porque la variable 'mensaje' ya fue creada arriba
print(mensaje + " - Universidad Catolica Andres Bello (UCAB)")
````

````python
# ¿Qué pasa si ejecutas esta celda SIN haber ejecutado las anteriores?
# (Si reiniciaste el kernel, prueba ejecutar SOLO esta celda)
# Obtendrás un error: NameError: name 'mensaje' is not defined

# Esto es normal. Significa que el kernel no conoce la variable 'mensaje'
# porque no has ejecutado la celda que la crea.

print("El orden de ejecución importa.")
print(f"La variable 'mensaje' contiene: {mensaje}")
````

### Reiniciar el Kernel

Si algo sale mal o quieres empezar de cero:

- **Menu > Kernel > Restart** — Reinicia el kernel (borra todas las variables)
- **Menu > Kernel > Restart & Run All** — Reinicia y ejecuta todas las celdas de arriba a abajo

En Google Colab:
- **Menu > Entorno de ejecución > Reiniciar entorno de ejecución**
- **Menu > Entorno de ejecución > Ejecutar todas**

## ⌨️ Sección 4: Atajos de teclado esenciales

Los atajos de teclado te permiten trabajar mucho más rápido. Hay dos **modos**:

### Modo de Comando (celda seleccionada pero NO editando)

| `Z` | Deshacer eliminación de celda |

| Atajo | Acción |

| `Ctrl + /` | Comentar/descomentar línea |
| `Ctrl + Z` | Deshacer |
| `Ctrl + Shift + Z` | Rehacer |
| `Esc` | Salir al modo de comando |

## 🎨 Sección 5: Markdown esencial

**Markdown** es un lenguaje de formato de texto simple pero poderoso. Es lo que usamos en las celdas de texto de Jupyter para crear documentos bien formateados.

### Títulos

Se crean con el símbolo `#`. Más `#` = título más pequeño:

```markdown
# Título 1 (el más grande)
## Título 2
### Título 3
#### Título 4
```

### Formato de texto

```markdown
**texto en negrita**
*texto en cursiva*
***negrita y cursiva***
`código en línea`
~~texto tachado~~
```

Resultado: **texto en negrita**, *texto en cursiva*, ***negrita y cursiva***, `código en línea`, ~~texto tachado~~

### Listas

**Lista sin orden:**
```markdown
- Primer elemento
- Segundo elemento
  - Sub-elemento
- Tercer elemento
```

Resultado:
- Primer elemento
- Segundo elemento
  - Sub-elemento
- Tercer elemento

**Lista numerada:**
```markdown
1. Primer paso
2. Segundo paso
3. Tercer paso
```

Resultado:
1. Primer paso
2. Segundo paso
3. Tercer paso

### Tablas

```markdown
| Carrera | Estudiantes | Sede |
|---------|-------------|------|
| Economía | 450 | Bárbula |
| Administración | 680 | Bárbula |
| Contaduría | 520 | La Morita |
```

Resultado:

| Carrera | Estudiantes | Sede |
|---------|-------------|------|
| Economía | 450 | Bárbula |
| Administración | 680 | Bárbula |
| Contaduría | 520 | La Morita |

### Enlaces e imágenes

```markdown
[Texto del enlace](https://www.uc.edu.ve)
![Texto alternativo](url_de_la_imagen.png)
```

### Fórmulas matemáticas (LaTeX)

Jupyter soporta fórmulas matemáticas usando la sintaxis LaTeX:

**En línea:** `$E = mc^2$` se ve como $E = mc^2$

**En bloque:**
```markdown
$$\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i$$
```

Se ve como:

$$\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i$$

Esto es especialmente útil para materias de Estadística, Econometría y Matemáticas.

## ✅ Sección 6: Buenas prácticas

Para que tus notebooks sean útiles y profesionales, sigue estas recomendaciones:

### 1. Usa un título claro
Siempre comienza con una celda Markdown que tenga:
- Título descriptivo
- Autor y fecha
- Breve descripción del contenido

### 2. Alterna texto y código
No pongas todo el código junto. Usa celdas de Markdown para explicar:
- **Qué** hace cada sección
- **Por qué** se hace de esa manera
- **Cómo** interpretar los resultados

### 3. Comenta tu código
Dentro de las celdas de código, usa `#` para agregar comentarios:

```python
# Calcular el promedio de calificaciones
promedio = sum(calificaciones) / len(calificaciones)
```

### 4. Reinicia y ejecuta todo antes de compartir
Antes de enviar tu notebook a alguien:
1. **Kernel > Restart & Run All** (o Entorno de ejecución > Ejecutar todas)
2. Verifica que todo se ejecuta sin errores
3. Esto garantiza que cualquier persona pueda reproducir tu trabajo

### 5. Guarda frecuentemente
- `Ctrl + S` para guardar
- Google Colab guarda automáticamente, pero puedes forzar con `Ctrl + S`

### 6. Usa nombres descriptivos para los archivos
- **Bien:** `analisis_matricula_2024.ipynb`
- **Mal:** `Untitled.ipynb` o `copia (2).ipynb`

````python
# Ejemplo de celda de código bien documentada

# Datos de ejemplo: calificaciones de una sección de Estadística I
calificaciones = [15, 12, 18, 9, 14, 7, 16, 11, 13, 10]

# Calcular estadísticas básicas
total_estudiantes = len(calificaciones)
promedio = sum(calificaciones) / total_estudiantes
nota_maxima = max(calificaciones)
nota_minima = min(calificaciones)

# Contar aprobados (nota >= 10 en escala de 0-20)
aprobados = sum(1 for nota in calificaciones if nota >= 10)
tasa_aprobacion = aprobados / total_estudiantes * 100

# Mostrar resultados
print("=" * 40)
print("RESUMEN DE CALIFICACIONES")
print("=" * 40)
print(f"Total de estudiantes: {total_estudiantes}")
print(f"Promedio:            {promedio:.1f}")
print(f"Nota máxima:         {nota_maxima}")
print(f"Nota mínima:         {nota_minima}")
print(f"Aprobados:           {aprobados} ({tasa_aprobacion:.1f}%)")
````
