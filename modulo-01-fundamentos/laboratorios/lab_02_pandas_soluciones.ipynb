## 📦 Configuración Inicial

```python
import pandas as pd

# Configurar Pandas para mostrar más filas y columnas
pd.set_option('display.max_columns', 15)
pd.set_option('display.max_rows', 20)
pd.set_option('display.width', 120)

print(f"Pandas versión: {pd.__version__}")
print("Configuración lista")
```

## ✅ Ejercicio 1: Cargar y explorar el dataset de matrícula

```python
# Ejercicio 1: Cargar y explorar matricula_faces.csv

# 1. Cargar el archivo CSV
matricula = pd.read_csv('../../datasets/universidad/matricula_faces.csv')
print("Archivo cargado exitosamente")

# 2. Mostrar las primeras 10 filas
print("\nPrimeras 10 filas:")
print(matricula.head(10))

# 3. Dimensiones del DataFrame
print(f"\nDimensiones: {matricula.shape[0]:,} filas x {matricula.shape[1]} columnas")

# 4. Tipos de datos
print("\nTipos de datos:")
print(matricula.dtypes)

# 5. Resumen con info()
print("\nResumen:")
matricula.info()
```

---

## ✅ Ejercicio 2: Valores únicos

```python
# Ejercicio 2: Valores únicos

# 1. Carreras únicas
print(f"Número de carreras únicas: {matricula['carrera'].nunique()}")
print(f"Carreras: {list(matricula['carrera'].unique())}")

# 2. Períodos académicos
periodos = sorted(matricula['periodo'].unique())
print(f"\nNúmero de períodos: {matricula['periodo'].nunique()}")
print(f"Primer período: {periodos[0]}")
print(f"Último período: {periodos[-1]}")

# 3. Sedes
print(f"\nNúmero de sedes: {matricula['sede'].nunique()}")
print(f"Sedes: {list(matricula['sede'].unique())}")

# 4. Tipos de ingreso
print(f"\nNúmero de tipos de ingreso: {matricula['tipo_ingreso'].nunique()}")
print(f"Tipos: {list(matricula['tipo_ingreso'].unique())}")
print("\nConteo por tipo de ingreso:")
print(matricula['tipo_ingreso'].value_counts())
```

---

## ✅ Ejercicio 3: Filtrado - Estudiantes de Administración activos

```python
# Ejercicio 3: Filtrar Administración activos

# 1. Filtrar: carrera = Administracion Y estatus = Activo
admin_activos = matricula[
    (matricula['carrera'] == 'Administracion') & 
    (matricula['estatus'] == 'Activo')
]

# 2. Cantidad de registros
print(f"Estudiantes de Administración activos: {len(admin_activos):,}")
total_admin = len(matricula[matricula['carrera'] == 'Administracion'])
print(f"Del total de Administración: {total_admin:,}")
print(f"Porcentaje activos: {len(admin_activos)/total_admin*100:.1f}%")

# 3. Distribución por género
print(f"\nDistribución por género:")
genero_counts = admin_activos['genero'].value_counts()
for genero, conteo in genero_counts.items():
    etiqueta = 'Femenino' if genero == 'F' else 'Masculino'
    pct = conteo / len(admin_activos) * 100
    print(f"  {etiqueta}: {conteo:,} ({pct:.1f}%)")

# 4. Distribución por sede
print(f"\nDistribución por sede:")
print(admin_activos['sede'].value_counts())

# 5. Edad promedio
print(f"\nEdad promedio: {admin_activos['edad'].mean():.1f} años")
```

---

## ✅ Ejercicio 4: GroupBy - Estudiantes por sede y estatus

```python
# Ejercicio 4: GroupBy por sede y estatus

# 1. Agrupar por sede y estatus
print("Estudiantes por sede y estatus:")
print("=" * 45)
grupo_sede_estatus = matricula.groupby(['sede', 'estatus']).size().reset_index(name='cantidad')
print(grupo_sede_estatus)

# 2. Sede con más estudiantes activos
activos_por_sede = matricula[matricula['estatus'] == 'Activo'].groupby('sede').size()
print(f"\nSede con más estudiantes activos: {activos_por_sede.idxmax()} ({activos_por_sede.max():,})")

# 3. Tasa de retiro por sede
print("\nTasa de retiro por sede:")
for sede in matricula['sede'].unique():
    datos_sede = matricula[matricula['sede'] == sede]
    total_sede = len(datos_sede)
    retirados_sede = len(datos_sede[datos_sede['estatus'] == 'Retirado'])
    tasa = retirados_sede / total_sede * 100
    print(f"  {sede}: {tasa:.1f}% ({retirados_sede:,} de {total_sede:,})")

# 4. Tabla cruzada
print("\nTabla cruzada: sede por estatus")
print("=" * 45)
tabla = pd.crosstab(matricula['sede'], matricula['estatus'], margins=True, margins_name='Total')
print(tabla)
```

---

## ✅ Ejercicio 5: Cargar y explorar rendimiento académico

```python
# Ejercicio 5: Rendimiento académico - promedio por carrera

# 1. Cargar el archivo
rendimiento = pd.read_csv('../../datasets/universidad/rendimiento_academico.csv')
print(f"Archivo cargado: {rendimiento.shape[0]:,} filas x {rendimiento.shape[1]} columnas")

# 2. Primeras filas y estructura
print("\nPrimeras 5 filas:")
print(rendimiento.head())
print("\nColumnas y tipos:")
print(rendimiento.dtypes)

# 3. Promedio agrupado por carrera
print("\nPromedio académico por carrera:")
print("=" * 40)
promedio_carrera = rendimiento.groupby('carrera')['promedio'].mean().round(2).sort_values(ascending=False)
print(promedio_carrera)

# 4. Carrera con mayor y menor promedio
print(f"\nCarrera con mayor promedio: {promedio_carrera.idxmax()} ({promedio_carrera.max()})")
print(f"Carrera con menor promedio: {promedio_carrera.idxmin()} ({promedio_carrera.min()})")
```

---

## ✅ Ejercicio 6: Filtrado avanzado - Estudiantes destacados

```python
# Ejercicio 6: Estudiantes con promedio >= 15

# 1. Filtrar estudiantes con promedio >= 15
destacados = rendimiento[rendimiento['promedio'] >= 15]

# 2. Total
print(f"Estudiantes destacados (promedio >= 15): {len(destacados):,}")
print(f"Del total de {len(rendimiento):,} ({len(destacados)/len(rendimiento)*100:.1f}%)")

# 3. Distribución por género
print(f"\nDistribución por género:")
genero_dest = destacados['genero'].value_counts()
for genero, conteo in genero_dest.items():
    etiqueta = 'Femenino' if genero == 'F' else 'Masculino'
    pct = conteo / len(destacados) * 100
    print(f"  {etiqueta}: {conteo} ({pct:.1f}%)")

# 4. Distribución por carrera
print(f"\nDistribución por carrera:")
carrera_dest = destacados['carrera'].value_counts()
for carrera, conteo in carrera_dest.items():
    # Calcular qué porcentaje de la carrera son destacados
    total_carrera = len(rendimiento[rendimiento['carrera'] == carrera])
    pct_carrera = conteo / total_carrera * 100
    print(f"  {carrera:25s}: {conteo:3d} ({pct_carrera:.1f}% de la carrera)")

# 5. Porcentaje con beca
becados = destacados['beca'].sum()  # True cuenta como 1
pct_becados = becados / len(destacados) * 100
print(f"\nEstudiantes destacados con beca: {becados} de {len(destacados)} ({pct_becados:.1f}%)")
```

---

## ✅ Ejercicio 7: Promedio por semestre

```python
# Ejercicio 7: Promedio por semestre

# 1. Agrupar por semestre y calcular promedio de 'promedio'
promedio_semestre = rendimiento.groupby('semestre')['promedio'].mean().round(2)
print("Promedio académico por semestre:")
print("=" * 35)
print(promedio_semestre)

# 2. Semestre con mayor promedio
print(f"\nSemestre con mayor promedio: Semestre {promedio_semestre.idxmax()} ({promedio_semestre.max()})")

# 3. Semestre con menor promedio
print(f"Semestre con menor promedio: Semestre {promedio_semestre.idxmin()} ({promedio_semestre.min()})")

# 4. Estadísticas más completas por semestre
print("\nEstadísticas completas por semestre:")
print("=" * 60)
stats_semestre = rendimiento.groupby('semestre').agg(
    n_estudiantes=('estudiante_id', 'count'),
    promedio_medio=('promedio', 'mean'),
    asistencia_media=('asistencia_pct', 'mean')
).round(1)
print(stats_semestre)

# 5. Observaciones sobre la tendencia
print("\nObservación:")
print("Revisa si el promedio tiende a subir en semestres avanzados")
print("(podría ser efecto de supervivencia: los de bajo rendimiento desertan).")
```

---

## ✅ Ejercicio 8: Crear una nueva columna calculada

```python
# Ejercicio 8: Columna tasa_aprobacion

# 1. Crear la nueva columna
rendimiento['tasa_aprobacion'] = rendimiento['materias_aprobadas'] / rendimiento['materias_inscritas']

# 2. Mostrar primeras 10 filas con la nueva columna
print("Primeras 10 filas con tasa_aprobacion:")
print(rendimiento[['estudiante_id', 'carrera', 'promedio', 'materias_inscritas', 
                    'materias_aprobadas', 'tasa_aprobacion']].head(10))

# 3. Tasa de aprobación promedio general
tasa_general = rendimiento['tasa_aprobacion'].mean()
print(f"\nTasa de aprobación promedio general: {tasa_general:.2%}")

# 4. Tasa de aprobación promedio por carrera
print("\nTasa de aprobación promedio por carrera:")
print("=" * 45)
tasa_carrera = rendimiento.groupby('carrera')['tasa_aprobacion'].mean().sort_values(ascending=False)
for carrera, tasa in tasa_carrera.items():
    print(f"  {carrera:25s}: {tasa:.2%}")

# 5. Estudiantes que aprobaron todo (tasa = 1.0)
aprobaron_todo = rendimiento[rendimiento['tasa_aprobacion'] == 1.0]
print(f"\nEstudiantes que aprobaron TODAS sus materias: {len(aprobaron_todo):,}")
print(f"Porcentaje del total: {len(aprobaron_todo)/len(rendimiento)*100:.1f}%")
print(f"Promedio de estos estudiantes: {aprobaron_todo['promedio'].mean():.1f}")
print(f"\nDistribución por carrera de los que aprobaron todo:")
print(aprobaron_todo['carrera'].value_counts())
```
