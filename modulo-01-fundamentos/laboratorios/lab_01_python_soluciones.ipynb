## ✅ Ejercicio 1: Variables de presupuesto departamental

```python
# Ejercicio 1: Variables de presupuesto departamental

# 1. Crear las variables con sus tipos correspondientes
nombre_departamento = "Economía"          # str
presupuesto_anual = 500000                 # int
porcentaje_ejecutado = 67.5                # float
tiene_deficit = False                       # bool
anio_fiscal = 2025                         # int

# 2. Calcular montos
monto_ejecutado = presupuesto_anual * porcentaje_ejecutado / 100
monto_disponible = presupuesto_anual - monto_ejecutado

# 3. Imprimir resumen formateado
print("=" * 50)
print("RESUMEN PRESUPUESTARIO")
print("=" * 50)
print(f"Departamento:        {nombre_departamento}")
print(f"Año fiscal:          {anio_fiscal}")
print(f"Presupuesto anual:   Bs. {presupuesto_anual:,.0f}")
print(f"Ejecutado:           Bs. {monto_ejecutado:,.0f} ({porcentaje_ejecutado}%)")
print(f"Disponible:          Bs. {monto_disponible:,.0f} ({100 - porcentaje_ejecutado}%)")
print(f"Tiene déficit:       {'Sí' if tiene_deficit else 'No'}")

# 4. Verificar tipos
print("\nTipos de datos:")
print(f"  nombre_departamento:  {type(nombre_departamento).__name__}")
print(f"  presupuesto_anual:    {type(presupuesto_anual).__name__}")
print(f"  porcentaje_ejecutado: {type(porcentaje_ejecutado).__name__}")
print(f"  tiene_deficit:        {type(tiene_deficit).__name__}")
print(f"  anio_fiscal:          {type(anio_fiscal).__name__}")
```

## ✅ Ejercicio 2: Operaciones con strings - Formatear nombres de profesores

```python
# Ejercicio 2: Formatear nombres de profesores
nombres = [
    "  maria garcia LOPEZ ",
    "CARLOS rodriguez",
    "ana MENDEZ perez  ",
    "  pedro jose RAMIREZ",
    "LAURA martinez SILVA"
]

# Variable para rastrear el nombre más largo
nombre_mas_largo = ""
longitud_maxima = 0

print("NOMBRES FORMATEADOS")
print("=" * 50)

nombres_formateados = []  # Lista para guardar los resultados

for nombre_original in nombres:
    # 1. Eliminar espacios sobrantes
    nombre_limpio = nombre_original.strip()
    
    # 2. Convertir a formato título
    nombre_titulo = nombre_limpio.title()
    
    # 3. Agregar prefijo "Prof."
    nombre_final = f"Prof. {nombre_titulo}"
    
    # 4. Imprimir el resultado
    print(f"  Original: '{nombre_original}'")
    print(f"  Formateado: '{nombre_final}'")
    print()
    
    nombres_formateados.append(nombre_final)
    
    # 5. Rastrear el más largo
    if len(nombre_final) > longitud_maxima:
        longitud_maxima = len(nombre_final)
        nombre_mas_largo = nombre_final

print(f"El nombre más largo es: '{nombre_mas_largo}' ({longitud_maxima} caracteres)")
```

## ✅ Ejercicio 3: Listas - Gestionar códigos de materias

```python
# Ejercicio 3: Gestionar códigos de materias
materias = ["ECO-101", "ADM-201", "CON-301", "EST-101", "MAT-201"]

# 1. Cantidad de materias
print(f"Cantidad inicial de materias: {len(materias)}")
print(f"Lista inicial: {materias}")

# 2. Agregar materias
materias.append("ING-101")
materias.append("SOC-201")
print(f"\nDespués de agregar ING-101 y SOC-201: {materias}")
print(f"Cantidad: {len(materias)}")

# 3. Eliminar MAT-201
materias.remove("MAT-201")
print(f"\nDespués de eliminar MAT-201: {materias}")
print(f"Cantidad: {len(materias)}")

# 4. Ordenar alfabéticamente
materias.sort()
print(f"\nOrdenada alfabéticamente: {materias}")

# 5. Imprimir lista final
print(f"\nLista final: {materias}")

# 6. Separar cada código en departamento y número
print("\nCATALOGO DE MATERIAS")
print(f"{'Código':10s} | {'Depto':6s} | {'Número':6s}")
print("-" * 30)
for materia in materias:
    partes = materia.split("-")
    departamento = partes[0]
    numero = partes[1]
    print(f"{materia:10s} | {departamento:6s} | {numero:6s}")
```

## ✅ Ejercicio 4: Diccionarios - Catálogo de materias

```python
# Ejercicio 4: Catálogo de materias

# Crear el catálogo con 5 materias
catalogo = {
    "ECO-101": {
        "nombre": "Introducción a la Economía",
        "creditos": 4,
        "semestre": 1,
        "tipo": "obligatoria"
    },
    "EST-201": {
        "nombre": "Estadística I",
        "creditos": 3,
        "semestre": 3,
        "tipo": "obligatoria"
    },
    "ADM-301": {
        "nombre": "Administración de Empresas",
        "creditos": 4,
        "semestre": 5,
        "tipo": "obligatoria"
    },
    "CON-201": {
        "nombre": "Contabilidad de Costos",
        "creditos": 3,
        "semestre": 4,
        "tipo": "obligatoria"
    },
    "ECO-401": {
        "nombre": "Economía Digital",
        "creditos": 2,
        "semestre": 8,
        "tipo": "electiva"
    }
}

# 1. Información de una materia específica
codigo = "EST-201"
materia = catalogo[codigo]
print(f"Información de {codigo}:")
print(f"  Nombre: {materia['nombre']}")
print(f"  Créditos: {materia['creditos']}")
print(f"  Semestre: {materia['semestre']}")
print(f"  Tipo: {materia['tipo']}")

# 2. Total de créditos
total_creditos = 0
for codigo, info in catalogo.items():
    total_creditos += info['creditos']
print(f"\nTotal de créditos del catálogo: {total_creditos}")

# 3. Contar obligatorias y electivas
obligatorias = 0
electivas = 0
for codigo, info in catalogo.items():
    if info['tipo'] == 'obligatoria':
        obligatorias += 1
    else:
        electivas += 1
print(f"Materias obligatorias: {obligatorias}")
print(f"Materias electivas:    {electivas}")

# 4. Imprimir catálogo completo formateado
print("\n" + "=" * 70)
print("CATÁLOGO DE MATERIAS")
print("=" * 70)
print(f"{'Código':10s} | {'Nombre':30s} | {'Créd':4s} | {'Sem':3s} | {'Tipo':12s}")
print("-" * 70)
for codigo, info in catalogo.items():
    print(f"{codigo:10s} | {info['nombre']:30s} | {info['creditos']:4d} | {info['semestre']:3d} | {info['tipo']:12s}")
```

## ✅ Ejercicio 5: Condicionales - Clasificar ejecución presupuestaria

```python
# Ejercicio 5: Clasificar ejecución presupuestaria
porcentajes = [92, 75, 55, 35, 18]

print("CLASIFICACIÓN DE EJECUCIÓN PRESUPUESTARIA")
print("=" * 65)
print(f"{'% Ejecutado':>12s} | {'Clasificación':22s} | {'Acción recomendada'}")
print("-" * 65)

for pct in porcentajes:
    # Clasificar según el porcentaje
    if pct >= 90:
        clasificacion = "Ejecución óptima"
        accion = "Mantener el ritmo"
    elif pct >= 70:
        clasificacion = "Ejecución aceptable"
        accion = "Acelerar procesos pendientes"
    elif pct >= 50:
        clasificacion = "Ejecución media"
        accion = "Revisar plan de gasto"
    elif pct >= 30:
        clasificacion = "Ejecución baja"
        accion = "Alerta: riesgo de subejercicio"
    else:
        clasificacion = "Ejecución crítica"
        accion = "Reunión urgente con dirección"
    
    print(f"{pct:>11d}% | {clasificacion:22s} | {accion}")
```

## ✅ Ejercicio 6: Bucle for - Procesar calificaciones

```python
# Ejercicio 6: Procesar calificaciones
calificaciones = [15, 8, 12, 19, 10, 7, 14, 16, 11, 6, 18, 13, 9, 17, 12, 10, 14, 8, 15, 11]

# Inicializar variables para los cálculos
suma_total = 0
aprobados = 0
reprobados = 0
nota_maxima = calificaciones[0]  # Iniciar con la primera nota
nota_minima = calificaciones[0]

# Contadores por categoría
categorias = {
    "Sobresaliente (18-20)": 0,
    "Distinguido (15-17)": 0,
    "Bueno (12-14)": 0,
    "Regular (10-11)": 0,
    "Reprobado (0-9)": 0
}

# Procesar cada calificación con el bucle
for nota in calificaciones:
    # 1. Sumar para el promedio
    suma_total += nota
    
    # 2. Contar aprobados y reprobados
    if nota >= 10:
        aprobados += 1
    else:
        reprobados += 1
    
    # 3. Encontrar máxima y mínima (sin usar max()/min())
    if nota > nota_maxima:
        nota_maxima = nota
    if nota < nota_minima:
        nota_minima = nota
    
    # 4. Clasificar la nota
    if nota >= 18:
        categorias["Sobresaliente (18-20)"] += 1
    elif nota >= 15:
        categorias["Distinguido (15-17)"] += 1
    elif nota >= 12:
        categorias["Bueno (12-14)"] += 1
    elif nota >= 10:
        categorias["Regular (10-11)"] += 1
    else:
        categorias["Reprobado (0-9)"] += 1

# Calcular promedio
promedio = suma_total / len(calificaciones)

# 5. Imprimir informe completo
print("=" * 50)
print("INFORME DE CALIFICACIONES - ESTADÍSTICA I")
print("=" * 50)
print(f"Total de estudiantes:  {len(calificaciones)}")
print(f"Promedio de la sección: {promedio:.1f}")
print(f"Nota máxima:           {nota_maxima}")
print(f"Nota mínima:           {nota_minima}")
print(f"Aprobados:             {aprobados} ({aprobados/len(calificaciones)*100:.1f}%)")
print(f"Reprobados:            {reprobados} ({reprobados/len(calificaciones)*100:.1f}%)")

print("\nDISTRIBUCIÓN POR CATEGORÍA:")
print("-" * 50)
for categoria, conteo in categorias.items():
    pct = conteo / len(calificaciones) * 100
    barra = "█" * conteo
    print(f"  {categoria:25s} | {conteo:2d} ({pct:5.1f}%) {barra}")
```

## ✅ Ejercicio 7: Funciones - Calcular índice académico

```python
# Ejercicio 7: Función calcular_indice_academico

def calcular_indice_academico(notas, creditos):
    """
    Calcula el índice académico ponderado de un estudiante.
    
    Parámetros:
        notas: lista de calificaciones (escala 0-20)
        creditos: lista de créditos por materia
    
    Retorna:
        Diccionario con índice, total de créditos, materias aprobadas y clasificación
    """
    # Calcular la suma ponderada: nota * créditos
    suma_ponderada = 0
    total_creditos = 0
    materias_aprobadas = 0
    
    for i in range(len(notas)):
        suma_ponderada += notas[i] * creditos[i]
        total_creditos += creditos[i]
        if notas[i] >= 10:
            materias_aprobadas += 1
    
    # Calcular el índice ponderado
    indice = suma_ponderada / total_creditos
    
    # Determinar la clasificación
    if indice >= 18:
        clasificacion = "Cum Laude"
    elif indice >= 16:
        clasificacion = "Magna"
    elif indice >= 14:
        clasificacion = "Distinguido"
    else:
        clasificacion = "Regular"
    
    return {
        "indice": round(indice, 2),
        "total_creditos": total_creditos,
        "materias_aprobadas": materias_aprobadas,
        "clasificacion": clasificacion
    }

# Probar la función
notas = [16, 14, 18, 12, 15]
creditos = [4, 3, 4, 3, 2]

resultado = calcular_indice_academico(notas, creditos)

print("RESULTADO DEL ÍNDICE ACADÉMICO")
print("=" * 40)
print(f"Índice ponderado:    {resultado['indice']}")
print(f"Total de créditos:  {resultado['total_creditos']}")
print(f"Materias aprobadas: {resultado['materias_aprobadas']} de {len(notas)}")
print(f"Clasificación:      {resultado['clasificacion']}")

# Verificación manual: (16*4 + 14*3 + 18*4 + 12*3 + 15*2) / (4+3+4+3+2)
# = (64 + 42 + 72 + 36 + 30) / 16 = 244 / 16 = 15.25
print(f"\nVerificación: (16*4 + 14*3 + 18*4 + 12*3 + 15*2) / 16 = {(16*4 + 14*3 + 18*4 + 12*3 + 15*2) / 16}")
```

## ✅ Ejercicio 8: Integración - Procesar estudiantes en riesgo

```python
# Ejercicio 8: Procesar estudiantes en riesgo
estudiantes = [
    {"nombre": "Ana Pérez", "carrera": "Economía", "promedio": 15.2, "asistencia": 92, "materias_reprobadas": 0},
    {"nombre": "Carlos Gómez", "carrera": "Administración", "promedio": 9.5, "asistencia": 60, "materias_reprobadas": 4},
    {"nombre": "Laura Torres", "carrera": "Contaduría", "promedio": 11.0, "asistencia": 78, "materias_reprobadas": 2},
    {"nombre": "Miguel López", "carrera": "Economía", "promedio": 7.8, "asistencia": 45, "materias_reprobadas": 5},
    {"nombre": "Sofía Herrera", "carrera": "RRII", "promedio": 17.5, "asistencia": 95, "materias_reprobadas": 0},
    {"nombre": "Pedro Castillo", "carrera": "Administración", "promedio": 12.3, "asistencia": 82, "materias_reprobadas": 1},
    {"nombre": "María Rivas", "carrera": "Contaduría", "promedio": 8.2, "asistencia": 55, "materias_reprobadas": 3},
    {"nombre": "José Navarro", "carrera": "Economía", "promedio": 14.1, "asistencia": 88, "materias_reprobadas": 1},
]

# 1. Función para evaluar riesgo
def evaluar_riesgo(estudiante):
    """
    Evalúa el nivel de riesgo académico de un estudiante.
    
    Criterios:
    - ALTO: promedio < 10 O asistencia < 60% O materias_reprobadas >= 4
    - MEDIO: promedio 10-12 O asistencia 60-75% O materias_reprobadas 2-3
    - BAJO: los demás casos
    """
    promedio = estudiante["promedio"]
    asistencia = estudiante["asistencia"]
    reprobadas = estudiante["materias_reprobadas"]
    
    # Verificar riesgo ALTO primero
    if promedio < 10 or asistencia < 60 or reprobadas >= 4:
        return "ALTO"
    # Verificar riesgo MEDIO
    elif promedio <= 12 or asistencia <= 75 or reprobadas >= 2:
        return "MEDIO"
    # Si no cumple ninguna condición anterior, es BAJO
    else:
        return "BAJO"

# 2. Aplicar la función a todos los estudiantes
conteo_riesgo = {"ALTO": 0, "MEDIO": 0, "BAJO": 0}
estudiantes_alto_riesgo = []
suma_promedios = 0

print("INFORME DE RIESGO ACADÉMICO")
print("=" * 75)
print(f"{'Nombre':20s} | {'Carrera':15s} | {'Prom':5s} | {'Asist':5s} | {'Rep':3s} | {'Riesgo':6s}")
print("-" * 75)

for est in estudiantes:
    riesgo = evaluar_riesgo(est)
    conteo_riesgo[riesgo] += 1
    suma_promedios += est["promedio"]
    
    # Guardar los de riesgo alto para el reporte
    if riesgo == "ALTO":
        estudiantes_alto_riesgo.append(est["nombre"])
    
    # Imprimir fila del estudiante
    print(f"{est['nombre']:20s} | {est['carrera']:15s} | {est['promedio']:5.1f} | {est['asistencia']:4d}% | {est['materias_reprobadas']:3d} | {riesgo:6s}")

# 3. Imprimir informe resumido
promedio_general = suma_promedios / len(estudiantes)

print("\n" + "=" * 75)
print("RESUMEN")
print("=" * 75)
print(f"Total de estudiantes:    {len(estudiantes)}")
print(f"Promedio general:        {promedio_general:.1f}")
print(f"")
print(f"Distribución por riesgo:")
for nivel, cantidad in conteo_riesgo.items():
    pct = cantidad / len(estudiantes) * 100
    indicador = "🔴" if nivel == "ALTO" else ("🟡" if nivel == "MEDIO" else "🟢")
    print(f"  {indicador} {nivel:6s}: {cantidad} estudiantes ({pct:.1f}%)")

print(f"\nEstudiantes en riesgo ALTO (requieren seguimiento):")
for nombre in estudiantes_alto_riesgo:
    print(f"  - {nombre}")
```
