## Parte 1: Errores Identificados (12 errores)

### Error 1: Sin título

| Campo | Respuesta |
|-------|----------|
| **Descripción** | El dashboard no tiene título, el usuario no sabe qué está viendo |
| **Elemento afectado** | Encabezado del dashboard |
| **Principio violado** | Narrativa con datos (Cole Nussbaumer) - contexto es esencial |
| **Impacto** | El usuario no puede entender rápidamente el propósito del dashboard |

**Solución**: Agregar título descriptivo que comunique el hallazgo principal o el tema.
- Ejemplo: "Rendimiento Académico FACES — 75% de aprobación en 2024"

---

### Error 2: Fondo degradado

| Campo | Respuesta |
|-------|----------|
| **Descripción** | Fondo degradado azul a morado que dificulta la lectura |
| **Elemento afectado** | Fondo completo del dashboard |
| **Principio violado** | Tufte - ratio tinta-datos (chartjunk); Accesibilidad - contraste |
| **Impacto** | Textos ilegibles, fatiga visual, distracción del contenido |

**Solución**: Usar fondo blanco (#FFFFFF) o gris muy claro (#F5F5F5).

---

### Error 3: 9 tarjetas KPI sin etiquetas

| Campo | Respuesta |
|-------|----------|
| **Descripción** | Demasiados KPIs (9) sin etiquetas que expliquen qué representa cada número |
| **Elemento afectado** | Fila de tarjetas KPI |
| **Principio violado** | Gestalt (similitud) - elementos indistinguibles; Carga cognitiva |
| **Impacto** | Usuario no sabe qué significa "78%" o "156" - números sin contexto |

**Solución**: 
- Reducir a 3-5 KPIs máximo
- Agregar etiquetas descriptivas (ej: "Tasa de Aprobación: 78%")
- Incluir unidades y período de referencia

---

### Error 4: Gráfico de torta 3D con 12 categorías

| Campo | Respuesta |
|-------|----------|
| **Descripción** | Gráfico de torta en 3D con 12 categorías y colores arcoíris |
| **Elemento afectado** | Gráfico de torta |
| **Principio violado** | Tufte - ratio tinta-datos; Few - demasiadas categorías; Cleveland & McGill - tortas difíciles de comparar |
| **Impacto** | Imposible comparar proporciones con precisión; el 3D distorsiona ángulos |

**Solución**: 
- Reemplazar por barras horizontales ordenadas
- O agrupar en máximo 5 categorías + "Otros"
- Eliminar 3D completamente

---

### Error 5: Eje Y truncado en gráfico de barras

| Campo | Respuesta |
|-------|----------|
| **Descripción** | El eje Y comienza en 800, no en 0, exagerando diferencias |
| **Elemento afectado** | Gráfico de barras |
| **Principio violado** | Honestidad visual (Tufte); Integridad gráfica |
| **Impacto** | Una barra de 1100 parece 4x más grande que una de 850, cuando solo es 30% mayor |

**Solución**: El eje Y en gráficos de barras siempre debe comenzar en 0.

````python
# Demostración visual del Error 5: Eje truncado
import matplotlib.pyplot as plt
import numpy as np

categorias = ['A', 'B', 'C', 'D', 'E']
valores = [1050, 980, 920, 890, 850]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# MAL: Eje truncado
axes[0].bar(categorias, valores, color='#ff6b6b')
axes[0].set_ylim(800, 1100)
axes[0].set_title('MAL: Eje truncado (800-1100)\nA parece 10x más grande que E', 
                  fontsize=12, color='red')
axes[0].set_ylabel('Valor')

# Añadir anotación
axes[0].annotate('', xy=(0, 850), xytext=(0, 1050),
                 arrowprops=dict(arrowstyle='<->', color='red', lw=2))
axes[0].text(0.3, 950, 'Parece\nhuge!', fontsize=10, color='red')

# BIEN: Eje desde cero
axes[1].bar(categorias, valores, color='#51cf66')
axes[1].set_ylim(0, 1200)
axes[1].set_title('BIEN: Eje desde 0\nDiferencia real es solo ~20%', 
                  fontsize=12, color='green')
axes[1].set_ylabel('Valor')

# Añadir línea de referencia
axes[1].axhline(y=1050, color='gray', linestyle='--', alpha=0.5)
axes[1].axhline(y=850, color='gray', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()

# Calcular la distorsión
print("Análisis de distorsión:")
print(f"Valor A: {valores[0]}, Valor E: {valores[4]}")
print(f"Diferencia real: {(valores[0] - valores[4]) / valores[4] * 100:.1f}%")
print(f"En el gráfico truncado, la barra A parece {(valores[0]-800)/(valores[4]-800):.1f}x más alta que E")
````

### Error 6: Demasiados colores sin significado

| Campo | Respuesta |
|-------|----------|
| **Descripción** | Gráfico de barras usa 7 colores diferentes sin razón semántica |
| **Elemento afectado** | Gráfico de barras 3D |
| **Principio violado** | Pre-atentivo (color como variable categórica innecesaria); Gestalt |
| **Impacto** | El usuario busca significado en los colores cuando no lo hay |

**Solución**: 
- Usar un solo color para todas las barras
- O usar color para resaltar solo 1-2 elementos importantes
- Reservar rojo/verde para indicar bueno/malo

---

### Error 7: 8 líneas en tonos similares de gris

| Campo | Respuesta |
|-------|----------|
| **Descripción** | Gráfico de líneas con 8 series en grises indistinguibles |
| **Elemento afectado** | Gráfico de líneas |
| **Principio violado** | Gestalt (similitud excesiva); Pre-atentivo (no hay foco) |
| **Impacto** | Imposible seguir una línea específica o extraer insights |

**Solución**:
- Resaltar solo 1-2 líneas importantes con color vivo
- Resto en gris muy claro como contexto
- O usar small multiples (un gráfico pequeño por serie)

---

### Error 8: Gráfico de barras en 3D

| Campo | Respuesta |
|-------|----------|
| **Descripción** | Las barras tienen efecto 3D con sombra y bisel |
| **Elemento afectado** | Gráfico de barras |
| **Principio violado** | Tufte - ratio tinta-datos; El 3D no aporta información |
| **Impacto** | Ocupa más espacio visual, dificulta lectura precisa de valores |

**Solución**: Eliminar todo efecto 3D. Usar barras planas (2D).

---

### Error 9: Título genérico "Datos"

| Campo | Respuesta |
|-------|----------|
| **Descripción** | El título del gráfico de barras es "Datos" - no comunica nada |
| **Elemento afectado** | Título del gráfico de barras |
| **Principio violado** | Narrativa - el título debe comunicar el hallazgo |
| **Impacto** | El usuario debe descifrar qué muestra el gráfico |

**Solución**: Título descriptivo con el hallazgo.
- Malo: "Datos"
- Regular: "Estudiantes por carrera"
- Bueno: "Administración lidera la matrícula con 1,050 estudiantes"

### Error 10: Tabla ilegible

| Campo | Respuesta |
|-------|----------|
| **Descripción** | Tabla con 25 filas, fuente de 8pt, sin formato condicional |
| **Elemento afectado** | Tabla de datos |
| **Principio violado** | Accesibilidad (fuente mínima 10-12pt); Few (demasiados datos) |
| **Impacto** | Imposible de leer; no resalta qué es importante |

**Solución**:
- Mostrar solo Top 5-10 filas
- Aumentar fuente a mínimo 10pt
- Agregar formato condicional (barras de datos o color)
- Permitir ordenamiento interactivo

---

### Error 11: Scatter plot con pocos puntos y título "Gráfico 4"

| Campo | Respuesta |
|-------|----------|
| **Descripción** | Scatter con solo 6 puntos, sin etiquetas de eje, título "Gráfico 4" |
| **Elemento afectado** | Scatter plot |
| **Principio violado** | Elección de gráfico; Con 6 datos, una tabla es mejor |
| **Impacto** | El scatter no revela patrones con tan pocos puntos |

**Solución**:
- Con solo 6 puntos, usar una tabla con los valores
- Si debe ser gráfico, usar barras o puntos etiquetados
- Agregar etiquetas de ejes y título descriptivo

---

### Error 12: Sin filtros ni interactividad

| Campo | Respuesta |
|-------|----------|
| **Descripción** | No hay segmentadores ni forma de filtrar los datos |
| **Elemento afectado** | Dashboard completo |
| **Principio violado** | Interactividad - el usuario debe poder explorar |
| **Impacto** | Vista estática, no permite responder preguntas específicas |

**Solución**:
- Agregar segmentador de carrera
- Agregar segmentador de período/fecha
- Configurar cross-filtering entre gráficos

````python
# Demostración: Demasiadas líneas vs líneas con foco
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
x = np.arange(1, 13)
series = [np.cumsum(np.random.randn(12)) + i*2 for i in range(8)]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# MAL: Todas las líneas en grises similares
grays = ['#666', '#777', '#888', '#999', '#aaa', '#bbb', '#ccc', '#ddd']
for i, s in enumerate(series):
    axes[0].plot(x, s, color=grays[i], linewidth=1.5)
axes[0].set_title('MAL: 8 líneas en grises similares\n¿Cuál es importante?', 
                  fontsize=12, color='red')
axes[0].set_xlabel('Mes')
axes[0].legend([f'Serie {i+1}' for i in range(8)], fontsize=6, loc='upper left')

# BIEN: Resaltar solo una o dos
for i, s in enumerate(series):
    if i == 3:  # Resaltar serie principal
        axes[1].plot(x, s, color='#2563eb', linewidth=3, label='Administración', zorder=10)
    elif i == 5:  # Segunda importante
        axes[1].plot(x, s, color='#dc2626', linewidth=2.5, label='Economía', zorder=9)
    else:
        axes[1].plot(x, s, color='#e5e7eb', linewidth=1, alpha=0.7)

axes[1].set_title('BIEN: Resaltar las series importantes\nResto como contexto', 
                  fontsize=12, color='green')
axes[1].set_xlabel('Mes')
axes[1].legend(loc='upper left')

plt.tight_layout()
plt.show()
````

## Parte 3: Diseño del Dashboard Mejorado

### Especificaciones del diseño propuesto

**Audiencia:** Directores de escuela y coordinadores académicos

**Pregunta de decisión:** ¿Qué carreras necesitan intervención académica?

### Layout propuesto

```
┌──────────────────────────────────────────────────────────────────┐

│  La tasa de aprobación alcanzó 75% en 2024 (+5% vs 2023)        │

├────────────┬────────────┬────────────┬───────────────────────────┤
│            │            │            │                           │
│  Total     │  Promedio  │  Tasa      │  Filtros:                │
│  Estud.    │  Notas     │  Aprob.    │  [Carrera ▼]             │

│            │            │            │                           │
├────────────┴────────────┴────────────┼───────────────────────────┤
│                                      │                           │
│  [Barras horizontales]               │  [Líneas de tendencia]   │
```

### Paleta de colores

| Uso | Color | Código |
|-----|-------|--------|
| Primario (positivo) | Azul | #2563EB |
| Secundario (contexto) | Gris | #94A3B8 |
| Alerta/Negativo | Rojo | #DC2626 |
| Éxito/Positivo | Verde | #059669 |
| Fondo | Blanco | #FFFFFF |
| Texto | Gris oscuro | #1E293B |

### Principios aplicados

1. **Ratio tinta-datos alto** (Tufte): Eliminamos 3D, fondos de color, bordes innecesarios
2. **Jerarquía visual** (Gestalt): KPIs arriba, detalle abajo; foco en lo importante
3. **Honestidad visual**: Ejes desde cero, comparaciones justas
4. **Accesibilidad**: Fuentes legibles (≥10pt), contraste adecuado
5. **Narrativa**: Título con hallazgo, no genérico

````python
# Dashboard mejorado completo
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

# Datos de ejemplo
carreras = ['Administración', 'Contaduría', 'Economía', 'RRII', 'Sociología']
estudiantes = [1050, 890, 750, 580, 245]
promedios = [14.5, 14.2, 13.8, 14.0, 12.8]
tasas = [78, 75, 72, 74, 62]

# Configuración
fig = plt.figure(figsize=(16, 12), facecolor='white')

# Título principal
fig.text(0.05, 0.96, 'Dashboard de Rendimiento Académico — FACES', 
         fontsize=18, fontweight='bold', color='#1e293b')
fig.text(0.05, 0.93, 'La tasa de aprobación alcanzó 75% en 2024 (+5% vs 2023)', 
         fontsize=12, color='#059669')
fig.text(0.95, 0.93, 'Actualizado: Enero 2025', 
         fontsize=10, color='#64748b', ha='right')

# KPIs
ax_kpi = fig.add_axes([0.05, 0.78, 0.6, 0.12])
ax_kpi.axis('off')

kpis = [
    ('Total Estudiantes', '3,001', ''),
    ('Promedio Notas', '14.2', ''),
    ('Tasa Aprobación', '75%', '▲ +5%')
]

for i, (label, value, delta) in enumerate(kpis):
    x_pos = 0.15 + i * 0.35
    
    # Caja del KPI
    rect = mpatches.FancyBboxPatch(
        (x_pos - 0.12, 0.1), 0.28, 0.8,
        boxstyle="round,pad=0.02,rounding_size=0.05",
        facecolor='white', edgecolor='#e2e8f0', linewidth=2
    )
    ax_kpi.add_patch(rect)
    
    # Valor
    ax_kpi.text(x_pos, 0.6, value, ha='center', va='center', 
                fontsize=28, fontweight='bold', color='#2563eb')
    # Etiqueta
    ax_kpi.text(x_pos, 0.25, label, ha='center', va='center', 
                fontsize=11, color='#64748b')
    # Delta (si existe)
    if delta:
        ax_kpi.text(x_pos + 0.08, 0.7, delta, ha='left', va='center', 
                    fontsize=10, color='#059669', fontweight='bold')

# Filtros (placeholder)
ax_filtro = fig.add_axes([0.7, 0.78, 0.25, 0.12])
ax_filtro.axis('off')
ax_filtro.text(0.5, 0.8, 'Filtros', ha='center', fontsize=12, fontweight='bold', color='#1e293b')
ax_filtro.add_patch(mpatches.FancyBboxPatch(
    (0.1, 0.4), 0.8, 0.3, boxstyle="round,pad=0.02",
    facecolor='#f8fafc', edgecolor='#cbd5e1'))
ax_filtro.text(0.5, 0.55, '▼ Todas las carreras', ha='center', fontsize=10)
ax_filtro.add_patch(mpatches.FancyBboxPatch(
    (0.1, 0.0), 0.8, 0.3, boxstyle="round,pad=0.02",
    facecolor='#f8fafc', edgecolor='#cbd5e1'))
ax_filtro.text(0.5, 0.15, '▼ 2024', ha='center', fontsize=10)

# Gráfico de barras - Estudiantes por carrera
ax1 = fig.add_subplot(2, 2, 3)
sorted_idx = np.argsort(estudiantes)
carreras_sorted = [carreras[i] for i in sorted_idx]
estudiantes_sorted = [estudiantes[i] for i in sorted_idx]

colors = ['#94a3b8'] * len(carreras_sorted)
colors[-1] = '#2563eb'  # Resaltar la mayor

bars = ax1.barh(carreras_sorted, estudiantes_sorted, color=colors, height=0.6)
ax1.set_title('Administración lidera la matrícula con 1,050 estudiantes', 
              fontsize=12, fontweight='bold', loc='left', color='#1e293b')
ax1.set_xlabel('Cantidad de estudiantes', fontsize=10, color='#64748b')
ax1.set_xlim(0, max(estudiantes_sorted) * 1.15)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)

# Etiquetas de datos
for bar, val in zip(bars, estudiantes_sorted):
    ax1.text(val + 20, bar.get_y() + bar.get_height()/2, 
             f'{val:,}', va='center', fontweight='bold', color='#1e293b')

# Gráfico de líneas - Tendencia
ax2 = fig.add_subplot(2, 2, 4)
periodos = ['2021-1', '2021-2', '2022-1', '2022-2', '2023-1', '2023-2', '2024-1', '2024-2']
promedios_tiempo = [13.2, 13.5, 13.4, 13.8, 14.0, 14.1, 14.3, 14.2]

ax2.plot(periodos, promedios_tiempo, marker='o', linewidth=2.5, 
         color='#2563eb', markersize=8, markerfacecolor='white', 
         markeredgewidth=2, markeredgecolor='#2563eb')
ax2.axhline(y=14, color='#dc2626', linestyle='--', alpha=0.7, linewidth=1.5)
ax2.text(7.1, 14.05, 'Meta: 14', fontsize=9, color='#dc2626')

ax2.set_title('El promedio superó la meta desde 2023-1', 
              fontsize=12, fontweight='bold', loc='left', color='#1e293b')
ax2.set_ylabel('Promedio de notas', fontsize=10, color='#64748b')
ax2.set_ylim(12.5, 15)
ax2.tick_params(axis='x', rotation=45)
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)

# Fill area above/below meta
ax2.fill_between(periodos, promedios_tiempo, 14, 
                 where=[p >= 14 for p in promedios_tiempo],
                 alpha=0.2, color='#059669')
ax2.fill_between(periodos, promedios_tiempo, 14, 
                 where=[p < 14 for p in promedios_tiempo],
                 alpha=0.2, color='#dc2626')

# Pie de página
fig.text(0.05, 0.02, 'Fuente: Sistema Académico UCAB | Período: 2023-2024', 
         fontsize=9, color='#94a3b8')

plt.tight_layout(rect=[0, 0.03, 1, 0.9])
plt.show()
````
