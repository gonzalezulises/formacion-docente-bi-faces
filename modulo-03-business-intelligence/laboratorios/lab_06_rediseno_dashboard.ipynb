## Referencia: Errores comunes y sus soluciones

Use esta tabla como referencia después de completar el ejercicio.

| Error | Principio | Solución |
|---|---|---|
| Sin título o título genérico | Narrativa | Título que comunique el hallazgo principal |
| Demasiados KPIs sin etiqueta | Gestalt (similitud) | Máximo 5 KPIs, con nombre y unidad |
| Gráfico de torta 3D | Tufte (ratio tinta-datos) | Barras horizontales o treemap |
| Eje truncado en barras | Honestidad visual | Eje desde cero en gráficos de barras |
| Demasiados colores | Pre-atentivo (color) | Paleta de 3-4 colores máximo |
| Líneas en grises similares | Gestalt (similitud) | Resaltar 1-2 líneas, el resto en gris claro |
| Tabla con fuente pequeña | Accesibilidad | Mostrar solo top 5-10, fuente ≥ 10pt |
| Scatter con pocos puntos | Elección de gráfico | Tabla o barras con 6 datos |
| Sin filtros interactivos | Interactividad | Agregar segmentadores |
| Fondo degradado | Tufte (chartjunk) | Fondo blanco o gris muy claro |
| Sin fecha de actualización | Contexto | Agregar fecha y fuente |
| 3D sin propósito | Tufte (ratio tinta-datos) | Todo en 2D |

````python
# Ejemplo visual: versión "mala" vs versión "buena" de un solo gráfico
import matplotlib.pyplot as plt
import numpy as np

carreras = ['Admin', 'Contad', 'Econ', 'RRII', 'Sociol']
valores = [950, 880, 850, 920, 870]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# VERSIÓN MALA
colors_bad = ['#ff0000', '#00ff00', '#ffff00', '#ff00ff', '#00ffff']
axes[0].bar(carreras, valores, color=colors_bad, edgecolor='black', linewidth=2)
axes[0].set_ylim(800, 1000)  # Eje truncado
axes[0].set_title('Datos', fontsize=10)  # Título genérico
axes[0].set_facecolor('#e6e6ff')  # Fondo de color
# Agregar grid excesivo
axes[0].grid(True, which='both', linewidth=1.5, color='gray')
axes[0].annotate('VERSIÓN MALA', xy=(0.5, 0.02), xycoords='axes fraction',
                 ha='center', fontsize=14, color='red', fontweight='bold')

# VERSIÓN BUENA
sorted_idx = np.argsort(valores)
carreras_sorted = [carreras[i] for i in sorted_idx]
valores_sorted = [valores[i] for i in sorted_idx]
colors_good = ['#94a3b8'] * 5
colors_good[-1] = '#2563eb'  # Resaltar la más alta

axes[1].barh(carreras_sorted, valores_sorted, color=colors_good)
axes[1].set_xlim(0, 1100)  # Eje desde cero
axes[1].set_title('Administración lidera la matrícula con 950 estudiantes', 
                   fontsize=11, fontweight='bold')
axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)

for i, (c, v) in enumerate(zip(carreras_sorted, valores_sorted)):
    axes[1].text(v + 10, i, str(v), va='center', fontweight='bold')

axes[1].annotate('VERSIÓN BUENA', xy=(0.5, 0.02), xycoords='axes fraction',
                 ha='center', fontsize=14, color='green', fontweight='bold')

plt.tight_layout()
plt.show()
````
