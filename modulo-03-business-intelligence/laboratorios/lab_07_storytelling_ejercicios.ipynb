## El Problema del "Mostrar Todo"

> *"Si intentas comunicar todo, no comunicas nada."*

La mayoría de presentaciones de datos fallan porque:
- Muestran **todos** los datos sin jerarquía
- Carecen de **narrativa** (son colecciones de gráficos)
- No tienen **llamado a la acción** claro
- Asumen que los datos "hablan por sí mismos"

**Los datos no hablan por sí mismos. Necesitan un narrador.**

## Parte 3: La Estructura de Presentación

Una presentación efectiva de datos sigue esta estructura:

```
┌──────────────────────────────────────────────────────────────┐
│  APERTURA (30 seg)                                           │
│  Hook: dato sorprendente, pregunta, provocación              │
│  "¿Sabían que hemos perdido el equivalente a 3 secciones?"   │
└──────────────────────────────────────────────────────────────┘
                              │
                              ▼
┌──────────────────────────────────────────────────────────────┐
│  CONTEXTO (2-3 min)                                          │
│  ¿Por qué estamos aquí? ¿Qué pregunta respondemos?           │
│  "Analizamos la matrícula de los últimos 5 años..."          │
└──────────────────────────────────────────────────────────────┘
                              │
                              ▼
┌──────────────────────────────────────────────────────────────┐
│  HALLAZGOS (5-10 min)                                        │
│  3-5 puntos clave, ordenados por importancia                 │
│  Cada hallazgo: afirmación + evidencia + implicación         │
└──────────────────────────────────────────────────────────────┘
                              │
                              ▼
┌──────────────────────────────────────────────────────────────┐
│  LLAMADO A LA ACCIÓN (1-2 min)                               │
│  ¿Qué queremos que haga la audiencia?                        │
│  "Proponemos implementar un programa de retención..."        │
└──────────────────────────────────────────────────────────────┘
```

## Resumen: Los 7 Principios del Storytelling con Datos

1. **Conoce tu audiencia**: ¿Qué necesitan saber? ¿Qué acción pueden tomar?

2. **Un mensaje central**: Si solo pudieran recordar una cosa, ¿cuál sería?

3. **Contexto-Conflicto-Conclusión**: Estructura narrativa clara

4. **Menos es más**: Eliminar todo lo que no contribuye al mensaje

5. **Diseño intencional**: Cada elemento (color, posición, tamaño) tiene propósito

6. **Guiar la atención**: Usar pre-attentive attributes para enfocar

7. **Llamado a la acción**: Terminar con recomendación específica
