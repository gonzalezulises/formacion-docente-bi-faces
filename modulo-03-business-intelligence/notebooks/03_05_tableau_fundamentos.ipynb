## 1. ¿Qué es Tableau?

Tableau es una plataforma de visualización de datos reconocida por su facilidad de uso y la calidad visual de sus gráficos. Su filosofía es "drag and drop": arrastrar campos para crear visualizaciones.

### Versiones

| Versión | Costo | Plataforma | Uso |
|---|---|---|---|
| **Tableau Public** | Gratuito | Windows / Mac | Crear y publicar dashboards (públicos) |
| **Tableau Desktop** | Licencia paga | Windows / Mac | Dashboards privados + más conectores |
| **Tableau Server/Cloud** | Enterprise | Web | Compartir en la organización |

> **Para este curso** usaremos **Tableau Public**, que es gratuito y suficiente para aprender los fundamentos.

## 2. Conexión a datos: dimensiones y medidas

### 2.1 Conectar un CSV

**Pasos en Tableau Public:**

1. Abrir Tableau Public
2. En la pantalla de inicio: **Conectar → Archivo de texto**
3. Seleccionar `rendimiento_academico.csv`
4. Tableau mostrará una vista previa de los datos
5. Clic en **Ir a la hoja de trabajo** (esquina inferior izquierda)

### 2.2 Dimensiones vs Medidas

Tableau clasifica automáticamente los campos:

| Tipo | Color | Qué son | Ejemplos |
|---|---|---|---|
| **Dimensiones** | Azul | Campos categóricos/cualitativos | Carrera, período, sexo, estado |
| **Medidas** | Verde | Campos numéricos/cuantitativos | Nota, edad, asistencia |

**Importante:** Tableau a veces clasifica mal. Si un campo numérico es realmente una categoría (como "año" o "código"), arrástrelo de Medidas a Dimensiones.

### 2.3 Discreto vs Continuo

Además de dimensión/medida, Tableau distingue:

| Tipo | Encabezado | Comportamiento |
|---|---|---|
| **Discreto** | Azul | Crea categorías (encabezados separados) |
| **Continuo** | Verde | Crea un eje continuo (rango numérico) |

Un campo puede cambiar entre discreto y continuo con clic derecho.

````python
# Veamos la clasificación de campos que hará Tableau
import pandas as pd

df = pd.read_csv('../datasets/universidad/rendimiento_academico.csv')

print("Clasificación de campos para Tableau:")
print("="*50)
for col in df.columns:
    dtype = df[col].dtype
    if dtype == 'object':
        tipo_tableau = 'DIMENSIÓN (texto)'
    elif 'int' in str(dtype) or 'float' in str(dtype):
        # Verificar si es realmente una categoría
        n_unique = df[col].nunique()
        if n_unique < 10 and col.lower() in ['id', 'codigo', 'año', 'semestre']:
            tipo_tableau = 'DIMENSIÓN (numérico → reclasificar)'
        else:
            tipo_tableau = f'MEDIDA (numérico, {n_unique} valores únicos)'
    else:
        tipo_tableau = 'DIMENSIÓN (fecha)'
    
    print(f"  {col:20s} → {tipo_tableau}")
````

## 3. Tipos de gráficos en Tableau

### 3.1 Gráfico de barras

**Para crear:**
1. Arrastrar `carrera` a **Filas**
2. Arrastrar `nota` a **Columnas** (Tableau la agrega como SUM por defecto)
3. Clic derecho en SUM(nota) → **Medida → Promedio**
4. Ordenar: clic en el ícono de ordenar en la barra de herramientas

**Personalizar:**
- Color: arrastrar `carrera` a **Color** en la tarjeta Marcas
- Etiquetas: arrastrar `nota` a **Etiqueta** en la tarjeta Marcas
- Título: doble clic en el título para editarlo

### 3.2 Gráfico de líneas

**Para crear:**
1. Arrastrar `periodo` a **Columnas**
2. Arrastrar `nota` a **Filas** → cambiar a AVG(nota)
3. Si quiere múltiples líneas: arrastrar `carrera` a **Color**

**Personalizar:**
- Línea de referencia: clic derecho en el eje → Agregar línea de referencia → Constante → valor meta
- Anotaciones: clic derecho en un punto → Anotar → Marca

### 3.3 Scatter plot

**Para crear:**
1. Arrastrar una medida a **Columnas** (ej: asistencia)
2. Arrastrar otra medida a **Filas** (ej: nota)
3. Tableau cambia automáticamente a scatter
4. Arrastrar `carrera` a **Color** para distinguir grupos

**Agregar línea de tendencia:**
- Menú Análisis → Línea de tendencia → Mostrar líneas de tendencia

### 3.4 Mapa

**Para crear** (requiere campo geográfico):
1. Arrastrar campo de ubicación a **Detalle** (Tableau detecta coordenadas)
2. Arrastrar medida a **Color** o **Tamaño**
3. Tableau genera el mapa automáticamente

### 3.5 Treemap

**Para crear:**
1. En "Mostrar" (Show Me, esquina superior derecha) → seleccionar treemap
2. Arrastrar dimensión a **Color** y medida a **Tamaño**
3. Útil para ver composición y jerarquías

## 4. Filtros y cálculos básicos

### 4.1 Filtros

Hay varias formas de filtrar en Tableau:

| Método | Cómo | Uso |
|---|---|---|
| **Arrastrar a Filtros** | Arrastrar campo a tarjeta Filtros | Filtro fijo |
| **Mostrar filtro** | Clic derecho en filtro → Mostrar filtro | Filtro interactivo visible |
| **Clic en gráfico** | Clic en una barra o punto | Filtro contextual temporal |
| **Filtro de contexto** | Clic derecho en filtro → Agregar al contexto | Filtro prioritario |

### 4.2 Campos calculados

Equivalente a las columnas calculadas de Power BI:

1. Clic derecho en el panel de datos → **Crear campo calculado**
2. Escribir la fórmula

**Ejemplos útiles:**

```
// Clasificación de notas
IF [Nota] >= 16 THEN "Excelente"
ELSEIF [Nota] >= 14 THEN "Bueno"
ELSEIF [Nota] >= 10 THEN "Aprobado"
ELSE "Reprobado"
END
```

```
// Tasa de aprobación
SUM(IF [Estado] = "Aprobado" THEN 1 ELSE 0 END) / COUNT([Estado])
```

```
// Diferencia con el promedio
[Nota] - WINDOW_AVG(AVG([Nota]))
```

### 4.3 Parámetros

Los parámetros permiten que el usuario controle un valor:

1. Clic derecho en panel de datos → **Crear parámetro**
2. Configurar tipo, rango y valor actual
3. Usar en un campo calculado
4. Mostrar el control del parámetro en el dashboard

**Ejemplo:** Parámetro "Meta" que el usuario puede cambiar, y una línea de referencia que se mueve.

## 5. Dashboards en Tableau

### 5.1 Crear un dashboard

1. Crear varias hojas de trabajo (cada una con un gráfico)
2. Clic en **Nuevo dashboard** (pestaña inferior)
3. Arrastrar hojas desde el panel izquierdo al canvas
4. Organizar el layout

### 5.2 Tamaño y layout

- **Fijo:** Tamaño exacto en píxeles (ej: 1200x800 para presentaciones)
- **Automático:** Se adapta a la pantalla
- **Rango:** Mínimo y máximo (recomendado para publicar en web)

### 5.3 Acciones

Las acciones son las interacciones entre elementos del dashboard:

| Acción | Qué hace | Cómo configurar |
|---|---|---|
| **Filtrar** | Al hacer clic en un gráfico, filtra los demás | Dashboard → Acciones → Agregar acción → Filtrar |
| **Resaltar** | Al hacer clic, resalta elementos relacionados | Dashboard → Acciones → Agregar acción → Resaltar |
| **URL** | Al hacer clic, abre un enlace | Útil para drill-down a reportes detallados |

### 5.4 Objetos adicionales

Además de gráficos, puede agregar:
- **Texto:** Títulos, subtítulos, notas
- **Imagen:** Logos, íconos
- **Página web:** Embeber contenido web
- **Contenedor:** Agrupar elementos horizontalmente o verticalmente

## 6. Comparación: Power BI vs Tableau

| Aspecto | Power BI | Tableau |
|---|---|---|
| **Costo versión gratuita** | Desktop gratuito (solo Windows) | Public gratuito (Win + Mac) |
| **Curva de aprendizaje** | Más fácil para usuarios de Excel | Más intuitivo para visualización |
| **Transformación de datos** | Power Query (muy potente) | Prep (separado) o básico en Desktop |
| **Lenguaje de cálculos** | DAX | Cálculos Tableau (más simple) |
| **Calidad visual** | Buena | Excelente (referente de la industria) |
| **Ecosistema** | Integración Microsoft (Teams, SharePoint, Excel) | Independiente |
| **Publicación** | Power BI Service (requiere licencia Pro) | Tableau Public (gratuito, público) |
| **Comunidad** | Grande, en crecimiento | Muy grande, establecida |
| **Mejor para** | Organizaciones con ecosistema Microsoft | Exploración visual y presentación |

### ¿Cuál elegir?

| Situación | Recomendación |
|---|---|
| La organización usa Microsoft 365 | **Power BI** (integración nativa) |
| Necesita publicar dashboards públicos | **Tableau Public** (gratuito) |
| Énfasis en calidad visual | **Tableau** |
| Transformaciones complejas de datos | **Power BI** (Power Query es superior) |
| Presupuesto limitado | **Ambos** tienen versión gratuita |
| Necesita compartir internamente | **Power BI** (más económico con licencia Microsoft) |

> **Para FACES:** Recomendamos Power BI como herramienta principal (por el ecosistema Microsoft de la universidad) y Tableau Public como complemento para dashboards públicos.

## 7. Ejercicio guiado: Recrear dashboard en Tableau

Vamos a recrear el dashboard de rendimiento académico que construimos en Power BI (sesión 03.02), ahora en Tableau.

### Paso 1: Conectar datos

1. Abrir Tableau Public
2. Conectar → Archivo de texto → `rendimiento_academico.csv`
3. Ir a Hoja de trabajo

### Paso 2: Crear Hoja 1 — Barras por carrera

1. Arrastrar `carrera` a **Filas**
2. Arrastrar `nota` a **Columnas** → clic derecho → Medida → **Promedio**
3. Clic en botón ordenar (descendente)
4. Arrastrar `carrera` a **Color**
5. Arrastrar AVG(nota) a **Etiqueta** → formato: 1 decimal
6. Título: Cambiar a "Promedio de notas por carrera"
7. Renombrar hoja: "Barras Carrera"

### Paso 3: Crear Hoja 2 — Líneas por período

1. Nueva hoja de trabajo
2. Arrastrar `periodo` a **Columnas**
3. Arrastrar `nota` a **Filas** → cambiar a AVG
4. Tipo de marca: **Línea**
5. Agregar línea de referencia: Análisis → Línea de referencia → Constante → 14
6. Título: "Evolución del promedio por período"
7. Renombrar hoja: "Tendencia"

### Paso 4: Crear Hoja 3 — Distribución

1. Nueva hoja de trabajo
2. Arrastrar `nota` a **Columnas** → clic derecho → **Discreto**
3. Arrastrar `Number of Records` (o COUNT) a **Filas**
4. Tipo de marca: **Barra**
5. Título: "Distribución de notas"
6. Renombrar hoja: "Histograma"

### Paso 5: Crear el dashboard

1. Clic en **Nuevo dashboard**
2. Tamaño: 1200 x 800
3. Arrastrar un objeto **Texto** arriba: "Dashboard de Rendimiento Académico — FACES"
4. Arrastrar "Barras Carrera" a la izquierda
5. Arrastrar "Tendencia" a la derecha arriba
6. Arrastrar "Histograma" a la derecha abajo
7. Configurar acción de filtro: Dashboard → Acciones → Agregar acción → Filtrar
   - Origen: Barras Carrera
   - Destino: Todas las hojas del dashboard
8. Probar: clic en una barra de carrera → los otros gráficos se filtran

### Paso 6: Publicar

1. Archivo → Guardar en Tableau Public
2. Iniciar sesión (crear cuenta si no tiene)
3. El dashboard se publica en su perfil de Tableau Public
4. Copiar la URL para compartir

````python
# Referencia visual: así se verían los tres gráficos del dashboard
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../datasets/universidad/rendimiento_academico.csv')

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Gráfico 1: Barras por carrera
if 'carrera' in df.columns and 'nota' in df.columns:
    promedios = df.groupby('carrera')['nota'].mean().sort_values()
    promedios.plot(kind='barh', ax=axes[0], color='#2563eb')
    axes[0].set_title('Promedio por carrera', fontweight='bold')
    axes[0].set_xlabel('Promedio')

# Gráfico 2: Líneas por período
if 'periodo' in df.columns:
    tendencia = df.groupby('periodo')['nota'].mean()
    tendencia.plot(ax=axes[1], marker='o', color='#2563eb', linewidth=2)
    axes[1].axhline(y=14, color='red', linestyle='--', alpha=0.5, label='Meta: 14')
    axes[1].set_title('Tendencia por período', fontweight='bold')
    axes[1].legend()
    axes[1].tick_params(axis='x', rotation=45)

# Gráfico 3: Histograma
df['nota'].hist(ax=axes[2], bins=15, color='#2563eb', edgecolor='white')
axes[2].set_title('Distribución de notas', fontweight='bold')
axes[2].set_xlabel('Nota')
axes[2].set_ylabel('Frecuencia')

plt.tight_layout()
plt.show()
print("\n→ En Tableau, estos 3 gráficos serían interactivos y se filtrarían entre sí.")
````
