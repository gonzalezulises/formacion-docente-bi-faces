## 1. Dato vs Métrica vs KPI

Estos tres conceptos se confunden frecuentemente. Es fundamental distinguirlos:

| Concepto | Definición | Ejemplo FACES |
|---|---|---|
| **Dato** | Valor crudo, sin procesar | "El estudiante Juan obtuvo 15 puntos" |
| **Métrica** | Dato procesado/agregado | "El promedio de notas del semestre es 14.2" |
| **KPI** | Métrica vinculada a un objetivo estratégico | "La tasa de aprobación es 78% (meta: 85%)" |

### Características de un buen KPI

Un KPI debe ser **SMART**:

- **S**pecific (Específico): ¿Qué se mide exactamente?
- **M**easurable (Medible): ¿Se puede cuantificar?
- **A**chievable (Alcanzable): ¿La meta es realista?
- **R**elevant (Relevante): ¿Importa para la toma de decisiones?
- **T**ime-bound (Temporal): ¿En qué período se mide?

### El error más común

Medir todo no es lo mismo que medir bien. Un dashboard con 30 KPIs no ayuda a nadie. La regla:

> **Máximo 5-7 KPIs por dashboard.** Si necesita más, probablemente necesita dashboards separados para diferentes audiencias.

## 2. KPIs académicos

Estos son los indicadores más relevantes para la gestión académica en FACES.

### 2.1 Tasa de deserción

| Aspecto | Detalle |
|---|---|
| **Fórmula** | (Estudiantes que abandonan en período t / Estudiantes inscritos al inicio del período t) x 100 |
| **Interpretación** | Porcentaje de estudiantes que abandonan sus estudios |
| **Fuente de datos** | Sistema de control de estudios |
| **Frecuencia** | Semestral |
| **Meta sugerida** | < 15% (varía por institución) |
| **Alerta** | > 20% requiere intervención inmediata |

### 2.2 Índice de rendimiento académico

| Aspecto | Detalle |
|---|---|
| **Fórmula** | Promedio ponderado de notas (nota x créditos) / Total de créditos |
| **Interpretación** | Rendimiento global considerando la carga académica |
| **Fuente de datos** | Actas de calificaciones |
| **Frecuencia** | Semestral |
| **Meta sugerida** | > 14/20 (escala venezolana) |

### 2.3 Tasa de graduación

| Aspecto | Detalle |
|---|---|
| **Fórmula** | (Graduados en año t / Cohorte de ingreso correspondiente) x 100 |
| **Interpretación** | Porcentaje de una cohorte que logra graduarse |
| **Fuente de datos** | Registro académico |
| **Frecuencia** | Anual |
| **Meta sugerida** | > 60% en el tiempo reglamentario |

### 2.4 Tiempo promedio de egreso

| Aspecto | Detalle |
|---|---|
| **Fórmula** | Promedio de (Fecha de egreso - Fecha de ingreso) de graduados |
| **Interpretación** | Cuánto tarda en promedio un estudiante en graduarse |
| **Fuente de datos** | Registro académico |
| **Frecuencia** | Anual |
| **Meta sugerida** | ≤ duración nominal de la carrera + 1 año |

```python
# Ejemplo: Calcular KPIs académicos con datos reales
import pandas as pd
import numpy as np

df = pd.read_csv('../datasets/universidad/rendimiento_academico.csv')

print("="*60)
print("KPIs ACADÉMICOS — FACES")
print("="*60)

# KPI 1: Tasa de aprobación (proxy de rendimiento)
if 'estado' in df.columns:
    tasa_aprobacion = (df['estado'] == 'Aprobado').mean() * 100
    print(f"\n📊 Tasa de aprobación: {tasa_aprobacion:.1f}%")
    print(f"   Meta: 85% | {'✅ Cumple' if tasa_aprobacion >= 85 else '⚠️ Por debajo de la meta'}")

# KPI 2: Promedio general de notas
if 'nota' in df.columns:
    promedio = df['nota'].mean()
    print(f"\n📊 Promedio general: {promedio:.1f}")
    print(f"   Meta: 14.0 | {'✅ Cumple' if promedio >= 14 else '⚠️ Por debajo de la meta'}")

# KPI 3: Tasa de aprobación por carrera
if 'carrera' in df.columns and 'estado' in df.columns:
    print(f"\n📊 Tasa de aprobación por carrera:")
    tasas = df.groupby('carrera')['estado'].apply(
        lambda x: (x == 'Aprobado').mean() * 100
    ).sort_values(ascending=False)
    for carrera, tasa in tasas.items():
        indicador = '✅' if tasa >= 85 else '⚠️' if tasa >= 70 else '🔴'
        print(f"   {indicador} {carrera}: {tasa:.1f}%")
```

## 3. KPIs organizacionales

Relevantes para la gestión del recurso humano de la facultad.

### 3.1 Rotación de personal

| Aspecto | Detalle |
|---|---|
| **Fórmula** | (Bajas en período t / Promedio de personal en período t) x 100 |
| **Interpretación** | Proporción de personal que se va |
| **Fuente de datos** | RRHH |
| **Frecuencia** | Anual |
| **Meta sugerida** | < 10% en instituciones educativas |

### 3.2 Satisfacción laboral

| Aspecto | Detalle |
|---|---|
| **Fórmula** | Promedio de puntuaciones en encuesta de satisfacción (escala 1-5) |
| **Interpretación** | Nivel de bienestar del personal |
| **Fuente de datos** | Encuesta anual |
| **Frecuencia** | Anual o semestral |
| **Meta sugerida** | > 3.5 / 5.0 |

### 3.3 Costo por empleado

| Aspecto | Detalle |
|---|---|
| **Fórmula** | Gasto total en personal / Número de empleados |
| **Interpretación** | Inversión promedio por persona |
| **Fuente de datos** | Presupuesto + Nómina |
| **Frecuencia** | Mensual o trimestral |

## 4. KPIs socioeconómicos

Indicadores del contexto social y económico que afectan a la comunidad universitaria.

### 4.1 Coeficiente de Gini

| Aspecto | Detalle |
|---|---|
| **Fórmula** | Medida estadística de desigualdad (0 = igualdad perfecta, 1 = desigualdad máxima) |
| **Interpretación** | Nivel de desigualdad en la distribución del ingreso |
| **Fuente de datos** | INE, Banco Mundial |
| **Frecuencia** | Anual |
| **Referencia** | Venezuela ~0.44 (2019), OCDE promedio ~0.32 |

### 4.2 Tasa de empleo

| Aspecto | Detalle |
|---|---|
| **Fórmula** | (Población ocupada / Población económicamente activa) x 100 |
| **Interpretación** | Proporción de la fuerza laboral que tiene empleo |
| **Fuente de datos** | INE, OIT |
| **Frecuencia** | Trimestral |
| **Relevancia FACES** | Empleabilidad de egresados |

### 4.3 Índice de pobreza

| Aspecto | Detalle |
|---|---|
| **Fórmula** | (Hogares bajo línea de pobreza / Total hogares) x 100 |
| **Interpretación** | Proporción de la población en condición de pobreza |
| **Fuente de datos** | ENCOVI, INE |
| **Frecuencia** | Anual |
| **Relevancia FACES** | Contexto socioeconómico de estudiantes |

### 4.4 Inflación

| Aspecto | Detalle |
|---|---|
| **Fórmula** | ((IPC actual - IPC anterior) / IPC anterior) x 100 |
| **Interpretación** | Variación porcentual del nivel general de precios |
| **Fuente de datos** | BCV, Observatorio de Finanzas |
| **Frecuencia** | Mensual |
| **Relevancia FACES** | Impacto en presupuesto y costos operativos |

## 5. KPIs financieros

Indicadores para evaluar la salud financiera y eficiencia en el uso de recursos.

### 5.1 ROI (Retorno sobre la Inversión)

| Aspecto | Detalle |
|---|---|
| **Fórmula** | ((Beneficio - Inversión) / Inversión) x 100 |
| **Interpretación** | Rendimiento por cada unidad monetaria invertida |
| **Ejemplo FACES** | ROI de un programa de tutorías = (ahorro por menor deserción - costo del programa) / costo del programa |
| **Frecuencia** | Por proyecto o anual |

### 5.2 Margen neto

| Aspecto | Detalle |
|---|---|
| **Fórmula** | (Ingresos netos / Ingresos totales) x 100 |
| **Interpretación** | Porcentaje del ingreso que queda después de todos los gastos |
| **Aplicación FACES** | Programas de extensión, cursos de postgrado |
| **Frecuencia** | Mensual o trimestral |

### 5.3 Liquidez

| Aspecto | Detalle |
|---|---|
| **Fórmula** | Activos corrientes / Pasivos corrientes |
| **Interpretación** | Capacidad de pagar obligaciones a corto plazo |
| **Meta** | > 1.0 (idealmente 1.5-2.0) |
| **Frecuencia** | Mensual |

### 5.4 Endeudamiento

| Aspecto | Detalle |
|---|---|
| **Fórmula** | (Pasivos totales / Activos totales) x 100 |
| **Interpretación** | Proporción de activos financiados con deuda |
| **Meta** | < 60% (varía por sector) |
| **Frecuencia** | Trimestral |

```python
# Ejemplo: Calcular KPIs financieros con datos de presupuesto
import pandas as pd

try:
    presupuesto = pd.read_csv('../datasets/universidad/presupuesto_facultad.csv')
    print("Estructura del dataset de presupuesto:")
    print(presupuesto.dtypes)
    print()
    print(presupuesto.head())
    
    # Calcular distribución del presupuesto
    if 'categoria' in presupuesto.columns and 'monto' in presupuesto.columns:
        total = presupuesto['monto'].sum()
        print(f"\nPresupuesto total: {total:,.0f}")
        print("\nDistribución por categoría:")
        dist = presupuesto.groupby('categoria')['monto'].sum().sort_values(ascending=False)
        for cat, monto in dist.items():
            pct = monto / total * 100
            print(f"  {cat}: {monto:,.0f} ({pct:.1f}%)")
except FileNotFoundError:
    print("Dataset de presupuesto no encontrado. Usando datos de ejemplo.")
    print("\nEjemplo de distribución presupuestaria:")
    ejemplo = {
        'Nómina docente': 55,
        'Nómina administrativa': 15,
        'Infraestructura': 12,
        'Investigación': 8,
        'Extensión': 5,
        'Otros': 5
    }
    for cat, pct in ejemplo.items():
        print(f"  {cat}: {pct}%")
```

## 6. Ficha técnica de un KPI

Todo KPI debe tener una ficha técnica que documente cómo se calcula e interpreta. Esto asegura que todos lo midan de la misma manera.

### Plantilla de ficha técnica

| Campo | Descripción |
|---|---|
| **Nombre del KPI** | Nombre claro y descriptivo |
| **Definición** | Qué mide y por qué importa |
| **Fórmula** | Cálculo exacto |
| **Unidad de medida** | %, número absoluto, ratio, moneda |
| **Fuente de datos** | De dónde se obtienen los datos |
| **Frecuencia** | Cada cuánto se actualiza |
| **Responsable** | Quién lo calcula y reporta |
| **Meta** | Valor objetivo |
| **Rango aceptable** | Verde / Amarillo / Rojo |
| **Audiencia** | A quién se reporta |

### Ejemplo completo

| Campo | Valor |
|---|---|
| **Nombre** | Tasa de aprobación por materia |
| **Definición** | Porcentaje de estudiantes que aprueban una materia en un período |
| **Fórmula** | (Aprobados / Total inscritos que presentaron evaluación final) x 100 |
| **Unidad** | Porcentaje (%) |
| **Fuente** | Actas de calificaciones — Control de Estudios |
| **Frecuencia** | Semestral |
| **Responsable** | Coordinación de la Escuela |
| **Meta** | ≥ 80% |
| **Rango** | 🟢 ≥80% / 🟡 60-79% / 🔴 <60% |
| **Audiencia** | Consejo de Escuela, Decanato |
