## 1. ¿Por qué necesitamos storytelling con datos?

Los datos por sí solos rara vez generan acción. Lo que mueve a las personas es una **historia**.

Considere estas dos formas de presentar la misma información:

**Versión A (solo datos):**
> "La tasa de deserción del semestre 2024-II fue 22%. La del semestre anterior fue 18%. La media nacional es 15%."

**Versión B (narrativa con datos):**
> "En el último semestre perdimos a 1 de cada 5 estudiantes. Eso es 220 jóvenes que no volverán a las aulas, 4 puntos más que el semestre anterior y 7 puntos por encima de la media nacional. Si no actuamos, para 2025 la carrera de Economía podría quedar con menos de 100 estudiantes activos."

La versión B usa los mismos números, pero:
- Los traduce a escala humana ("220 jóvenes")
- Establece una tendencia ("4 puntos más")
- Contextualiza ("7 por encima de la media")
- Proyecta consecuencias ("podría quedar con menos de 100")
- Llama a la acción ("si no actuamos")

## 2. Framework de storytelling con datos

Basado en el trabajo de Cole Nussbaumer Knaflic (*Storytelling with Data*), simplificado para contexto universitario.

### Los 3 componentes

```
┌──────────────────────────────────────────────────────┐
│                  STORYTELLING CON DATOS               │
│                                                      │
│   ┌──────────┐   ┌──────────┐   ┌──────────────┐   │
│   │  DATOS   │ + │  VISUAL  │ + │  NARRATIVA   │   │
│   │          │   │          │   │              │   │
│   │ Los      │   │ El       │   │ La historia  │   │
│   │ números  │   │ gráfico  │   │ que conecta  │   │
│   │ correctos│   │ adecuado │   │ ambos        │   │
│   └──────────┘   └──────────┘   └──────────────┘   │
│                                                      │
│   Sin narrativa = reporte técnico                    │
│   Sin datos = opinión                                │
│   Sin visual = informe denso                         │
│   Los tres juntos = comunicación que genera acción   │
└──────────────────────────────────────────────────────┘
```

### Proceso en 6 pasos

1. **Entender el contexto**: ¿Quién es la audiencia? ¿Qué decisión necesitan tomar?
2. **Elegir la visualización adecuada**: Aplicar lo aprendido en la sesión 03.01
3. **Eliminar el desorden**: Quitar todo lo que no contribuya al mensaje
4. **Dirigir la atención**: Usar atributos pre-atentivos para guiar la mirada
5. **Pensar como diseñador**: Forma sigue función, accesibilidad, estética limpia
6. **Contar la historia**: Contexto → Conflicto → Resolución

## 3. Estructura narrativa: Contexto → Conflicto → Resolución

Toda buena historia (incluso con datos) sigue una estructura dramática:

### Contexto (¿Dónde estamos?)

Establece el escenario. Responde:
- ¿Cuál es la situación actual?
- ¿Cuáles son las referencias o metas?
- ¿Qué datos de fondo necesita conocer la audiencia?

**Ejemplo:** "FACES tiene 3,200 estudiantes activos en 5 carreras. La meta institucional de retención es del 85%."

### Conflicto (¿Cuál es el problema u oportunidad?)

Presenta la tensión. Responde:
- ¿Qué está pasando que requiere atención?
- ¿Cómo se compara con la meta o el período anterior?
- ¿Cuál es la magnitud del problema?

**Ejemplo:** "Sin embargo, la deserción ha subido de 15% a 22% en tres semestres. Economía y Contaduría concentran el 70% de las bajas. Cada estudiante que deserta representa una inversión perdida de X bolívares."

### Resolución (¿Qué debemos hacer?)

Propone el camino adelante. Responde:
- ¿Qué sugieren los datos?
- ¿Cuáles son las opciones?
- ¿Cuál es la recomendación?

**Ejemplo:** "Los datos muestran que el 60% de las deserciones ocurren antes del 3er semestre. Un programa de tutorías focalizado en los primeros 3 semestres de Economía y Contaduría podría reducir la deserción en 30%, según la experiencia de la Escuela de Administración que lo implementó en 2023."

### La clave: el "¿y qué?"

Después de cada dato, pregunte: **"¿Y qué?"**

- "La deserción es 22%" → ¿Y qué? → "Es la más alta en 5 años"
- "Es la más alta en 5 años" → ¿Y qué? → "Si continúa, perdemos acreditación"
- "Perdemos acreditación" → ¿Y qué? → "Necesitamos un plan de retención ya"

Siga preguntando hasta llegar a la decisión o acción.

````python
# Ejemplo: Construir una narrativa paso a paso
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('../datasets/universidad/rendimiento_academico.csv')

# Paso 1: Explorar los datos para encontrar la historia
print("Explorando los datos para encontrar la historia...")
print(f"\nTotal de registros: {len(df)}")

if 'carrera' in df.columns and 'nota' in df.columns:
    resumen = df.groupby('carrera').agg(
        promedio=('nota', 'mean'),
        total=('nota', 'count'),
        aprobados=('estado', lambda x: (x == 'Aprobado').sum()) if 'estado' in df.columns else ('nota', 'count')
    ).round(1)
    print("\nResumen por carrera:")
    print(resumen)
````

````python
# Paso 2: Crear un gráfico que cuente la historia
# En vez de un gráfico genérico, uno con narrativa

if 'carrera' in df.columns and 'nota' in df.columns:
    promedios = df.groupby('carrera')['nota'].mean().sort_values()
    
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Colores: resaltar la más baja en rojo, la más alta en verde
    colors = ['#94a3b8'] * len(promedios)
    colors[0] = '#dc2626'   # La más baja
    colors[-1] = '#059669'  # La más alta
    
    bars = ax.barh(promedios.index, promedios.values, color=colors)
    
    # Línea de meta
    ax.axvline(x=14, color='#f59e0b', linestyle='--', linewidth=2, label='Meta: 14.0')
    
    # Título con hallazgo (no descriptivo)
    ax.set_title(f'{promedios.index[0]} tiene el promedio más bajo: {promedios.values[0]:.1f}',
                 fontsize=14, fontweight='bold', pad=20)
    ax.set_xlabel('Promedio de notas')
    
    # Etiquetas de valor
    for bar, val in zip(bars, promedios.values):
        ax.text(val + 0.1, bar.get_y() + bar.get_height()/2, 
                f'{val:.1f}', va='center', fontweight='bold')
    
    ax.legend()
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
    plt.tight_layout()
    plt.show()
````

## 4. Reportes ejecutivos

Un reporte ejecutivo no es un informe técnico. Es un documento breve que facilita la toma de decisiones.

### Qué incluir

1. **Resumen ejecutivo** (1 párrafo): Lo más importante primero
2. **KPIs principales** (3-5 indicadores con semáforo)
3. **Hallazgo clave** (1 gráfico con su interpretación)
4. **Recomendaciones** (2-3 acciones concretas)
5. **Próximos pasos** (quién hace qué y cuándo)

### Qué NO incluir

- Metodología detallada (poner en anexo si es necesario)
- Tablas con muchos números (resumir en gráficos)
- Jerga técnica sin explicar
- Más de 2 páginas (idealmente 1)
- Gráficos decorativos sin mensaje claro

### Estructura visual sugerida

```
┌────────────────────────────────────────────────────┐
│ REPORTE EJECUTIVO — [Título descriptivo]           │
│ Período: [fechas] | Elaborado por: [nombre]        │
├────────────────────────────────────────────────────┤
│                                                    │
│ RESUMEN: [1 párrafo con el hallazgo principal y    │
│ la recomendación clave]                            │
│                                                    │
├──────────┬──────────┬──────────┬──────────────────┤
│ 🟢 KPI 1 │ 🟡 KPI 2 │ 🔴 KPI 3 │ Semáforo visual  │
├──────────┴──────────┴──────────┴──────────────────┤
│                                                    │
│ [1 GRÁFICO CLAVE con título-hallazgo]              │
│                                                    │
├────────────────────────────────────────────────────┤
│ RECOMENDACIONES:                                   │
│ 1. [Acción concreta + responsable + plazo]         │
│ 2. [Acción concreta + responsable + plazo]         │
│ 3. [Acción concreta + responsable + plazo]         │
└────────────────────────────────────────────────────┘
```

## 5. De la tabla al insight

Transformar números en argumentos requiere práctica. Aquí hay técnicas concretas:

### 5.1 Traducir a escala humana

| Dato crudo | Traducción |
|---|---|
| "22% de deserción" | "1 de cada 5 estudiantes abandona" |
| "Incremento de 4 puntos porcentuales" | "40 estudiantes adicionales perdidos" |
| "Promedio de 12.3" | "La mayoría reprueba por menos de 2 puntos" |

### 5.2 Usar comparaciones

| Dato aislado | Con comparación |
|---|---|
| "Tasa de graduación: 45%" | "Mientras la media nacional es 60%, FACES gradúa solo al 45%" |
| "3,200 estudiantes" | "30% menos que hace 5 años" |

### 5.3 Conectar causa-efecto

No solo describir qué pasa, sino **por qué** importa:

- "La matrícula cayó 30%" → **por lo tanto** → "los ingresos por aranceles se redujeron en X"
- "El 60% deserta antes del 3er semestre" → **lo que sugiere** → "el problema está en la adaptación inicial"

### 5.4 Usar el contraste temporal

- **Antes/después:** "Antes del programa: 22% deserción. Después: 14%"
- **Tendencia:** "Tercer semestre consecutivo de descenso"
- **Proyección:** "De continuar esta tendencia, para 2026..."

## 6. Audiencias: adaptar el mensaje

La misma información se presenta de forma diferente según quién la recibe.

### Directivos (Decano, Consejo de Facultad)

- **Quieren:** Decisiones claras, impacto financiero/institucional
- **Formato:** 1 página, KPIs con semáforo, recomendaciones concretas
- **Lenguaje:** Sin tecnicismos estadísticos
- **Ejemplo:** "La inversión en tutorías ($X) podría reducir las pérdidas por deserción ($Y) en un factor de 3:1"

### Técnicos (Coordinadores, analistas)

- **Quieren:** Detalle metodológico, validez de los datos, comparaciones
- **Formato:** 3-5 páginas, tablas detalladas, intervalos de confianza
- **Lenguaje:** Términos estadísticos apropiados
- **Ejemplo:** "La correlación entre asistencia y nota es r=0.67 (p<0.01), controlando por carrera y semestre"

### Público general (Estudiantes, prensa, comunidad)

- **Quieren:** Relevancia personal, impacto social
- **Formato:** Infografía, datos puntuales, historias individuales
- **Lenguaje:** Cotidiano, sin jerga
- **Ejemplo:** "8 de cada 10 egresados de Administración consiguen empleo en menos de 6 meses"
