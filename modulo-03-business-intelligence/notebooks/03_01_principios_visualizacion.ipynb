## 1. ¿Por qué importa la visualización?

Los datos por sí solos no generan decisiones. La visualización es el puente entre el análisis y la acción. Una buena visualización:

- **Reduce la carga cognitiva**: el cerebro procesa imágenes 60,000 veces más rápido que texto
- **Revela patrones** que las tablas ocultan
- **Facilita la comunicación** con audiencias diversas
- **Genera confianza** en los datos cuando se hace correctamente

> *"La visualización de datos es la representación visual de la información para facilitar su comprensión"* — Edward Tufte

## 2. Elección del gráfico según tipo de variable y relación

La primera pregunta antes de crear cualquier gráfico es: **¿qué quiero comunicar?**

### Guía de selección

| Tipo de relación | Variables | Gráfico recomendado | Ejemplo FACES |
|---|---|---|---|
| **Comparación categórica** | 1 categórica + 1 numérica | Barras (horizontales o verticales) | Matrícula por carrera |
| **Evolución temporal** | 1 temporal + 1 numérica | Líneas | Tasa de deserción por semestre |
| **Distribución** | 1 numérica | Histograma, box plot | Distribución de notas |
| **Correlación** | 2 numéricas | Scatter plot | Horas de estudio vs calificación |
| **Composición** | 1 categórica (partes de un todo) | Barras apiladas, treemap | Presupuesto por departamento |
| **Proporción** | 1 categórica (pocas categorías) | Donut (máx 5 segmentos) | Distribución por sexo |
| **Geográfica** | Ubicación + numérica | Mapa coroplético | Procedencia de estudiantes |

### Errores frecuentes en la elección

- **Gráfico de torta con más de 5 categorías**: imposible comparar ángulos similares
- **Barras para series temporales**: se pierde la continuidad del tiempo
- **Scatter sin suficientes puntos**: con menos de 20 observaciones, una tabla es más clara
- **Líneas para datos categóricos sin orden**: las líneas implican continuidad

```python
# Ejemplo: Elección de gráficos con datos de FACES
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Cargar datos de rendimiento académico
df = pd.read_csv('../datasets/universidad/rendimiento_academico.csv')
print(f"Dataset: {df.shape[0]} filas, {df.shape[1]} columnas")
print(f"Columnas: {list(df.columns)}")
df.head()
```

```python
# Demostración: gráfico correcto vs incorrecto
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# INCORRECTO: torta con muchas categorías
if 'carrera' in df.columns:
    conteo = df['carrera'].value_counts()
    axes[0].pie(conteo, labels=conteo.index, autopct='%1.1f%%')
    axes[0].set_title('INCORRECTO: Torta con muchas categorías', color='red', fontweight='bold')

    # CORRECTO: barras horizontales
    conteo.sort_values().plot(kind='barh', ax=axes[1], color='#2563eb')
    axes[1].set_title('CORRECTO: Barras horizontales', color='green', fontweight='bold')
    axes[1].set_xlabel('Cantidad de estudiantes')

plt.tight_layout()
plt.show()
```

## 3. Principios de Gestalt aplicados a visualización

Los principios de la psicología Gestalt explican cómo el cerebro organiza la información visual. Son fundamentales para diseñar visualizaciones que se interpreten correctamente.

### 3.1 Proximidad

Los elementos cercanos se perciben como grupo. En un gráfico, agrupar barras relacionadas comunica que pertenecen a la misma categoría.

**Aplicación:** En un gráfico de barras agrupadas, las barras del mismo grupo deben estar más juntas entre sí que separadas del siguiente grupo.

### 3.2 Similitud

Elementos que comparten características visuales (color, forma, tamaño) se perciben como parte del mismo grupo.

**Aplicación:** Usar el mismo color para la misma serie de datos en todo el dashboard. Si "Administración" es azul en un gráfico, debe ser azul en todos.

### 3.3 Continuidad

El ojo sigue líneas y curvas de forma natural. Las interrupciones rompen el flujo de lectura.

**Aplicación:** En gráficos de líneas, no interrumpir la serie. Si hay datos faltantes, usar línea punteada, no dejar un hueco.

### 3.4 Cierre

El cerebro completa formas incompletas. Podemos sugerir fronteras sin dibujarlas explícitamente.

**Aplicación:** No es necesario enmarcar cada gráfico con bordes gruesos. El espacio en blanco y la alineación son suficientes.

### 3.5 Figura-fondo

Distinguimos elementos principales (figura) del contexto (fondo). El contraste es la clave.

**Aplicación:** Resaltar el dato clave con color saturado; el resto en gris tenue. Esto dirige la atención.

```python
# Ejemplo: Principio de figura-fondo
# Resaltar una categoría específica vs mostrar todas iguales

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

if 'carrera' in df.columns:
    conteo = df['carrera'].value_counts().head(6)
    
    # Sin énfasis: todas las barras iguales
    conteo.plot(kind='bar', ax=axes[0], color='#2563eb')
    axes[0].set_title('Sin énfasis: ¿cuál es el mensaje?', fontweight='bold')
    axes[0].tick_params(axis='x', rotation=45)
    
    # Con énfasis: figura-fondo
    colors = ['#d1d5db'] * len(conteo)
    colors[0] = '#dc2626'  # Resaltar la más alta
    conteo.plot(kind='bar', ax=axes[1], color=colors)
    axes[1].set_title('Con énfasis: la carrera con más matrícula', fontweight='bold')
    axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

## 4. Atributos pre-atentivos (Pre-attentive attributes)

Son características visuales que el cerebro procesa **antes** de la atención consciente (en menos de 250 milisegundos). Son la herramienta más poderosa del diseñador de visualizaciones.

### Atributos principales

| Atributo | Uso recomendado | Ejemplo |
|---|---|---|
| **Color (tono)** | Distinguir categorías | Cada carrera un color distinto |
| **Color (intensidad)** | Mostrar magnitud | Mapa de calor de notas |
| **Tamaño** | Comparar magnitudes | Burbujas proporcionales a matrícula |
| **Posición** | Comparar valores con precisión | Eje Y en gráfico de barras |
| **Orientación** | Distinguir categorías (limitado) | Barras horizontales vs verticales |
| **Forma** | Distinguir categorías (limitado) | Marcadores en scatter: círculo, triángulo |

### Jerarquía de precisión

No todos los atributos son igual de precisos para comparar valores:

1. **Posición** (más preciso) — por eso los gráficos de barras funcionan tan bien
2. **Longitud**
3. **Ángulo / Pendiente**
4. **Área** — el cerebro subestima diferencias de área
5. **Volumen** (menos preciso) — por eso el 3D es problemático
6. **Color (intensidad)** — útil para tendencias, no para valores exactos

> **Regla práctica:** Si necesitas que el usuario compare valores exactos, usa posición (barras). Si solo necesitas que vea tendencias, puedes usar color o tamaño.

```python
# Ejemplo: Atributos pre-atentivos
# El color hace que un punto "salte" inmediatamente

np.random.seed(42)
n = 50
x = np.random.randn(n)
y = np.random.randn(n)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Sin atributo pre-atentivo
axes[0].scatter(x, y, color='#2563eb', s=60)
axes[0].set_title('¿Puede encontrar el punto atípico?', fontweight='bold')
axes[0].set_xlabel('Variable X')
axes[0].set_ylabel('Variable Y')

# Con color como atributo pre-atentivo
colors = ['#d1d5db'] * n
colors[23] = '#dc2626'  # Un punto resaltado
sizes = [60] * n
sizes[23] = 200
axes[1].scatter(x, y, color=colors, s=sizes)
axes[1].set_title('Ahora sí: color + tamaño dirigen la atención', fontweight='bold')
axes[1].set_xlabel('Variable X')
axes[1].set_ylabel('Variable Y')

plt.tight_layout()
plt.show()
```

## 5. Distorsiones comunes en visualización

Una visualización puede ser técnicamente correcta pero **visualmente engañosa**. Estas son las distorsiones más frecuentes:

### 5.1 Ejes truncados

Cuando el eje Y no comienza en cero, diferencias pequeñas parecen enormes.

- **Cuándo es válido truncar:** En gráficos de líneas donde el rango es estrecho y la tendencia importa más que la magnitud absoluta
- **Cuándo NO truncar:** En gráficos de barras (la longitud de la barra representa la magnitud)

### 5.2 Escalas manipuladas

Usar escalas no lineales sin indicarlo, o cambiar la escala entre gráficos que se comparan.

### 5.3 Áreas engañosas

Cuando se usan íconos o burbujas para representar cantidades, el área debe ser proporcional al valor, no el radio. Duplicar el radio cuadruplica el área.

### 5.4 3D innecesario

Los gráficos 3D añaden complejidad visual sin añadir información. La perspectiva distorsiona la percepción de los valores:
- Las barras al fondo parecen más pequeñas
- Los ángulos de la torta se distorsionan
- Se pierde precisión en la lectura

> **Regla:** Si puedes representar lo mismo en 2D, hazlo en 2D.

### 5.5 Colores engañosos

- Usar rojo/verde para positivo/negativo (problemas de daltonismo)
- Paletas de colores que no son perceptualmente uniformes
- Demasiados colores (más de 7 categorías de color)

```python
# Ejemplo: Eje truncado — la misma data, dos historias

valores = [95, 96, 97, 94, 98]
periodos = ['2021-I', '2021-II', '2022-I', '2022-II', '2023-I']

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Eje truncado: parece una montaña rusa
axes[0].bar(periodos, valores, color='#dc2626')
axes[0].set_ylim(93, 99)
axes[0].set_title('ENGAÑOSO: Eje truncado\n"¡Caída dramática en 2022!"', 
                   color='red', fontweight='bold')
axes[0].set_ylabel('Tasa de asistencia (%)')

# Eje completo: la realidad
axes[1].bar(periodos, valores, color='#2563eb')
axes[1].set_ylim(0, 100)
axes[1].set_title('HONESTO: Eje desde cero\n"Asistencia estable y alta"', 
                   color='green', fontweight='bold')
axes[1].set_ylabel('Tasa de asistencia (%)')

plt.tight_layout()
plt.show()
```

## 6. Diseño para audiencias no técnicas

En el contexto universitario, frecuentemente presentamos datos a directivos, consejos y comités que no son especialistas en datos. Principios clave:

### 6.1 Menos es más

- **Un gráfico, un mensaje.** Si el gráfico necesita explicación extensa, probablemente está mal diseñado.
- Eliminar chartjunk: líneas de cuadrícula excesivas, bordes, sombras, fondos de color.
- Ratio tinta-datos de Tufte: maximizar la proporción de tinta que representa datos vs decoración.

### 6.2 Títulos que comunican el hallazgo

| Título débil | Título fuerte |
|---|---|
| "Matrícula por carrera" | "Administración concentra el 35% de la matrícula" |
| "Evolución de deserción" | "La deserción bajó 12% después de las tutorías" |
| "Presupuesto 2024" | "El 60% del presupuesto se destina a nómina" |

### 6.3 Anotaciones y contexto

- Señalar eventos relevantes en la línea de tiempo ("aquí se implementó el programa")
- Incluir metas o benchmarks como líneas de referencia
- Agregar datos clave directamente en el gráfico (etiquetas de datos en barras importantes)

### 6.4 Accesibilidad

- Paletas amigables con daltonismo (azul-naranja en vez de rojo-verde)
- Texto legible (mínimo 12pt en dashboards)
- Contraste suficiente entre elementos

```python
# Ejemplo: Antes y después — diseño para directivos

periodos = ['2020-I', '2020-II', '2021-I', '2021-II', '2022-I', '2022-II', '2023-I']
desercion = [18, 16, 22, 19, 15, 13, 11]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# ANTES: gráfico genérico
axes[0].plot(periodos, desercion, 'bo-')
axes[0].set_title('Deserción por período')
axes[0].set_ylabel('Porcentaje')
axes[0].grid(True)
axes[0].tick_params(axis='x', rotation=45)

# DESPUÉS: diseño orientado a la decisión
axes[1].plot(periodos, desercion, color='#2563eb', linewidth=2.5, marker='o', markersize=8)
axes[1].axhline(y=15, color='#dc2626', linestyle='--', alpha=0.7, label='Meta institucional: 15%')
axes[1].axvline(x=3, color='#059669', linestyle=':', alpha=0.5)
axes[1].annotate('Inicio programa\nde tutorías', xy=(3, 19), xytext=(4, 21),
                 arrowprops=dict(arrowstyle='->', color='#059669'),
                 fontsize=10, color='#059669', fontweight='bold')
axes[1].set_title('La deserción bajó 39% desde el programa de tutorías', 
                   fontweight='bold', fontsize=12)
axes[1].set_ylabel('Tasa de deserción (%)')
axes[1].legend(loc='upper right')
axes[1].tick_params(axis='x', rotation=45)
axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```
