## 3. Python básico rápido

No necesitas ser programador para hacer data wrangling. Estos son los conceptos mínimos que usaremos.

### 3.1 Variables y tipos de datos

````python
# Texto (string)
facultad = "FACES"
carrera = "Administración"

# Números
estudiantes_inscritos = 1250
promedio_general = 14.7

# Booleano (verdadero/falso)
semestre_activo = True

# Verificar tipos
print(type(facultad))             # <class 'str'>
print(type(estudiantes_inscritos)) # <class 'int'>
print(type(promedio_general))      # <class 'float'>
print(type(semestre_activo))       # <class 'bool'>
````

### 3.2 Listas y diccionarios

````python
# Lista: colección ordenada de elementos
carreras = ["Administración", "Contaduría", "Economía"]
print(carreras[0])       # "Administración" (índice empieza en 0)
print(len(carreras))     # 3

# Agregar un elemento
carreras.append("Estadística")
print(carreras)
````

````python
# Diccionario: pares clave-valor (como una ficha de datos)
estudiante = {
    "cedula": "V-25123456",
    "nombre": "María González",
    "carrera": "Contaduría",
    "semestre": 6,
    "promedio": 15.2
}

print(estudiante["nombre"])   # "María González"
print(estudiante["promedio"]) # 15.2

# Recorrer un diccionario
for clave, valor in estudiante.items():
    print(f"{clave}: {valor}")
````

### 3.3 Funciones

Una función es un bloque de código reutilizable:

````python
def calcular_indice_aprobacion(aprobados, total):
    """Calcula el porcentaje de aprobación de una materia."""
    if total == 0:
        return 0
    return round((aprobados / total) * 100, 1)

# Ejemplo: 85 aprobados de 120 inscritos
indice = calcular_indice_aprobacion(85, 120)
print(f"Índice de aprobación: {indice}%")
````

### 3.4 Condicionales y bucles

````python
# Condicional
nota = 12
if nota >= 10:
    print("Aprobado")
else:
    print("Reprobado")

# Bucle for
notas = [15, 8, 12, 18, 6, 14]
aprobados = 0
for nota in notas:
    if nota >= 10:
        aprobados += 1

print(f"Aprobados: {aprobados} de {len(notas)}")
````

## 4. Primer contacto con Pandas

Pandas es LA herramienta para manipular datos tabulares en Python. Un `DataFrame` es como una hoja de cálculo.

````python
# Exploración rápida
print(f"Filas: {len(df)}")
print(f"Columnas: {list(df.columns)}")
print()
df.describe()
````
