## 💻 Sección 1: ¿Qué es Power BI?

### El ecosistema Power BI

| Componente | Descripción | Costo |
|-----------|-------------|-------|
| **Power BI Desktop** | Aplicación Windows para crear reportes | Gratuito |
| **Power BI Service** | Portal web para publicar y compartir | Gratuito (limitado) / Pro ($9.99/mes) |
| **Power BI Mobile** | App móvil para consumir reportes | Gratuito |
| **Power BI Report Server** | Servidor local (on-premise) | Licencia Premium |

### Power BI vs otras herramientas

| Característica | Excel | Power BI | Tableau | Google Looker Studio |
|---------------|-------|----------|---------|---------------------|
| **Costo** | Licencia Office | Desktop gratuito | $70/mes | Gratuito |
| **Tamaño datos** | ~1M filas | ~1B filas | ~1B filas | Depende de fuente |
| **Interactividad** | Limitada | Alta | Alta | Media |
| **Actualización automática** | Manual | Programable | Programable | En tiempo real |
| **Curva de aprendizaje** | Baja | Media | Media-alta | Baja |
| **Modelo de datos** | Tablas dinámicas | Modelo relacional | Modelo relacional | Mezcla de datos |
| **Lenguaje de fórmulas** | Fórmulas Excel | DAX | Campos calculados | N/A |
| **Plataforma** | Windows/Mac | **Solo Windows** | Windows/Mac | Web (cualquiera) |

### Los 3 pilares de Power BI

1. **Power Query** — Para **extraer y transformar** datos (ETL). Equivale a la limpieza de datos que hicimos con pandas.
2. **Modelo de datos** — Para **relacionar tablas** entre sí. Equivale a los JOINs que aprendimos en SQL.
3. **DAX (Data Analysis Expressions)** — Para crear **medidas y cálculos** dinámicos. Equivale a funciones agregadas en SQL o fórmulas en Excel.

## 🛠️ Sección 2: Preparar datos para Power BI

Antes de abrir Power BI, vamos a preparar nuestros datasets usando Python. Esto nos permite:
- Verificar la calidad de los datos
- Limpiar y transformar lo necesario
- Exportar archivos CSV limpios y listos para importar

Prepararemos dos conjuntos de datos:
1. **Presupuesto de la facultad** — Para un dashboard de gestión presupuestaria
2. **Ventas retail** — Para un dashboard de análisis comercial

````python
# ============================================================
# Configuración inicial
# ============================================================

import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings('ignore')

# Crear carpeta para exportaciones si no existe
os.makedirs('exports_power_bi', exist_ok=True)

print("✅ Bibliotecas cargadas")
print(f"📁 Carpeta de exportación: exports_power_bi/")
````

````python
# ============================================================
# Dataset 1: Presupuesto de la facultad
# ============================================================

df_presupuesto = pd.read_csv('../../datasets/universidad/presupuesto_facultad.csv')

print("💰 PRESUPUESTO DE LA FACULTAD")
print(f"   Registros: {len(df_presupuesto):,}")
print(f"   Columnas:  {list(df_presupuesto.columns)}")
print(f"\n   Años disponibles: {sorted(df_presupuesto['anio'].unique())}")
print(f"   Partidas:         {df_presupuesto['partida'].unique().tolist()}")
print(f"   Fuentes:          {df_presupuesto['fuente'].unique().tolist()}")

print("\n   Primeras filas:")
display(df_presupuesto.head())

print("\n   Estadísticas descriptivas:")
display(df_presupuesto[['presupuestado', 'ejecutado']].describe().round(2))
````

````python
# ============================================================
# Limpieza y enriquecimiento del presupuesto
# ============================================================

# Crear columna de fecha completa para Power BI
df_presupuesto['fecha'] = pd.to_datetime(
    df_presupuesto['anio'].astype(str) + '-' + df_presupuesto['mes'].astype(str).str.zfill(2) + '-01'
)

# Calcular la ejecución presupuestaria (% ejecutado vs presupuestado)
df_presupuesto['pct_ejecucion'] = round(
    (df_presupuesto['ejecutado'] / df_presupuesto['presupuestado']) * 100, 1
)

# Calcular la diferencia (sub-ejecución o sobre-ejecución)
df_presupuesto['diferencia'] = round(
    df_presupuesto['ejecutado'] - df_presupuesto['presupuestado'], 2
)

# Clasificar la ejecución
df_presupuesto['clasificacion'] = pd.cut(
    df_presupuesto['pct_ejecucion'],
    bins=[0, 70, 90, 100, float('inf')],
    labels=['Sub-ejecutado', 'En rango', 'Adecuado', 'Sobre-ejecutado']
)

# Añadir nombre del trimestre
df_presupuesto['trimestre'] = 'T' + ((df_presupuesto['mes'] - 1) // 3 + 1).astype(str)

print("✅ Columnas añadidas:")
print(f"   {list(df_presupuesto.columns)}")
print("\n   Muestra con nuevas columnas:")
display(df_presupuesto[['fecha', 'partida', 'presupuestado', 'ejecutado', 
                         'pct_ejecucion', 'diferencia', 'clasificacion', 'trimestre']].head(10))
````

````python
# ============================================================
# Exportar presupuesto limpio
# ============================================================

df_presupuesto.to_csv('exports_power_bi/presupuesto_limpio.csv', index=False, encoding='utf-8-sig')

print("✅ Presupuesto exportado: exports_power_bi/presupuesto_limpio.csv")
print(f"   Registros: {len(df_presupuesto):,}")
print(f"   Columnas:  {len(df_presupuesto.columns)}")
print("\n💡 Nota: Usamos encoding='utf-8-sig' para que Power BI reconozca los acentos correctamente.")
````

````python
# ============================================================
# Dataset 2: Ventas retail
# ============================================================

df_ventas = pd.read_csv('../../datasets/empresarial/ventas_retail.csv')

print("🛍️ VENTAS RETAIL")
print(f"   Registros: {len(df_ventas):,}")
print(f"   Columnas:  {list(df_ventas.columns)}")
print(f"\n   Tiendas:     {sorted(df_ventas['tienda'].unique())}")
print(f"   Regiones:    {df_ventas['region'].unique().tolist()}")
print(f"   Categorías:  {df_ventas['categoria'].unique().tolist()}")

print("\n   Primeras filas:")
display(df_ventas.head())

print("\n   Estadísticas:")
display(df_ventas[['cantidad', 'ingreso', 'costo']].describe().round(2))
````

````python
# ============================================================
# Limpieza y enriquecimiento de ventas
# ============================================================

# Convertir fecha a datetime
df_ventas['fecha'] = pd.to_datetime(df_ventas['fecha'])

# Calcular utilidad (margen bruto)
df_ventas['utilidad'] = round(df_ventas['ingreso'] - df_ventas['costo'], 2)

# Calcular margen porcentual
df_ventas['margen_pct'] = round(
    (df_ventas['utilidad'] / df_ventas['ingreso']) * 100, 1
)

# Extraer componentes de fecha para análisis temporal en Power BI
df_ventas['anio'] = df_ventas['fecha'].dt.year
df_ventas['mes'] = df_ventas['fecha'].dt.month
df_ventas['nombre_mes'] = df_ventas['fecha'].dt.strftime('%B')
df_ventas['dia_semana'] = df_ventas['fecha'].dt.strftime('%A')
df_ventas['trimestre'] = 'T' + df_ventas['fecha'].dt.quarter.astype(str)

# Precio promedio por unidad
df_ventas['precio_unitario'] = round(df_ventas['ingreso'] / df_ventas['cantidad'], 2)

print("✅ Columnas añadidas:")
print(f"   {list(df_ventas.columns)}")
print("\n   Muestra:")
display(df_ventas[['fecha', 'tienda', 'categoria', 'ingreso', 'costo', 
                    'utilidad', 'margen_pct', 'trimestre']].head(10))
````

````python
# ============================================================
# Crear tabla de dimensión: Calendario (para Power BI)
# ============================================================
# Una tabla de calendario es fundamental en Power BI para
# análisis temporal. Contiene una fila por cada día.

fecha_min = df_ventas['fecha'].min()
fecha_max = df_ventas['fecha'].max()

calendario = pd.DataFrame({
    'fecha': pd.date_range(start=fecha_min, end=fecha_max, freq='D')
})
calendario['anio'] = calendario['fecha'].dt.year
calendario['mes'] = calendario['fecha'].dt.month
calendario['nombre_mes'] = calendario['fecha'].dt.strftime('%B')
calendario['trimestre'] = 'T' + calendario['fecha'].dt.quarter.astype(str)
calendario['dia_semana'] = calendario['fecha'].dt.strftime('%A')
calendario['semana_anio'] = calendario['fecha'].dt.isocalendar().week.astype(int)

print("📅 TABLA DE CALENDARIO")
print(f"   Rango: {fecha_min.date()} a {fecha_max.date()}")
print(f"   Días:  {len(calendario):,}")
display(calendario.head(10))
````

````python
# ============================================================
# Exportar todos los archivos para Power BI
# ============================================================

df_ventas.to_csv('exports_power_bi/ventas_limpio.csv', index=False, encoding='utf-8-sig')
calendario.to_csv('exports_power_bi/calendario.csv', index=False, encoding='utf-8-sig')

print("✅ Archivos exportados:")
print("   1. exports_power_bi/presupuesto_limpio.csv")
print("   2. exports_power_bi/ventas_limpio.csv")
print("   3. exports_power_bi/calendario.csv")
print("\n💡 Estos archivos están listos para importar en Power BI Desktop.")
print("   Recuerda usar la codificación UTF-8 al importar.")
````

## 🗂️ Sección 4: Modelo de datos

### ¿Qué es un modelo de datos?

Un modelo de datos define **cómo se relacionan las tablas entre sí**. En Power BI, las relaciones permiten que los filtros se propaguen de una tabla a otra.

### Esquema estrella (Star Schema)

El patrón más común en BI es el **esquema estrella**:

```
                    ┌─────────────┐
                    │ Calendario  │  (Dimensión)
                    │ fecha       │
                    │ anio, mes   │
                    └──────┬──────┘
                           │
                           │ fecha
                           │
┌─────────────┐    ┌─────┬──────┐    ┌─────────────┐
│ Tiendas     │    │   VENTAS     │    │ Categorías   │
│ (Dimensión) │────│   (Hechos)   │────│ (Dimensión)  │
│ tienda      │    │ fecha       │    │ categoria   │
│ region      │    │ tienda      │    └─────────────┘
└─────────────┘    │ categoria   │
                    │ cantidad    │
                    │ ingreso     │
                    │ costo       │
                    └─────────────┘
```

- **Tabla de hechos** (Fact table): Contiene los datos transaccionales (ventas, cantidades, montos)
- **Tablas de dimensión** (Dimension tables): Contienen atributos descriptivos (fecha, tienda, categoría)

### Crear relaciones en Power BI

1. Ve a la vista **"Modelo"** (Model View) en el panel izquierdo (icono de diagrama)
2. Power BI intenta detectar relaciones automáticamente. Verifica que:
   - `ventas_limpio[fecha]` ↔ `calendario[fecha]`
3. Si falta una relación:
   - Arrastra la columna `fecha` de una tabla hacia la columna `fecha` de la otra
   - Configura: Cardinalidad **"Muchos a uno"** (*:1), Dirección del filtro cruzado **"Ambas"**

### ¿Por qué importa el modelo?

Sin relaciones correctas, los filtros (slicers) no funcionan entre tablas. Por ejemplo, si filtras por "T1" en el calendario, necesitas que esa selección se propague a la tabla de ventas para ver solo ventas del primer trimestre.

## 🧮 Sección 5: DAX Básico

**DAX** (Data Analysis Expressions) es el lenguaje de fórmulas de Power BI. Es similar a las fórmulas de Excel pero opera sobre **tablas y columnas** en lugar de celdas.

### Medidas vs Columnas calculadas

| Concepto | Cuándo usarlo | Ejemplo |
|----------|-------------|--------|
| **Medida** | Cálculos dinámicos que cambian con filtros | Total de ventas, promedio |
| **Columna calculada** | Valor fijo por cada fila | Categoría de margen (alto/bajo) |

> **Regla general:** Usa **medidas** el 90% del tiempo. Las columnas calculadas consumen más memoria.

### Cómo crear una medida

1. En la vista de **Reporte**, selecciona la tabla `ventas_limpio` en el panel derecho
2. Pestaña **Modelado** (Modeling) → **"Nueva medida"** (New Measure)
3. Escribe la fórmula DAX en la barra de fórmulas superior

### Medidas esenciales para nuestros datos

#### Medida 1: Total de ventas

```dax
Total Ventas = SUM(ventas_limpio[ingreso])
```

Esta medida suma todos los ingresos. Cuando aplicas un filtro (por ejemplo, seleccionas "Alimentos" en un slicer), la medida automáticamente recalcula solo para esa categoría.

#### Medida 2: Total de costos

```dax
Total Costos = SUM(ventas_limpio[costo])
```

#### Medida 3: Utilidad total

```dax
Utilidad Total = [Total Ventas] - [Total Costos]
```

Observa que puedes **referenciar otras medidas** usando corchetes.

#### Medida 4: Margen porcentual

```dax
Margen % = DIVIDE([Utilidad Total], [Total Ventas], 0)
```

`DIVIDE` es preferible a usar `/` porque maneja automáticamente la división entre cero (devuelve 0 en lugar de error).

#### Medida 5: Cantidad de transacciones

```dax
Transacciones = COUNTROWS(ventas_limpio)
```

#### Medida 6: Promedio móvil de ventas (30 días)

```dax
Promedio Movil 30D = 
AVERAGEX(
    DATESINPERIOD(
        calendario[fecha],
        MAX(calendario[fecha]),
        -30,
        DAY
    ),
    [Total Ventas]
)
```

Esta medida es más avanzada. Calcula el promedio de ventas de los últimos 30 días desde cualquier punto seleccionado.

### Funciones DAX esenciales

| Función | Descripción | Ejemplo |
|---------|-----------|--------|
| `SUM()` | Suma una columna | `SUM(tabla[columna])` |
| `AVERAGE()` | Promedio de una columna | `AVERAGE(tabla[columna])` |
| `COUNT()` | Cuenta valores no vacíos | `COUNT(tabla[columna])` |
| `COUNTROWS()` | Cuenta filas de una tabla | `COUNTROWS(tabla)` |
| `DISTINCTCOUNT()` | Cuenta valores únicos | `DISTINCTCOUNT(tabla[columna])` |
| `DIVIDE()` | División segura | `DIVIDE(numerador, denominador, alternativo)` |
| `CALCULATE()` | Cambia el contexto de filtro | `CALCULATE([Medida], filtro)` |
| `FILTER()` | Filtra una tabla | `FILTER(tabla, condicion)` |
| `ALL()` | Ignora todos los filtros | `ALL(tabla)` |

### CALCULATE: La función más poderosa de DAX

`CALCULATE` modifica el **contexto de filtro** de una medida. Ejemplo:

```dax
Ventas Alimentos = 
CALCULATE(
    [Total Ventas],
    ventas_limpio[categoria] = "Alimentos"
)
```

Esta medida siempre muestra las ventas de Alimentos, sin importar qué categoría esté seleccionada en los filtros.

```dax
% del Total General = 
DIVIDE(
    [Total Ventas],
    CALCULATE([Total Ventas], ALL(ventas_limpio)),
    0
)
```

Aquí `ALL()` dentro de `CALCULATE` ignora todos los filtros, dando el total general. Así puedes calcular qué porcentaje del total representa cada segmento.

## 🎨 Sección 6: Crear visualizaciones

### Panel de visualizaciones

En la vista de **Reporte**, el panel derecho tiene tres secciones:
1. **Visualizaciones** — Tipos de gráfico disponibles
2. **Campos** — Columnas y medidas de tus tablas
3. **Formato** — Opciones de diseño (colores, títulos, etc.)

### Guía de visualizaciones principales

#### 1. Gráfico de barras (Bar Chart)

**Uso ideal:** Comparar valores entre categorías

**Pasos:**
1. Haz clic en el icono de **gráfico de barras agrupadas** en el panel de visualizaciones
2. Arrastra `categoria` al eje Y
3. Arrastra `ingreso` (o la medida `Total Ventas`) al eje X
4. Opcionalmente, arrastra `region` a la **Leyenda** para ver barras agrupadas por región

**Tips:**
- Ordena las barras por valor (clic en los tres puntos → "Ordenar por")
- Usa colores consistentes: un solo color si no hay subcategorías

---

#### 2. Gráfico de líneas (Line Chart)

**Uso ideal:** Mostrar tendencias en el tiempo

**Pasos:**
1. Haz clic en el icono de **gráfico de líneas**
2. Arrastra `fecha` al eje X
3. Arrastra `Total Ventas` al eje Y
4. Power BI automáticamente agregará una jerarquía de tiempo (Año > Trimestre > Mes > Día)

**Tips:**
- Haz clic derecho en la jerarquía de fecha para elegir el nivel de detalle
- Agrega una línea de tendencia: Formato → Analítica → Línea de tendencia

---

#### 3. Gráfico circular (Pie Chart)

**Uso ideal:** Composición (partes de un todo) con **pocas categorías** (3-5)

**Pasos:**
1. Haz clic en el icono de **gráfico circular**
2. Arrastra `region` a **Leyenda**
3. Arrastra `Total Ventas` a **Valores**

**Tips:**
- Evita gráficos circulares con más de 5 categorías (usa barras en su lugar)
- Muestra etiquetas de porcentaje: Formato → Etiquetas de detalle

---

#### 4. Tarjeta (Card)

**Uso ideal:** Mostrar un KPI único y destacado (total ventas, margen, transacciones)

**Pasos:**
1. Haz clic en el icono de **tarjeta**
2. Arrastra la medida `Total Ventas` al campo **Valor**
3. Formatea el número: Formato → Etiqueta de datos → Formato de visualización → Moneda o Número

**Tips:**
- Usa tarjetas múltiples (Multi-row card) para mostrar varios KPIs juntos
- Agrega un título descriptivo: "Ventas Totales (2023)"

---

#### 5. Segmentador (Slicer)

**Uso ideal:** Filtro visual interactivo que afecta a todas las demás visualizaciones

**Pasos:**
1. Haz clic en el icono de **segmentador de datos**
2. Arrastra `categoria` o `region` al campo **Valor**
3. Elige el estilo: lista, desplegable o botones

**Tips:**
- Coloca slicers en la parte superior o lateral del dashboard
- Usa slicers de fecha para filtrar períodos
- "Ctrl + clic" permite seleccionar múltiples valores

---

#### 6. Mapa (Map)

**Uso ideal:** Datos geográficos (ventas por región/estado/país)

**Pasos:**
1. Haz clic en el icono de **mapa** (o mapa con burbujas)
2. Arrastra `region` a **Ubicación**
3. Arrastra `Total Ventas` a **Tamaño**

**Tips:**
- Los mapas requieren datos geográficos reconocibles (países, estados, ciudades)
- Para datos internos (como nuestras regiones), un gráfico de barras puede ser más claro

## 📰 Sección 7: Diseñar un dashboard

### Principios de diseño

Un buen dashboard no es solo un conjunto de gráficos. Sigue estos principios:

#### 1. Jerarquía visual

```
┌──────────────────────────────────────────────────┐
│  TÍTULO DEL DASHBOARD     [Filtro 1] [Filtro 2]  │
├────────┬────────┬────────┬────────┬─────────────┤
│ KPI 1  │ KPI 2  │ KPI 3  │ KPI 4  │             │  ← Zona superior:
│ $125K  │ 32.5%  │ 5,039  │  10    │             │    Métricas clave
├────────┴────────┴────────┼────────┴─────────────┤
│                           │                     │
│   Gráfico principal        │  Gráfico secundario │  ← Zona media:
│   (tendencia temporal)     │  (composición)      │    Análisis principal
│                           │                     │
├───────────────────────────┼─────────────────────┤
│                           │                     │
│   Gráfico de detalle       │   Tabla resumen     │  ← Zona inferior:
│   (barras por categoría)  │                     │    Detalle
│                           │                     │
└───────────────────────────┴─────────────────────┘
```

#### 2. Paleta de colores

- Usa **máximo 5-6 colores** en todo el dashboard
- Colores institucionales de FACES/UC: azul, dorado, blanco
- Paleta sugerida: `#1F4E79` (azul oscuro), `#2E75B6` (azul medio), `#4BACC6` (cian), `#F4B942` (dorado), `#E74C3C` (rojo para alertas)
- Usa rojo/verde solo para indicar bueno/malo (con precaución por daltonismo)

#### 3. Interactividad

- **Cross-filtering:** Al hacer clic en una barra de un gráfico, los demás gráficos se filtran automáticamente
- Para desactivar cross-filtering entre dos visualizaciones específicas: selecciona la visualización fuente → Formato → Editar interacciones → elige "Ninguno" en la visualización destino

#### 4. Marcadores (Bookmarks)

Los marcadores guardan el estado actual del dashboard (filtros aplicados, página visible) para crear **navegación tipo presentación**:

1. Pestaña **Vista** (View) → **Panel de marcadores**
2. Configura los filtros y visualizaciones como quieras
3. Haz clic en **"Añadir"** en el panel de marcadores
4. Crea un botón y asígnale la acción "Marcador" para navegar al hacer clic

#### 5. Regla de los 5 segundos

> Un usuario debe entender el mensaje principal del dashboard en **5 segundos o menos**.

Si necesita más tiempo, simplifica: menos gráficos, títulos más claros, KPIs más grandes.

### Ejercicio 1: Dashboard "Gestión Presupuestaria FACES"

Usando el archivo `presupuesto_limpio.csv`, crea un dashboard en Power BI con:

**KPIs (tarjetas):**
- Total presupuestado (suma de toda la columna `presupuestado`)
- Total ejecutado (suma de `ejecutado`)
- Porcentaje de ejecución global

**Visualizaciones:**
1. Gráfico de líneas: Presupuestado vs Ejecutado por mes (dos líneas en el mismo gráfico)
2. Gráfico de barras: Ejecución por partida (Personal, Materiales, Servicios, etc.)
3. Gráfico circular: Distribución del presupuesto por fuente (Ordinario, Propio, Especial)
4. Tabla con formato condicional: Partida | Presupuestado | Ejecutado | % Ejecución (colorear en rojo < 70%, verde > 90%)

**Filtros:**
- Segmentador por año
- Segmentador por partida

**Medidas DAX sugeridas:**

```dax
Total Presupuestado = SUM(presupuesto_limpio[presupuestado])
Total Ejecutado = SUM(presupuesto_limpio[ejecutado])
Ejecucion % = DIVIDE([Total Ejecutado], [Total Presupuestado], 0)
```

````python
# ============================================================
# Verificación: Resumen de datos del presupuesto
# ============================================================
# Ejecuta esta celda para tener los números de referencia
# que deberías ver en tu dashboard de Power BI.

print("💰 VALORES ESPERADOS EN EL DASHBOARD DE PRESUPUESTO")
print("=" * 55)

total_presup = df_presupuesto['presupuestado'].sum()
total_ejec = df_presupuesto['ejecutado'].sum()
pct_ejec = (total_ejec / total_presup) * 100

print(f"\n   Total Presupuestado: ${total_presup:,.2f}")
print(f"   Total Ejecutado:     ${total_ejec:,.2f}")
print(f"   % Ejecución:         {pct_ejec:.1f}%")

print("\n   Por partida:")
resumen = df_presupuesto.groupby('partida').agg(
    presupuestado=('presupuestado', 'sum'),
    ejecutado=('ejecutado', 'sum')
).assign(
    pct_ejecucion=lambda x: round((x['ejecutado'] / x['presupuestado']) * 100, 1)
).sort_values('presupuestado', ascending=False)

display(resumen)

print("\n   Por fuente:")
resumen_fuente = df_presupuesto.groupby('fuente')['presupuestado'].sum()
for fuente, monto in resumen_fuente.items():
    pct = monto / total_presup * 100
    print(f"   - {fuente}: ${monto:,.2f} ({pct:.1f}%)")
````

### Ejercicio 2: Dashboard "Ventas Retail"

Usando los archivos `ventas_limpio.csv` y `calendario.csv`, crea un dashboard con:

**KPIs:**
- Total Ventas
- Utilidad Total
- Margen %
- Número de transacciones

**Visualizaciones:**
1. Línea: Tendencia de ventas mensuales
2. Barras horizontales: Ventas por categoría (ordenadas de mayor a menor)
3. Barras agrupadas: Ventas por región, con colores por categoría
4. Tabla: Top 10 tiendas por ventas

**Filtros:**
- Segmentador de fecha (rango)
- Segmentador por categoría
- Segmentador por región

````python
# ============================================================
# Verificación: Resumen de datos de ventas
# ============================================================

print("🛍️ VALORES ESPERADOS EN EL DASHBOARD DE VENTAS")
print("=" * 55)

total_ventas = df_ventas['ingreso'].sum()
total_costos = df_ventas['costo'].sum()
utilidad = total_ventas - total_costos
margen = (utilidad / total_ventas) * 100
transacciones = len(df_ventas)

print(f"\n   Total Ventas:      ${total_ventas:,.2f}")
print(f"   Total Costos:      ${total_costos:,.2f}")
print(f"   Utilidad:          ${utilidad:,.2f}")
print(f"   Margen:            {margen:.1f}%")
print(f"   Transacciones:     {transacciones:,}")

print("\n   Ventas por categoría:")
ventas_cat = df_ventas.groupby('categoria')['ingreso'].sum().sort_values(ascending=False)
for cat, monto in ventas_cat.items():
    pct = monto / total_ventas * 100
    print(f"   - {cat}: ${monto:,.2f} ({pct:.1f}%)")

print("\n   Ventas por región:")
ventas_reg = df_ventas.groupby('region')['ingreso'].sum().sort_values(ascending=False)
for reg, monto in ventas_reg.items():
    pct = monto / total_ventas * 100
    print(f"   - {reg}: ${monto:,.2f} ({pct:.1f}%)")
````
