## 👁️ Sección 1: ¿Por qué visualizar?

### El Cuarteto de Anscombe

En 1973, el estadístico **Francis Anscombe** creó 4 conjuntos de datos con **exactamente las mismas estadísticas** (media, varianza, correlación, regresión lineal) pero **distribuciones completamente diferentes**. Solo la visualización revela la diferencia.

````python
# ============================================================
# El Cuarteto de Anscombe: mismas estadísticas, formas diferentes
# ============================================================

# Los 4 conjuntos de datos de Anscombe
anscombe = sns.load_dataset('anscombe')

# Verificar que las estadísticas son (casi) idénticas
print('📊 Estadísticas de los 4 conjuntos (son prácticamente iguales):')
print('=' * 60)
for dataset in ['I', 'II', 'III', 'IV']:
    datos = anscombe[anscombe['dataset'] == dataset]
    print(f'  Dataset {dataset}: media_x={datos["x"].mean():.1f}, '
          f'media_y={datos["y"].mean():.2f}, '
          f'corr={datos["x"].corr(datos["y"]):.3f}')
````

````python
# ============================================================
# Pero al visualizarlos... ¡son completamente diferentes!
# ============================================================

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

for i, (dataset, ax) in enumerate(zip(['I', 'II', 'III', 'IV'], axes.flat)):
    datos = anscombe[anscombe['dataset'] == dataset]
    ax.scatter(datos['x'], datos['y'], color='steelblue', s=60, edgecolor='white')
    
    # Añadir línea de regresión (misma para los 4)
    m, b = np.polyfit(datos['x'], datos['y'], 1)
    x_line = np.linspace(datos['x'].min(), datos['x'].max(), 100)
    ax.plot(x_line, m * x_line + b, color='red', linestyle='--', alpha=0.7)
    
    ax.set_title(f'Dataset {dataset}', fontweight='bold')
    ax.set_xlabel('x')
    ax.set_ylabel('y')

plt.suptitle('Cuarteto de Anscombe: Mismas estadísticas, formas diferentes', 
             fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print('\n📌 Lección: Las estadísticas descriptivas solas NO son suficientes.')
print('   Siempre hay que VISUALIZAR los datos antes de sacar conclusiones.')
````

## 📰 Sección 2: Matplotlib básico

**Matplotlib** es la biblioteca base de visualización en Python. Todo lo demás (seaborn, plotly) se construye sobre ella.

### Anatomía de una figura

```
Figure (lienzo completo)
└── Axes (el gráfico individual)
    ├── Title (título)
    ├── X-axis (eje X con etiqueta)
    ├── Y-axis (eje Y con etiqueta)
    ├── Plot area (el gráfico en sí)
    └── Legend (leyenda)
```

````python
# ============================================================
# Gráfico de línea: tendencia de matrícula por período
# ============================================================

# Contar estudiantes por período
matricula_por_periodo = df_matricula.groupby('periodo').size().reset_index(name='estudiantes')
# Ordenar por período
matricula_por_periodo = matricula_por_periodo.sort_values('periodo')

# Crear el gráfico
plt.figure(figsize=(14, 5))
plt.plot(matricula_por_periodo['periodo'], 
         matricula_por_periodo['estudiantes'],
         color='steelblue', linewidth=2, marker='o', markersize=5)

plt.xlabel('Período académico')
plt.ylabel('Número de estudiantes')
plt.title('Evolución de la Matrícula FACES por Período')
plt.xticks(rotation=45, ha='right')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print('\n📌 Elementos del gráfico: plt.plot() para la línea, plt.xlabel/ylabel para ejes,')
print('   plt.title para título, plt.xticks para rotar etiquetas, plt.show() para mostrar.')
````

## 📊 Sección 3: Gráficos de barras

Los gráficos de barras son ideales para **comparar cantidades entre categorías**.

````python
# ============================================================
# Gráfico de barras: estudiantes por carrera
# ============================================================

conteo_carrera = df_rendimiento['carrera'].value_counts()

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Barras verticales con Matplotlib
axes[0].bar(conteo_carrera.index, conteo_carrera.values, 
            color=['steelblue', 'coral', 'seagreen', 'mediumpurple'],
            edgecolor='white')
axes[0].set_xlabel('Carrera')
axes[0].set_ylabel('Número de estudiantes')
axes[0].set_title('Estudiantes por Carrera (barras verticales)')
axes[0].tick_params(axis='x', rotation=15)

# Barras horizontales para comparar promedios
promedio_carrera = df_rendimiento.groupby('carrera')['promedio'].mean().sort_values()
axes[1].barh(promedio_carrera.index, promedio_carrera.values, 
             color=['mediumpurple', 'seagreen', 'coral', 'steelblue'],
             edgecolor='white')
axes[1].set_xlabel('Promedio académico')
axes[1].set_title('Promedio Académico por Carrera')

# Añadir valores en las barras
for i, v in enumerate(promedio_carrera.values):
    axes[1].text(v + 0.05, i, f'{v:.2f}', va='center', fontweight='bold')

plt.tight_layout()
plt.show()
````

````python
# ============================================================
# Barplot con seaborn (incluye barras de error automáticas)
# ============================================================

fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(data=df_rendimiento, x='carrera', y='promedio', 
            palette='Set2', ci=95, ax=ax)
ax.set_xlabel('Carrera')
ax.set_ylabel('Promedio académico')
ax.set_title('Promedio por Carrera con Intervalos de Confianza (95%)')
ax.tick_params(axis='x', rotation=15)
plt.tight_layout()
plt.show()

print('\n📌 Seaborn calcula automáticamente los intervalos de confianza')
print('   (las líneas negras sobre cada barra indican la incertidumbre)')
````

## 📈 Sección 4: Gráficos de línea

Los gráficos de línea son ideales para **series temporales**: datos que cambian a lo largo del tiempo.

````python
# ============================================================
# Serie temporal: presupuesto vs ejecutado
# ============================================================

# Agregar presupuesto por año y mes (total de todas las partidas)
pres_mensual = df_presupuesto.groupby(['anio', 'mes']).agg(
    presupuestado=('presupuestado', 'sum'),
    ejecutado=('ejecutado', 'sum')
).reset_index()

# Crear columna de fecha para el eje X
pres_mensual['fecha'] = pd.to_datetime(
    pres_mensual['anio'].astype(str) + '-' + pres_mensual['mes'].astype(str) + '-01'
)
pres_mensual = pres_mensual.sort_values('fecha')

# Graficar
fig, ax = plt.subplots(figsize=(14, 6))

ax.plot(pres_mensual['fecha'], pres_mensual['presupuestado'], 
        color='steelblue', linewidth=2, label='Presupuestado', alpha=0.8)
ax.plot(pres_mensual['fecha'], pres_mensual['ejecutado'], 
        color='coral', linewidth=2, label='Ejecutado', alpha=0.8)

ax.fill_between(pres_mensual['fecha'], 
                pres_mensual['presupuestado'], 
                pres_mensual['ejecutado'],
                alpha=0.1, color='gray')

ax.set_xlabel('Fecha')
ax.set_ylabel('Monto (Bs.)')
ax.set_title('Presupuesto vs Ejecución — FACES (mensual)')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print('\n📌 La zona sombreada entre ambas líneas muestra la brecha')
print('   entre lo presupuestado y lo realmente ejecutado.')
````

## ⭕ Sección 5: Scatter plots (gráficos de dispersión)

Los scatter plots muestran la **relación entre dos variables numéricas**. Cada punto representa una observación.

````python
# ============================================================
# Scatter plot: asistencia vs promedio, por carrera
# ============================================================

fig, ax = plt.subplots(figsize=(12, 7))

colores = {'Economia': 'steelblue', 'Contaduria': 'coral', 
           'Administracion': 'seagreen', 'Relaciones Industriales': 'mediumpurple'}

for carrera, color in colores.items():
    datos = df_rendimiento[df_rendimiento['carrera'] == carrera]
    ax.scatter(datos['asistencia_pct'], datos['promedio'], 
              c=color, label=carrera, alpha=0.5, s=30, edgecolor='white')

ax.set_xlabel('Asistencia (%)')
ax.set_ylabel('Promedio académico')
ax.set_title('Relación entre Asistencia y Promedio Académico')
ax.legend(title='Carrera', loc='lower right')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Calcular la correlación
corr = df_rendimiento['asistencia_pct'].corr(df_rendimiento['promedio'])
print(f'\n📌 Correlación entre asistencia y promedio: {corr:.3f}')
print('   Cada punto es un estudiante. El color indica su carrera.')
````

## 📊 Sección 6: Histogramas y distribuciones

Los histogramas muestran **cómo se distribuyen** los valores de una variable.

````python
# ============================================================
# Histogramas con Matplotlib y Seaborn
# ============================================================

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# Matplotlib básico
axes[0].hist(df_rendimiento['promedio'], bins=30, color='steelblue', 
             edgecolor='white', alpha=0.8)
axes[0].set_title('plt.hist() básico')
axes[0].set_xlabel('Promedio')
axes[0].set_ylabel('Frecuencia')

# Seaborn histplot con KDE
sns.histplot(df_rendimiento['promedio'], bins=30, kde=True, 
             color='coral', ax=axes[1])
axes[1].set_title('sns.histplot() + KDE')
axes[1].set_xlabel('Promedio')

# Seaborn KDE solo (distribución suavizada)
for carrera in df_rendimiento['carrera'].unique():
    datos = df_rendimiento[df_rendimiento['carrera'] == carrera]['promedio']
    sns.kdeplot(datos, label=carrera, ax=axes[2], linewidth=2)
axes[2].set_title('sns.kdeplot() por carrera')
axes[2].set_xlabel('Promedio')
axes[2].legend(fontsize=9)

plt.suptitle('Tres formas de visualizar distribuciones', fontsize=14, 
             fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
````

````python
# ============================================================
# Distribución de salarios del dataset de nómina
# ============================================================

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histograma de salarios
sns.histplot(df_nomina['salario_mensual_usd'], bins=30, kde=True, 
             color='seagreen', ax=axes[0])
axes[0].axvline(df_nomina['salario_mensual_usd'].mean(), color='red', 
                linestyle='--', label=f'Media: ${df_nomina["salario_mensual_usd"].mean():.0f}')
axes[0].axvline(df_nomina['salario_mensual_usd'].median(), color='orange', 
                linestyle='--', label=f'Mediana: ${df_nomina["salario_mensual_usd"].median():.0f}')
axes[0].set_title('Distribución de Salarios')
axes[0].set_xlabel('Salario mensual (USD)')
axes[0].legend()

# KDE por departamento (top 4)
top_deptos = df_nomina['departamento'].value_counts().head(4).index
for depto in top_deptos:
    datos = df_nomina[df_nomina['departamento'] == depto]['salario_mensual_usd']
    sns.kdeplot(datos, label=depto, ax=axes[1], linewidth=2)
axes[1].set_title('Distribución salarial por departamento')
axes[1].set_xlabel('Salario mensual (USD)')
axes[1].legend(fontsize=9)

plt.tight_layout()
plt.show()
````

## 🗺️ Sección 8: Heatmaps

Los heatmaps usan **color** para representar valores en una matriz. Son ideales para **matrices de correlación** y **tablas cruzadas**.

````python
# ============================================================
# Heatmap de correlación con estilo profesional
# ============================================================

# Calcular correlaciones de las variables numéricas de rendimiento
cols_num = ['semestre', 'edad', 'promedio', 'materias_inscritas', 
            'materias_aprobadas', 'materias_reprobadas', 'asistencia_pct']
correlaciones = df_rendimiento[cols_num].corr()

# Máscara para el triángulo superior
mask = np.triu(np.ones_like(correlaciones, dtype=bool))

fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(correlaciones, mask=mask, annot=True, fmt='.2f',
            cmap='RdBu_r', center=0, vmin=-1, vmax=1,
            square=True, linewidths=0.5, ax=ax,
            cbar_kws={'label': 'Coeficiente de correlación', 'shrink': 0.8})

ax.set_title('Matriz de Correlación \u2014 Rendimiento Académico FACES', fontsize=14)
plt.tight_layout()
plt.show()

print('\n📌 Colores rojos = correlación positiva, azules = negativa')
print('   El número en cada celda indica la fuerza (-1 a +1)')
````

## 📰 Sección 9: Subplots y figuras compuestas

Las figuras compuestas permiten mostrar **múltiples perspectivas** de un mismo tema en una sola imagen.

````python
# ============================================================
# Figura compuesta: 4 vistas del rendimiento académico
# ============================================================

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Panel 1: Distribución del promedio
sns.histplot(df_rendimiento['promedio'], bins=30, kde=True, 
             color='steelblue', ax=axes[0, 0])
axes[0, 0].set_title('Distribución del Promedio')
axes[0, 0].set_xlabel('Promedio')

# Panel 2: Boxplot por carrera
sns.boxplot(data=df_rendimiento, x='carrera', y='promedio', 
            palette='Set2', ax=axes[0, 1])
axes[0, 1].set_title('Promedio por Carrera')
axes[0, 1].tick_params(axis='x', rotation=15)

# Panel 3: Scatter asistencia vs promedio
sns.scatterplot(data=df_rendimiento, x='asistencia_pct', y='promedio',
                hue='carrera', alpha=0.4, s=20, ax=axes[1, 0], legend=False)
axes[1, 0].set_title('Asistencia vs Promedio')
axes[1, 0].set_xlabel('Asistencia (%)')
axes[1, 0].set_ylabel('Promedio')

# Panel 4: Conteo por carrera y género
ct = pd.crosstab(df_rendimiento['carrera'], df_rendimiento['genero'])
ct.plot(kind='bar', ax=axes[1, 1], color=['coral', 'steelblue'], edgecolor='white')
axes[1, 1].set_title('Estudiantes por Carrera y Género')
axes[1, 1].set_xlabel('Carrera')
axes[1, 1].set_ylabel('Cantidad')
axes[1, 1].tick_params(axis='x', rotation=15)
axes[1, 1].legend(title='Género')

plt.suptitle('Dashboard: Rendimiento Académico FACES', 
             fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
````

## ✨ Sección 10: Personalización y figuras de calidad

Para presentaciones académicas y publicaciones, necesitamos figuras **pulidas y profesionales**.

````python
# ============================================================
# Estilos disponibles en Matplotlib
# ============================================================

estilos = ['seaborn-v0_8-whitegrid', 'seaborn-v0_8-darkgrid', 
           'ggplot', 'fivethirtyeight']

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

for ax, estilo in zip(axes.flat, estilos):
    with plt.style.context(estilo):
        datos = df_rendimiento.groupby('carrera')['promedio'].mean()
        ax.bar(datos.index, datos.values, color=['steelblue', 'coral', 'seagreen', 'mediumpurple'])
        ax.set_title(f'Estilo: {estilo}', fontsize=12)
        ax.set_ylabel('Promedio')
        ax.tick_params(axis='x', rotation=20, labelsize=9)

plt.suptitle('Comparación de estilos de Matplotlib', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
````

````python
# ============================================================
# Figura pulida: calidad de publicación académica
# ============================================================

# Preparar datos
resumen = df_rendimiento.groupby('carrera').agg(
    promedio=('promedio', 'mean'),
    std=('promedio', 'std'),
    n=('promedio', 'count')
).sort_values('promedio', ascending=True)

# Error estándar para barras de error
resumen['se'] = resumen['std'] / np.sqrt(resumen['n'])

# Colores personalizados (paleta profesional)
colores = ['#2196F3', '#4CAF50', '#FF9800', '#9C27B0']

# Crear la figura
fig, ax = plt.subplots(figsize=(10, 6))

barras = ax.barh(range(len(resumen)), resumen['promedio'], 
                 xerr=resumen['se'] * 1.96,  # IC 95%
                 color=colores, edgecolor='white', linewidth=1.5,
                 capsize=5, error_kw={'linewidth': 1.5})

# Etiquetas y formato
ax.set_yticks(range(len(resumen)))
ax.set_yticklabels(resumen.index, fontsize=12)
ax.set_xlabel('Promedio académico (escala 0-20)', fontsize=12)
ax.set_title('Promedio Académico por Carrera \u2014 FACES, UC', 
             fontsize=14, fontweight='bold', pad=20)

# Añadir valores en las barras
for i, (valor, se) in enumerate(zip(resumen['promedio'], resumen['se'])):
    ax.text(valor + se * 1.96 + 0.1, i, f'{valor:.2f}', 
            va='center', fontsize=11, fontweight='bold')

# Añadir nota al pie
ax.text(0.5, -0.12, 
        f'Fuente: Datos simulados de rendimiento académico FACES (n={len(df_rendimiento):,}). '
        f'Barras de error: IC 95%.', 
        transform=ax.transAxes, fontsize=9, color='gray', ha='center')

# Limpiar bordes
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.grid(axis='x', alpha=0.3)
plt.tight_layout()

# Guardar la figura
# plt.savefig('promedio_por_carrera.png', dpi=300, bbox_inches='tight')

plt.show()

print('\n📌 Para guardar: plt.savefig("archivo.png", dpi=300, bbox_inches="tight")')
print('   Formatos: .png (presentaciones), .pdf (publicaciones), .svg (web)')
````

### Guía rápida: ¿Qué gráfico usar?

| Pregunta | Tipo de gráfico | Función |
|----------|---------------|--------|
| ¿Cómo se distribuyen los datos? | Histograma, KDE | `sns.histplot()`, `sns.kdeplot()` |
| ¿Cómo se comparan categorías? | Barras | `plt.bar()`, `sns.barplot()` |
| ¿Cómo cambia algo en el tiempo? | Líneas | `plt.plot()` |
| ¿Hay relación entre dos variables? | Scatter | `plt.scatter()`, `sns.scatterplot()` |
| ¿Hay outliers o diferencias de dispersión? | Boxplot | `sns.boxplot()` |
| ¿Qué variables correlacionan? | Heatmap | `sns.heatmap()` |
| ¿Múltiples vistas de un tema? | Subplots | `plt.subplots()` |

## 📋 Resumen

En esta sesión aprendimos:

| Tipo de gráfico | Función principal | Caso de uso |
|----------------|-----------------|------------|
| Línea | `plt.plot()` | Tendencias temporales |
| Barras | `plt.bar()` / `sns.barplot()` | Comparar categorías |
| Histograma | `plt.hist()` / `sns.histplot()` | Distribuciones |
| KDE | `sns.kdeplot()` | Distribuciones suavizadas |
| Scatter | `plt.scatter()` / `sns.scatterplot()` | Relación entre 2 variables |
| Boxplot | `sns.boxplot()` | Distribución + outliers |
| Heatmap | `sns.heatmap()` | Correlaciones, tablas cruzadas |
| Subplots | `plt.subplots()` | Múltiples gráficos en una figura |

### Principios de buena visualización

1. **Simplicidad:** Menos es más. Elimina elementos que no aportan información
2. **Claridad:** Títulos descriptivos, ejes etiquetados, leyendas cuando hay múltiples series
3. **Honestidad:** No distorsionar escalas ni truncar ejes para manipular la percepción
4. **Contexto:** Incluir fuentes, notas y unidades de medida
5. **Color con propósito:** Usar color para codificar información, no solo para decorar
