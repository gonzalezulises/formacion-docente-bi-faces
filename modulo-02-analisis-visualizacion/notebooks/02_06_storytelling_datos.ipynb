## 📖 Sección 1: ¿Qué es el storytelling con datos?

**Storytelling con datos** es la práctica de construir una **narrativa** alrededor de datos y visualizaciones para comunicar un mensaje de forma clara y persuasiva.

### La visualización sola no basta

Imagínate esta situación: presentas un gráfico de barras con la matrícula por carrera en un consejo de facultad. El decano mira el gráfico y dice: *"¿Y qué significa esto? ¿Qué quieres que hagamos?"*

El gráfico mostró los datos, pero **no contó la historia**.

### El arco narrativo de los datos

Toda buena historia de datos sigue un arco de **tres actos**, similar al cine:

| Acto | Nombre | Pregunta que responde | Ejemplo académico |
|------|--------|----------------------|-------------------|
| 1 | **Planteamiento** (Setup) | ¿Dónde estamos? ¿Cuál es el contexto? | "FACES tuvo 6,000 estudiantes en 2015" |
| 2 | **Conflicto** (Conflict) | ¿Qué cambió? ¿Cuál es el problema? | "En 2024, solo quedan 2,300. La matrícula cayó un 62%" |
| 3 | **Resolución** (Resolution) | ¿Qué debemos hacer? | "Las carreras más afectadas necesitan programas de retención" |

### El framework de Cole Nussbaumer Knaflic

La autora de *Storytelling with Data* propone 6 principios:

1. **Entiende el contexto** — ¿Quién es tu audiencia? ¿Qué quieres que hagan?
2. **Elige una visualización adecuada** — No toda data merece un gráfico circular
3. **Elimina lo innecesario** — Menos es más. Quita todo lo que no contribuya al mensaje
4. **Dirige la atención** — Usa color, tamaño y posición para guiar la mirada
5. **Piensa como un diseñador** — Accesibilidad, alineación, espacio en blanco
6. **Cuenta una historia** — Inicio, nudo y desenlace

## 🎭 Sección 2: Anatomía de una buena historia de datos

### Las 3 C del storytelling con datos

#### 1. Contexto (¿Para quién?)

Antes de crear un solo gráfico, responde:

- **¿Quién es tu audiencia?** — Decano, consejo de facultad, Ministerio, estudiantes, público general
- **¿Qué saben ya?** — ¿Conocen el tema? ¿Necesitan contexto básico?
- **¿Qué les importa?** — Al decano le importan los números; al ministerio, las tendencias nacionales
- **¿Qué quieres que HAGAN?** — Esta es la pregunta más importante. ¿Aprobar un presupuesto? ¿Cambiar una política?

#### 2. Contenido (¿Qué dicen los datos?)

- **Hallazgo principal:** El dato o tendencia más importante
- **Evidencia de soporte:** Datos que refuerzan el hallazgo principal
- **Contrapunto:** Datos que matizan o complejizan la historia (credibilidad)

#### 3. Call to Action (¿Qué hacer?)

- Toda presentación de datos debe terminar con una **recomendación específica**
- No: *"Los datos muestran que la matrícula bajó"* (descriptivo)
- Sí: *"Proponemos un programa de retención estudiantil para las 3 carreras más afectadas, con un costo estimado de X"* (accionable)

### Ejemplo del contexto universitario

| Elemento | Ejemplo |
|----------|--------|
| **Audiencia** | Consejo de Facultad de FACES |
| **Lo que saben** | Saben que la matrícula ha caído, pero no la magnitud |
| **Lo que les importa** | Mantener la acreditación y los recursos asignados |
| **Acción deseada** | Aprobar un plan de retención estudiantil |
| **Hallazgo principal** | La matrícula cayó 62% en 10 años |
| **Evidencia** | Economía perdió más que Contaduría |
| **Call to action** | Programa de becas + tutorías para las 2 carreras más afectadas |

## 🎨 Sección 3: Principios de diseño visual

### Principios de la Gestalt

La psicología de la Gestalt explica cómo el cerebro organiza la información visual:

| Principio | Descripción | Aplicación en gráficos |
|-----------|-----------|---------------------|
| **Proximidad** | Elementos cercanos se perciben como grupo | Agrupar barras relacionadas |
| **Similitud** | Elementos con el mismo color/forma se agrupan | Usar el mismo color para la misma serie |
| **Cierre** | El cerebro completa formas incompletas | Las líneas de cuadrícula no necesitan ser continuas |
| **Continuidad** | Los ojos siguen líneas y curvas | Las líneas de tendencia guían la mirada |
| **Figura-fondo** | Separamos un elemento del fondo | Resaltar una barra oscureciendo las demás |

### Atributos pre-atentivos

Son características visuales que el cerebro procesa **instantáneamente** (en menos de 500 milisegundos), sin esfuerzo consciente:

- **Color** (el más poderoso)
- **Tamaño**
- **Posición**
- **Orientación**
- **Forma**

**Regla práctica:** Si quieres que algo destaque, cámbialo en **una sola dimensión** (solo color, o solo tamaño). No uses las 5 a la vez.

### Ratio datos-tinta (Data-Ink Ratio)

Concepto de Edward Tufte: la proporción de "tinta" en un gráfico que representa datos reales vs. decoración.

**Eliminar:**
- Bordes gruesos alrededor de barras
- Fondos de color
- Líneas de cuadrícula innecesarias
- Efectos 3D (nunca, jamás)
- Leyendas cuando hay pocas series (mejor etiquetar directamente)

Veamos esto en práctica con un ejemplo de **antes y después**:

````python
# ============================================================
# Ejemplo ANTES: Gráfico recargado (lo que NO debemos hacer)
# ============================================================
# Usamos indicadores del BCV para mostrar inflación

# Preparar datos
datos_inflacion = df_bcv[['fecha', 'inflacion_mensual']].dropna().copy()
datos_inflacion['anio'] = datos_inflacion['fecha'].dt.year
inflacion_anual = datos_inflacion.groupby('anio')['inflacion_mensual'].mean().reset_index()

# Gráfico RECARGADO (anti-patrón)
fig, ax = plt.subplots(figsize=(10, 6))

# Colores aleatorios por barra (mal uso del color)
colores_random = plt.cm.Set3(np.linspace(0, 1, len(inflacion_anual)))

barras = ax.bar(inflacion_anual['anio'], inflacion_anual['inflacion_mensual'],
                color=colores_random, edgecolor='black', linewidth=2)

# Cuadrícula excesiva
ax.grid(True, which='both', axis='both', linestyle='-', linewidth=1, color='gray')
ax.set_facecolor('#F0F0F0')  # Fondo gris

# Título genérico y redundante
ax.set_title('Gráfico de Barras de la Inflación Mensual Promedio por Año\n(Fuente: BCV, elaboración propia)', 
             fontsize=11, fontweight='bold')
ax.set_xlabel('Año')
ax.set_ylabel('Inflación Mensual Promedio (%)')

# Etiquetas rotadas (difíciles de leer)
ax.set_xticks(inflacion_anual['anio'])
ax.set_xticklabels(inflacion_anual['anio'], rotation=45)

# Bordes del gráfico
for spine in ax.spines.values():
    spine.set_visible(True)
    spine.set_linewidth(2)

plt.tight_layout()
plt.suptitle('ANTES: Gráfico recargado', y=1.02, fontsize=14, color='red', fontweight='bold')
plt.show()

print("\n❌ Problemas de este gráfico:")
print("   1. Colores aleatorios sin significado")
print("   2. Bordes gruesos y fondo gris innecesario")
print("   3. Cuadrícula excesiva")
print("   4. Título genérico que no cuenta una historia")
print("   5. No dirige la atención a ningún punto clave")
````

````python
# ============================================================
# Ejemplo DESPUÉS: Gráfico limpio y con narrativa
# ============================================================

fig, ax = plt.subplots(figsize=(10, 6))

# Color uniforme con destaque estratégico
# Destacamos los años de mayor inflación en rojo
colores = [COLORES['rojo'] if v > 15 else COLORES['gris_claro'] 
           for v in inflacion_anual['inflacion_mensual']]

barras = ax.bar(inflacion_anual['anio'], inflacion_anual['inflacion_mensual'],
                color=colores, width=0.7)

# Título que cuenta una historia
ax.set_title('La inflación mensual promedio superó el 15% en los años críticos',
             fontsize=14, fontweight='bold', color=COLORES['azul_oscuro'], pad=15)

# Subtítulo con contexto
ax.text(0.0, 1.02, 'Inflación mensual promedio en Venezuela, datos del BCV',
        transform=ax.transAxes, fontsize=10, color=COLORES['gris'], style='italic')

# Etiquetar solo las barras destacadas
for barra, valor in zip(barras, inflacion_anual['inflacion_mensual']):
    if valor > 15:
        ax.text(barra.get_x() + barra.get_width() / 2, barra.get_height() + 0.5,
                f'{valor:.1f}%', ha='center', va='bottom', fontweight='bold',
                color=COLORES['rojo'], fontsize=10)

# Etiquetas limpias
ax.set_ylabel('Inflación mensual promedio (%)', color=COLORES['gris'])
ax.set_xticks(inflacion_anual['anio'])
ax.tick_params(colors=COLORES['gris'])

# Línea de referencia
ax.axhline(y=15, color=COLORES['rojo'], linestyle='--', linewidth=1, alpha=0.5)
ax.text(inflacion_anual['anio'].max() + 0.3, 15, 'Umbral 15%', 
        color=COLORES['rojo'], fontsize=9, va='center', alpha=0.7)

# Cuadrícula mínima
ax.grid(axis='y', alpha=0.3, linewidth=0.5)
ax.set_axisbelow(True)

# Eliminar bordes superiores y derechos
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color(COLORES['gris_claro'])
ax.spines['bottom'].set_color(COLORES['gris_claro'])

plt.tight_layout()
plt.show()

print("\n✅ Mejoras aplicadas:")
print("   1. Color con propósito: rojo = años críticos, gris = el resto")
print("   2. Título narrativo que dice QUÉ pasa, no solo qué se muestra")
print("   3. Etiquetas solo en los puntos importantes")
print("   4. Línea de referencia para dar contexto (umbral del 15%)")
print("   5. Diseño limpio: sin bordes gruesos, cuadrícula mínima, espacio respirable")
````

## 💡 Sección 4: Elegir el gráfico correcto

### Árbol de decisión

La pregunta no es *"¿Qué gráfico se ve más bonito?"* sino *"¿Qué quiero que la audiencia entienda?"*

| ¿Qué quiero mostrar? | Gráfico recomendado | Evitar |
|---------------------|--------------------|---------|
| **Comparación** entre categorías | Barras (horizontal si hay muchas) | Circular con 10+ categorías |
| **Tendencia** en el tiempo | Líneas | Barras (dificultan ver tendencia) |
| **Distribución** de una variable | Histograma, boxplot | Circular |
| **Relación** entre dos variables | Dispersión (scatter) | Barras |
| **Composición** (partes de un todo) | Barras apiladas, circular (pocas partes) | Circular con >5 partes |
| **Un solo número destacado** | Tarjeta / KPI grande | Gráfico completo |

Veamos cada tipo con datos reales:

````python
# ============================================================
# Demostración: Cada tipo de gráfico con su mejor uso
# ============================================================

# Preparar datos de matrícula para las demostraciones
# Extraer año del periodo (formato: "2015-I" -> 2015)
df_matricula['anio'] = df_matricula['periodo'].str[:4].astype(int)

# Matrícula total por año
matricula_anual = df_matricula.groupby('anio').size().reset_index(name='total')

# Matrícula por carrera
matricula_carrera = df_matricula.groupby('carrera').size().reset_index(name='total')
matricula_carrera = matricula_carrera.sort_values('total', ascending=True)

# Distribución de edades
edades = df_matricula['edad'].dropna()

# Relación edad-matrícula por año y carrera
cruce = df_matricula.groupby(['anio', 'carrera']).size().reset_index(name='total')

# Composición por estatus
composicion = df_matricula.groupby('estatus').size().reset_index(name='total')
composicion['pct'] = round(composicion['total'] / composicion['total'].sum() * 100, 1)

print("✅ Datos preparados para las demostraciones")
````

````python
# ============================================================
# Panel de 4 gráficos: cada uno para su propósito
# ============================================================

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Cada pregunta tiene su gráfico ideal', 
             fontsize=16, fontweight='bold', color=COLORES['azul_oscuro'], y=1.02)

# --- 1. COMPARACIÓN: Barras horizontales ---
ax1 = axes[0, 0]
ax1.barh(matricula_carrera['carrera'], matricula_carrera['total'],
         color=COLORES['azul_medio'])
ax1.set_title('COMPARACIÓN: Matrícula por carrera', fontweight='bold', fontsize=11)
ax1.set_xlabel('Total de registros')
for i, v in enumerate(matricula_carrera['total']):
    ax1.text(v + 20, i, f'{v:,}', va='center', fontsize=9)

# --- 2. TENDENCIA: Líneas ---
ax2 = axes[0, 1]
ax2.plot(matricula_anual['anio'], matricula_anual['total'], 
         color=COLORES['azul_oscuro'], linewidth=2.5, marker='o', markersize=4)
ax2.fill_between(matricula_anual['anio'], matricula_anual['total'], 
                 alpha=0.1, color=COLORES['azul_medio'])
ax2.set_title('TENDENCIA: Matrícula anual', fontweight='bold', fontsize=11)
ax2.set_ylabel('Total de registros')

# --- 3. DISTRIBUCIÓN: Histograma ---
ax3 = axes[1, 0]
ax3.hist(edades, bins=20, color=COLORES['cian'], edgecolor='white', linewidth=0.5)
ax3.axvline(edades.median(), color=COLORES['rojo'], linestyle='--', linewidth=1.5, label=f'Mediana: {edades.median():.0f}')
ax3.set_title('DISTRIBUCIÓN: Edades de ingreso', fontweight='bold', fontsize=11)
ax3.set_xlabel('Edad')
ax3.set_ylabel('Frecuencia')
ax3.legend()

# --- 4. COMPOSICIÓN: Barras apiladas / circular ---
ax4 = axes[1, 1]
colores_est = [COLORES['verde'], COLORES['azul_medio'], COLORES['rojo']]
wedges, texts, autotexts = ax4.pie(composicion['total'], 
                                    labels=composicion['estatus'],
                                    colors=colores_est,
                                    autopct='%1.1f%%',
                                    pctdistance=0.75,
                                    startangle=90)
for autotext in autotexts:
    autotext.set_fontsize(10)
    autotext.set_fontweight('bold')
ax4.set_title('COMPOSICIÓN: Estatus de matrícula', fontweight='bold', fontsize=11)

# Ajustes generales
for ax in [ax1, ax2, ax3]:
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.grid(axis='y' if ax != ax1 else 'x', alpha=0.3)

plt.tight_layout()
plt.show()

print("\n📌 Guía rápida:")
print("   - Comparar valores → Barras (horizontales si hay textos largos)")
print("   - Ver tendencias    → Líneas")
print("   - Ver distribución  → Histograma o boxplot")
print("   - Ver composición   → Circular (solo si hay 3-5 categorías)")
````

## 🎬 Sección 5: Caso práctico — Presentar la situación de FACES

Ahora construiremos una **narrativa completa** usando los datos de matrícula de FACES. Imaginemos que debemos presentar esta información al Consejo de Facultad.

### Nuestro plan narrativo

| Gráfico | Acto | Mensaje | Emoción |
|---------|------|---------|--------|
| 1 | Planteamiento | La matrícula ha caído sostenidamente | Preocupación |
| 2 | Conflicto | No todas las carreras sufren igual | Urgencia focalizada |
| 3 | Insight adicional | La composición de género también cambió | Complejidad |

````python
# ============================================================
# Preparar datos para la narrativa
# ============================================================

# Agrupar por año (combinando ambos semestres)
tendencia = df_matricula.groupby('anio').size().reset_index(name='total')

# Matrícula por carrera y año
por_carrera_anio = df_matricula.groupby(['anio', 'carrera']).size().reset_index(name='total')

# Composición por género y año
por_genero_anio = df_matricula.groupby(['anio', 'genero']).size().unstack(fill_value=0)
por_genero_anio['total'] = por_genero_anio.sum(axis=1)
por_genero_anio['pct_F'] = round(por_genero_anio['F'] / por_genero_anio['total'] * 100, 1)
por_genero_anio['pct_M'] = round(por_genero_anio['M'] / por_genero_anio['total'] * 100, 1)

# Datos clave para la narrativa
anio_inicio = tendencia['anio'].min()
anio_fin = tendencia['anio'].max()
mat_inicio = tendencia[tendencia['anio'] == anio_inicio]['total'].values[0]
mat_fin = tendencia[tendencia['anio'] == anio_fin]['total'].values[0]
cambio_pct = round((mat_fin - mat_inicio) / mat_inicio * 100, 1)

print(f"📊 Datos clave para la narrativa:")
print(f"   Período: {anio_inicio} a {anio_fin}")
print(f"   Matrícula inicial ({anio_inicio}): {mat_inicio:,}")
print(f"   Matrícula final ({anio_fin}): {mat_fin:,}")
print(f"   Cambio: {cambio_pct}%")
````

````python
# ============================================================
# GRÁFICO 1 - PLANTEAMIENTO: La tendencia general
# ============================================================

fig, ax = plt.subplots(figsize=(11, 6))

# Línea principal
ax.plot(tendencia['anio'], tendencia['total'], 
        color=COLORES['azul_oscuro'], linewidth=3, marker='o', markersize=6, zorder=3)

# Área sombreada
ax.fill_between(tendencia['anio'], tendencia['total'], 
                alpha=0.08, color=COLORES['azul_oscuro'])

# Anotar puntos clave
ax.annotate(f'{mat_inicio:,} registros',
            xy=(anio_inicio, mat_inicio), 
            xytext=(anio_inicio + 0.5, mat_inicio + 30),
            fontsize=11, fontweight='bold', color=COLORES['azul_oscuro'],
            arrowprops=dict(arrowstyle='->', color=COLORES['azul_oscuro'], lw=1.5))

ax.annotate(f'{mat_fin:,} registros',
            xy=(anio_fin, mat_fin),
            xytext=(anio_fin - 1.5, mat_fin + 40),
            fontsize=11, fontweight='bold', color=COLORES['rojo'],
            arrowprops=dict(arrowstyle='->', color=COLORES['rojo'], lw=1.5))

# Título narrativo
ax.set_title(f'La matrícula de FACES cayó un {abs(cambio_pct)}% en {anio_fin - anio_inicio} años',
             fontsize=15, fontweight='bold', color=COLORES['azul_oscuro'], pad=20)
ax.text(0.0, 1.02, f'Registros totales de matrícula por año, período {anio_inicio}-{anio_fin}',
        transform=ax.transAxes, fontsize=10, color=COLORES['gris'], style='italic')

# Formato
ax.set_ylabel('Registros de matrícula', color=COLORES['gris'])
ax.set_xticks(tendencia['anio'])
ax.tick_params(colors=COLORES['gris'])
ax.grid(axis='y', alpha=0.3)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color(COLORES['gris_claro'])
ax.spines['bottom'].set_color(COLORES['gris_claro'])

plt.tight_layout()
plt.show()

print("\n📖 Narrativa (Acto 1 - Planteamiento):")
print(f"   \"FACES registró {mat_inicio:,} entradas de matrícula en {anio_inicio}.")
print(f"   Para {anio_fin}, esta cifra se redujo a {mat_fin:,}, una caída del {abs(cambio_pct)}%.")
print(f"   La tendencia ha sido consistentemente descendente a lo largo de la década.\"")
````

````python
# ============================================================
# GRÁFICO 2 - CONFLICTO: ¿Qué carreras sufren más?
# ============================================================

fig, ax = plt.subplots(figsize=(11, 6))

# Colores por carrera
carreras = ['Economia', 'Administracion', 'Contaduria', 'Relaciones Industriales']
colores_carrera = {
    'Economia': COLORES['rojo'],
    'Administracion': COLORES['azul_medio'],
    'Contaduria': COLORES['cian'],
    'Relaciones Industriales': COLORES['dorado']
}

# Calcular caída porcentual por carrera para la narrativa
caidas = {}

for carrera in carreras:
    datos = por_carrera_anio[por_carrera_anio['carrera'] == carrera].sort_values('anio')
    color = colores_carrera[carrera]
    
    ax.plot(datos['anio'], datos['total'], 
            color=color, linewidth=2.5, marker='o', markersize=4, label=carrera)
    
    # Etiquetar al final de cada línea
    ultimo_anio = datos['anio'].iloc[-1]
    ultimo_valor = datos['total'].iloc[-1]
    ax.text(ultimo_anio + 0.2, ultimo_valor, carrera,
            fontsize=9, color=color, fontweight='bold', va='center')
    
    # Calcular caída
    primer_valor = datos['total'].iloc[0]
    caida = round((ultimo_valor - primer_valor) / primer_valor * 100, 1)
    caidas[carrera] = caida

# Título narrativo
carrera_mas_afectada = min(caidas, key=caidas.get)
caida_max = caidas[carrera_mas_afectada]

ax.set_title(f'{carrera_mas_afectada} fue la carrera más afectada ({caida_max}%)',
             fontsize=15, fontweight='bold', color=COLORES['azul_oscuro'], pad=20)
ax.text(0.0, 1.02, 'Registros de matrícula por carrera, evolución anual',
        transform=ax.transAxes, fontsize=10, color=COLORES['gris'], style='italic')

# Formato
ax.set_ylabel('Registros de matrícula', color=COLORES['gris'])
ax.set_xticks(tendencia['anio'])
ax.tick_params(colors=COLORES['gris'])
ax.grid(axis='y', alpha=0.3)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color(COLORES['gris_claro'])
ax.spines['bottom'].set_color(COLORES['gris_claro'])

# Sin leyenda tradicional (etiquetas directas son más claras)

plt.tight_layout()
plt.show()

print("\n📖 Narrativa (Acto 2 - Conflicto):")
print("   \"Sin embargo, la caída no fue uniforme entre carreras:")
for carrera, caida in sorted(caidas.items(), key=lambda x: x[1]):
    print(f"   - {carrera}: {caida}%")
print(f"   {carrera_mas_afectada} requiere atención prioritaria.\"")
````

````python
# ============================================================
# GRÁFICO 3 - INSIGHT ADICIONAL: Cambio en composición de género
# ============================================================

fig, ax = plt.subplots(figsize=(11, 6))

anios = por_genero_anio.index
pct_f = por_genero_anio['pct_F']
pct_m = por_genero_anio['pct_M']

# Áreas apiladas
ax.fill_between(anios, 0, pct_f, alpha=0.3, color=COLORES['dorado'], label='Femenino')
ax.fill_between(anios, pct_f, 100, alpha=0.3, color=COLORES['azul_medio'], label='Masculino')

# Líneas de tendencia
ax.plot(anios, pct_f, color=COLORES['dorado'], linewidth=2.5, marker='o', markersize=4)
ax.plot(anios, pct_m, color=COLORES['azul_medio'], linewidth=2.5, marker='s', markersize=4)

# Línea de referencia 50%
ax.axhline(y=50, color=COLORES['gris'], linestyle=':', linewidth=1, alpha=0.5)
ax.text(anio_fin + 0.2, 50, '50%', color=COLORES['gris'], fontsize=9, va='center')

# Etiquetas directas al final
ax.text(anio_fin + 0.2, pct_f.iloc[-1], f'Femenino\n{pct_f.iloc[-1]:.1f}%', 
        color=COLORES['dorado'], fontsize=10, fontweight='bold', va='center')
ax.text(anio_fin + 0.2, pct_m.iloc[-1], f'Masculino\n{pct_m.iloc[-1]:.1f}%', 
        color=COLORES['azul_medio'], fontsize=10, fontweight='bold', va='center')

# Anotar año de cruce si existe
diferencias = pct_f - pct_m
cruces = diferencias[diferencias.shift() * diferencias < 0]
if len(cruces) > 0:
    anio_cruce = cruces.index[0]
    ax.annotate(f'Cruce en {anio_cruce}',
                xy=(anio_cruce, 50),
                xytext=(anio_cruce - 2, 42),
                fontsize=10, fontweight='bold', color=COLORES['rojo'],
                arrowprops=dict(arrowstyle='->', color=COLORES['rojo'], lw=1.5))

# Determinar la tendencia para el título
pct_f_inicio = pct_f.iloc[0]
pct_f_fin = pct_f.iloc[-1]
cambio_genero = 'creció' if pct_f_fin > pct_f_inicio else 'disminuyó'

# Título narrativo
ax.set_title(f'La participación femenina {cambio_genero}: de {pct_f_inicio:.1f}% a {pct_f_fin:.1f}%',
             fontsize=15, fontweight='bold', color=COLORES['azul_oscuro'], pad=20)
ax.text(0.0, 1.02, 'Composición porcentual por género en la matrícula de FACES',
        transform=ax.transAxes, fontsize=10, color=COLORES['gris'], style='italic')

# Formato
ax.set_ylabel('Porcentaje (%)', color=COLORES['gris'])
ax.set_ylim(0, 100)
ax.set_xticks(anios)
ax.tick_params(colors=COLORES['gris'])
ax.grid(axis='y', alpha=0.3)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color(COLORES['gris_claro'])
ax.spines['bottom'].set_color(COLORES['gris_claro'])

plt.tight_layout()
plt.show()

print("\n📖 Narrativa (Acto 3 - Insight adicional):")
print(f"   \"Además de la caída en números absolutos, la composición de género")
print(f"   también se transformó. La participación femenina pasó de {pct_f_inicio:.1f}%")
print(f"   a {pct_f_fin:.1f}%. Este dato es relevante para el diseño de")
print(f"   políticas de retención diferenciadas.\"")
````

````python
# ============================================================
# SÍNTESIS NARRATIVA: La historia completa
# ============================================================

print("=" * 70)
print("📖 NARRATIVA COMPLETA: LA SITUACIÓN DE FACES")
print("=" * 70)
print(f"""
PLANTEAMIENTO:
La Facultad de Ciencias Económicas y Sociales (FACES) de la Universidad
de Carabobo ha experimentado una caída significativa en su matrícula.
En {anio_inicio}, se registraron {mat_inicio:,} entradas; para {anio_fin}, la cifra 
se redujo a {mat_fin:,}, una disminución del {abs(cambio_pct)}%.

CONFLICTO:
La caída no fue uniforme. Mientras algunas carreras mantuvieron 
cierta estabilidad relativa, otras sufrieron pérdidas drásticas:
""")
for carrera, caida in sorted(caidas.items(), key=lambda x: x[1]):
    indicador = "⚠️" if caida < -40 else "🟡" if caida < -20 else "✅"
    print(f"   {indicador} {carrera}: {caida}%")

print(f"""
INSIGHT ADICIONAL:
La composición de género también se transformó durante este período.
La participación femenina pasó de {pct_f_inicio:.1f}% a {pct_f_fin:.1f}%.

RECOMENDACIÓN:
Se propone implementar un programa de retención estudiantil focalizado
en las carreras con mayor caída, con estrategias diferenciadas que
consideren la composición demográfica actual de cada programa.
""")
print("=" * 70)
````

## 📋 Resumen

En esta sesión aprendimos:

| Concepto | Descripción |
|----------|------------|
| **Storytelling con datos** | Construir narrativas alrededor de datos para mover a la acción |
| **Arco narrativo** | Planteamiento → Conflicto → Resolución |
| **Las 3 C** | Contexto (audiencia), Contenido (datos), Call to action (acción) |
| **Principios Gestalt** | Proximidad, similitud, cierre, continuidad, figura-fondo |
| **Atributos pre-atentivos** | Color, tamaño, posición para dirigir la mirada |
| **Data-ink ratio** | Eliminar tinta que no representa datos |
| **Títulos narrativos** | "La matrícula cayó 62%" vs "Gráfico de matrícula" |
| **Color con propósito** | Gris para contexto, color para el dato clave |
| **Elección de gráfico** | Depende de la pregunta, no de la estética |
| **Anotaciones** | Texto dentro del gráfico que guía la interpretación |

### Checklist para cada visualización

Antes de presentar cualquier gráfico, verifica:

- [ ] ¿El título cuenta lo que pasa, no solo lo que se muestra?
- [ ] ¿El color tiene un propósito (no es decorativo)?
- [ ] ¿Eliminé todo lo innecesario (bordes, cuadrículas, 3D, efectos)?
- [ ] ¿La audiencia puede entender el mensaje en 5 segundos?
- [ ] ¿Hay un call to action claro?
- [ ] ¿El tipo de gráfico es el correcto para la pregunta?
