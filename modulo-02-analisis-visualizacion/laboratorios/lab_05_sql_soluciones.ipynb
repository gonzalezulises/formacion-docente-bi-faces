## ⚙️ Configuración inicial

Ejecuta la siguiente celda para cargar los datasets en una base de datos SQLite en memoria.

````python
# ============================================================
# Configuración inicial: cargar CSVs en SQLite
# ============================================================

import sqlite3
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

# Crear base de datos en memoria
conexion = sqlite3.connect(':memory:')

# Cargar rendimiento académico
df_rendimiento = pd.read_csv('../../datasets/universidad/rendimiento_academico.csv')
df_rendimiento.to_sql('rendimiento', conexion, index=False, if_exists='replace')

# Cargar matrícula
df_matricula = pd.read_csv('../../datasets/universidad/matricula_faces.csv')
df_matricula.to_sql('matricula', conexion, index=False, if_exists='replace')

print('✅ Base de datos SQLite creada con 2 tablas:')
print(f'   rendimiento: {len(df_rendimiento):,} registros, columnas: {list(df_rendimiento.columns)}')
print(f'   matricula:   {len(df_matricula):,} registros, columnas: {list(df_matricula.columns)}')

# Función auxiliar para ejecutar consultas
def consulta(sql, conn=conexion):
    """Ejecuta una consulta SQL y devuelve el resultado como DataFrame."""
    print(f'\033[94m{sql}\033[0m')
    print()
    resultado = pd.read_sql_query(sql, conn)
    return resultado

print('\n✅ Función consulta() lista para usar')
print('   Ejemplo: resultado = consulta("SELECT * FROM rendimiento LIMIT 5")')
````

````python
# ============================================================
# Verificación: ver las primeras filas de cada tabla
# ============================================================

print('=== Tabla rendimiento ===')
print(consulta("SELECT * FROM rendimiento LIMIT 3").to_string(index=False))

print('\n=== Tabla matricula ===')
print(consulta("SELECT * FROM matricula LIMIT 3").to_string(index=False))
````

## ✏️ Ejercicio 1: SELECT con WHERE

Escribe una consulta que seleccione **todos los estudiantes** de la carrera **Contaduria** de la tabla `rendimiento`.

Muestra las columnas: `estudiante_id`, `carrera`, `semestre`, `promedio`, `asistencia_pct`

Limita el resultado a **15 filas**.

````python
# ============================================================
# Ejercicio 1: SELECT con WHERE — SOLUCIÓN
# ============================================================

resultado = consulta("""
    SELECT estudiante_id, carrera, semestre, promedio, asistencia_pct
    FROM rendimiento
    WHERE carrera = 'Contaduria'
    LIMIT 15
""")
print(resultado.to_string(index=False))

# --- Explicación ---
# SELECT especifica las 5 columnas que queremos ver.
# WHERE carrera = 'Contaduria' filtra solo los registros de esa carrera.
# LIMIT 15 restringe la salida a las primeras 15 filas.
# En SQL, los valores de texto van entre comillas simples ('Contaduria').
````

## ✏️ Ejercicio 2: COUNT con GROUP BY y ORDER BY

Cuenta el **número de registros** en la tabla `matricula` agrupados por `estatus`.

Ordena los resultados de mayor a menor (`ORDER BY ... DESC`).

**Resultado esperado:** Una tabla con 3 filas (Activo, Egresado, Retirado) y su conteo.

````python
# ============================================================
# Ejercicio 2: COUNT con GROUP BY — SOLUCIÓN
# ============================================================

resultado = consulta("""
    SELECT estatus, COUNT(*) AS cantidad
    FROM matricula
    GROUP BY estatus
    ORDER BY cantidad DESC
""")
print(resultado.to_string(index=False))

# --- Explicación ---
# COUNT(*) cuenta todas las filas en cada grupo.
# AS cantidad le da un nombre legible a la columna calculada.
# GROUP BY estatus agrupa los registros por su valor de estatus
#   (Activo, Egresado, Retirado), creando un conteo por cada uno.
# ORDER BY cantidad DESC ordena de mayor a menor conteo.
````

## ✏️ Ejercicio 3: AVG con WHERE y GROUP BY

Calcula el **promedio académico promedio** (`AVG(promedio)`) por `carrera`, pero **solo para estudiantes en semestre 5 o superior** (`semestre >= 5`).

Incluye también el conteo de estudiantes en cada carrera y la asistencia promedio.

Ordena por promedio académico descendente.

````python
# ============================================================
# Ejercicio 3: AVG con WHERE y GROUP BY — SOLUCIÓN
# ============================================================

resultado = consulta("""
    SELECT carrera,
           COUNT(*) AS estudiantes,
           ROUND(AVG(promedio), 2) AS promedio_general,
           ROUND(AVG(asistencia_pct), 1) AS asistencia_promedio
    FROM rendimiento
    WHERE semestre >= 5
    GROUP BY carrera
    ORDER BY promedio_general DESC
""")
print(resultado.to_string(index=False))

# --- Explicación ---
# WHERE semestre >= 5 filtra ANTES de agrupar, dejando solo estudiantes
#   de semestres avanzados (5to en adelante).
# COUNT(*) cuenta cuántos estudiantes hay en cada carrera (post-filtro).
# AVG(promedio) calcula la media aritmética del promedio académico.
# AVG(asistencia_pct) calcula la asistencia media por carrera.
# ROUND(..., 2) redondea a 2 decimales para legibilidad.
# ORDER BY promedio_general DESC muestra primero la carrera con mejor promedio.
````

## ✏️ Ejercicio 4: JOIN entre tablas

Combina información de ambas tablas. Crea:

1. Un **resumen de matrícula** por carrera: total registros y % de retirados
2. Un **resumen de rendimiento** por carrera: promedio y asistencia
3. Un **JOIN** entre ambos resúmenes usando `carrera` como clave

**Pista:** Usa subconsultas (una para cada resumen) y luego únelas con INNER JOIN.

```sql
SELECT m.carrera, m.total_registros, m.pct_retirados,
       r.promedio_general, r.asistencia_promedio
FROM (
    SELECT carrera, COUNT(*) AS total_registros, ...
    FROM matricula
    GROUP BY carrera
) m
INNER JOIN (
    SELECT carrera, ROUND(AVG(promedio), 2) AS promedio_general, ...
    FROM rendimiento
    GROUP BY carrera
) r ON m.carrera = r.carrera
```

````python
# ============================================================
# Ejercicio 4: JOIN entre matrícula y rendimiento — SOLUCIÓN
# ============================================================

resultado = consulta("""
    SELECT m.carrera,
           m.total_registros,
           m.pct_retirados,
           r.promedio_general,
           r.asistencia_promedio
    FROM (
        SELECT carrera,
               COUNT(*) AS total_registros,
               ROUND(SUM(CASE WHEN estatus = 'Retirado' THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 1) AS pct_retirados
        FROM matricula
        GROUP BY carrera
    ) m
    INNER JOIN (
        SELECT carrera,
               ROUND(AVG(promedio), 2) AS promedio_general,
               ROUND(AVG(asistencia_pct), 1) AS asistencia_promedio
        FROM rendimiento
        GROUP BY carrera
    ) r ON m.carrera = r.carrera
""")
print(resultado.to_string(index=False))

# --- Explicación ---
# Esta consulta usa dos subconsultas (tablas derivadas) unidas con INNER JOIN:
#
# Subconsulta "m" (matrícula):
#   - COUNT(*) cuenta el total de registros por carrera.
#   - CASE WHEN estatus = 'Retirado' THEN 1 ELSE 0 END convierte el estatus
#     en 1/0 para poder sumar solo los retirados.
#   - Se multiplica por 100.0 (no 100) para forzar división decimal.
#
# Subconsulta "r" (rendimiento):
#   - Calcula promedio y asistencia media por carrera.
#
# INNER JOIN une ambas subconsultas por carrera, combinando datos
# de matrícula y rendimiento en una sola vista.
````

## ✏️ Ejercicio 5: Subconsulta — Por encima del promedio general

Encuentra todos los **estudiantes cuyo promedio es superior al promedio general** de toda la facultad.

Muestra: `estudiante_id`, `carrera`, `promedio`

Ordena por promedio descendente y limita a 20 resultados.

**Pista:** Usa una subconsulta en el WHERE:
```sql
WHERE promedio > (SELECT AVG(promedio) FROM rendimiento)
```

````python
# ============================================================
# Ejercicio 5: Subconsulta - por encima del promedio — SOLUCIÓN
# ============================================================

resultado = consulta("""
    SELECT estudiante_id, carrera, promedio
    FROM rendimiento
    WHERE promedio > (SELECT AVG(promedio) FROM rendimiento)
    ORDER BY promedio DESC
    LIMIT 20
""")
print(resultado.to_string(index=False))

# --- Explicación ---
# La subconsulta (SELECT AVG(promedio) FROM rendimiento) calcula
# el promedio general de TODA la facultad (un solo número).
# El WHERE compara cada estudiante contra ese promedio global.
# Solo se muestran los estudiantes que superan la media.
# ORDER BY promedio DESC pone primero a los de mayor promedio.
# LIMIT 20 restringe la salida a los 20 mejores.
````

## ✏️ Ejercicio 6: GROUP BY con HAVING

Encuentra las **carreras donde el promedio académico general es menor a 12 puntos**.

Muestra: carrera, total estudiantes, promedio general, asistencia promedio.

**Recuerda:** `HAVING` filtra después de agrupar, a diferencia de `WHERE` que filtra antes.

````python
# ============================================================
# Ejercicio 6: GROUP BY con HAVING — SOLUCIÓN
# ============================================================

resultado = consulta("""
    SELECT carrera,
           COUNT(*) AS estudiantes,
           ROUND(AVG(promedio), 2) AS promedio_general,
           ROUND(AVG(asistencia_pct), 1) AS asistencia_promedio
    FROM rendimiento
    GROUP BY carrera
    HAVING AVG(promedio) < 12
""")
print(resultado.to_string(index=False))

# --- Explicación ---
# GROUP BY carrera agrupa todos los registros por carrera.
# HAVING AVG(promedio) < 12 filtra DESPUÉS de agrupar.
#   - Diferencia clave: WHERE filtra filas individuales ANTES de agrupar.
#   - HAVING filtra grupos DESPUÉS de calcular las agregaciones.
# Solo se muestran las carreras cuyo promedio agregado sea < 12.
# Nota: HAVING usa AVG(promedio), no el alias promedio_general,
#   porque los alias no están disponibles en la cláusula HAVING en SQLite estándar.
````

## ✏️ Ejercicio 7: Consulta compleja — Tasa de retiro y rendimiento

Crea una consulta que muestre, **para cada carrera**:

1. El número total de registros en matrícula
2. El número de retirados
3. La **tasa de retiro** (porcentaje de retirados sobre el total)
4. El **promedio académico** (de la tabla rendimiento)
5. La **asistencia promedio** (de la tabla rendimiento)

Ordena por tasa de retiro descendente.

**Pista:** Necesitas un JOIN entre un resumen de matrícula y un resumen de rendimiento. Para la tasa de retiro usa:
```sql
ROUND(SUM(CASE WHEN estatus = 'Retirado' THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 1) AS tasa_retiro
```

````python
# ============================================================
# Ejercicio 7: Consulta compleja - tasa de retiro + rendimiento — SOLUCIÓN
# ============================================================

resultado = consulta("""
    SELECT m.carrera,
           m.total_matricula,
           m.retirados,
           m.tasa_retiro,
           r.promedio,
           r.asistencia
    FROM (
        SELECT carrera,
               COUNT(*) AS total_matricula,
               SUM(CASE WHEN estatus = 'Retirado' THEN 1 ELSE 0 END) AS retirados,
               ROUND(SUM(CASE WHEN estatus = 'Retirado' THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 1) AS tasa_retiro
        FROM matricula
        GROUP BY carrera
    ) m
    INNER JOIN (
        SELECT carrera,
               ROUND(AVG(promedio), 2) AS promedio,
               ROUND(AVG(asistencia_pct), 1) AS asistencia
        FROM rendimiento
        GROUP BY carrera
    ) r ON m.carrera = r.carrera
    ORDER BY tasa_retiro DESC
""")
print(resultado.to_string(index=False))

# --- Explicación ---
# Esta es la consulta más completa del laboratorio. Combina dos análisis:
#
# Subconsulta "m" (matrícula):
#   - COUNT(*) = total de registros de matrícula por carrera.
#   - SUM(CASE WHEN estatus = 'Retirado' THEN 1 ELSE 0 END) = conteo de retirados.
#     CASE WHEN actúa como un IF: si el estatus es 'Retirado' suma 1, si no suma 0.
#   - tasa_retiro = (retirados / total) * 100, redondeado a 1 decimal.
#     Se usa 100.0 (con decimal) para evitar división entera.
#
# Subconsulta "r" (rendimiento):
#   - Calcula promedio y asistencia media por carrera.
#
# INNER JOIN une ambos resúmenes por carrera.
# ORDER BY tasa_retiro DESC muestra primero la carrera con más deserción.
#
# Valor analítico: permite correlacionar visualmente si las carreras
# con mayor tasa de retiro también tienen menor promedio o asistencia.
````

## ✏️ Ejercicio 8: Tu propia consulta analítica

Diseña tu **propia consulta SQL** para responder una pregunta analítica sobre los datos de FACES.

**Requisitos:**
1. Escribe la pregunta que quieres responder en un comentario
2. Escribe la consulta SQL
3. Interpreta los resultados

**Ideas de preguntas:**
- ¿Cuál es la tasa de aprobación (materias_aprobadas / materias_inscritas) por carrera?
- ¿Los estudiantes becados tienen mejor promedio que los no becados?
- ¿Cómo ha evolucionado la matrícula por período y carrera?
- ¿Qué sede (Barbula vs La Morita) tiene mayor tasa de retiro?
- ¿Hay diferencias de rendimiento por género y carrera?

````python
# ============================================================
# Ejercicio 8a: Tasa de aprobación por carrera — SOLUCIÓN
# ============================================================

# Pregunta: ¿Cuál es la tasa de aprobación por carrera?
# (materias_aprobadas / materias_inscritas * 100)

resultado = consulta("""
    SELECT carrera,
           COUNT(*) AS estudiantes,
           ROUND(AVG(materias_inscritas), 1) AS materias_inscritas_prom,
           ROUND(AVG(materias_aprobadas), 1) AS materias_aprobadas_prom,
           ROUND(AVG(materias_aprobadas) * 100.0 / AVG(materias_inscritas), 1) AS tasa_aprobacion
    FROM rendimiento
    GROUP BY carrera
    ORDER BY tasa_aprobacion DESC
""")
print(resultado.to_string(index=False))

# Interpretación: La tasa de aprobación muestra qué proporción de las
# materias inscritas son finalmente aprobadas. Una tasa baja puede indicar
# dificultad académica, problemas de preparación o carga excesiva.
````

````python
# ============================================================
# Ejercicio 8b: Becados vs No Becados — SOLUCIÓN
# ============================================================

# Pregunta: ¿Los estudiantes becados tienen mejor promedio que los no becados?

resultado = consulta("""
    SELECT beca,
           COUNT(*) AS estudiantes,
           ROUND(AVG(promedio), 2) AS promedio_general,
           ROUND(AVG(asistencia_pct), 1) AS asistencia_promedio,
           ROUND(AVG(materias_aprobadas) * 100.0 / AVG(materias_inscritas), 1) AS tasa_aprobacion
    FROM rendimiento
    GROUP BY beca
""")
print(resultado.to_string(index=False))

# Interpretación: Comparar becados (True) vs no becados (False) permite
# evaluar si el programa de becas se asocia con mejor rendimiento.
# Nota: esto es correlación, no causalidad — las becas podrían otorgarse
# a estudiantes que ya tienen buen rendimiento.
````

````python
# ============================================================
# Ejercicio 8c: Tasa de retiro por sede — SOLUCIÓN
# ============================================================

# Pregunta: ¿Qué sede tiene mayor tasa de retiro?

resultado = consulta("""
    SELECT sede,
           COUNT(*) AS total,
           SUM(CASE WHEN estatus = 'Retirado' THEN 1 ELSE 0 END) AS retirados,
           ROUND(SUM(CASE WHEN estatus = 'Retirado' THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 1) AS tasa_retiro,
           SUM(CASE WHEN estatus = 'Egresado' THEN 1 ELSE 0 END) AS egresados,
           ROUND(SUM(CASE WHEN estatus = 'Egresado' THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 1) AS tasa_egreso
    FROM matricula
    GROUP BY sede
    ORDER BY tasa_retiro DESC
""")
print(resultado.to_string(index=False))

# Interpretación: Comparar sedes permite identificar si hay factores
# geográficos o institucionales que afectan la retención estudiantil.
# La tasa de egreso complementa el análisis: una sede puede tener
# menos retiro pero también menos egresados (más activos).
````

````python
# ============================================================
# Ejercicio 8d: Rendimiento por género y carrera — SOLUCIÓN
# ============================================================

# Pregunta: ¿Hay diferencias de rendimiento por género y carrera?

resultado = consulta("""
    SELECT carrera,
           genero,
           COUNT(*) AS estudiantes,
           ROUND(AVG(promedio), 2) AS promedio,
           ROUND(AVG(asistencia_pct), 1) AS asistencia,
           ROUND(AVG(materias_aprobadas), 1) AS aprobadas_prom
    FROM rendimiento
    GROUP BY carrera, genero
    ORDER BY carrera, genero
""")
print(resultado.to_string(index=False))

# Interpretación: El GROUP BY con dos columnas (carrera, genero) crea
# un grupo para cada combinación. Esto permite analizar brechas de
# género dentro de cada carrera. Las diferencias pueden orientar
# políticas de equidad o programas de apoyo focalizados.
````

### SQL Cheat Sheet

```sql
-- Básico
SELECT col1, col2 FROM tabla WHERE condicion;

-- Agregación
SELECT col, COUNT(*), AVG(val) FROM tabla GROUP BY col HAVING AVG(val) > 10;

-- JOIN
SELECT a.col1, b.col2 FROM tabla_a a INNER JOIN tabla_b b ON a.clave = b.clave;

-- Subconsulta
SELECT * FROM tabla WHERE valor > (SELECT AVG(valor) FROM tabla);

-- CASE WHEN
SELECT SUM(CASE WHEN condicion THEN 1 ELSE 0 END) AS conteo FROM tabla;
```
