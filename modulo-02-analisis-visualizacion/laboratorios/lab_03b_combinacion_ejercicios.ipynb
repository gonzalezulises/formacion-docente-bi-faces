## 📋 Resumen

En este laboratorio practicaste las operaciones fundamentales de combinación de datos:

| Operación | Función | Cuándo usarla |
|-----------|---------|---------------|
| Concatenar | `pd.concat()` | Apilar filas de DataFrames con las mismas columnas |
| Merge inner | `pd.merge(how='inner')` | Conservar solo registros con coincidencia en ambos lados |
| Merge left | `pd.merge(how='left')` | Conservar todos los registros del DataFrame izquierdo |
| Diagnóstico | `indicator=True` | Verificar cuántos registros coincidieron en el cruce |
| NaN check | `.isnull().sum()` | Detectar registros sin coincidencia después de un left merge |

```
Cargar datasets → Explorar columnas → Identificar claves → Concatenar/Merge → Diagnosticar → Interpretar
```
