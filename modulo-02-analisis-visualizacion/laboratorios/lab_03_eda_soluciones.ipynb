```python
# ============================================================
# Configuración inicial
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12
sns.set_palette('Set2')
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)

print('✅ Bibliotecas cargadas exitosamente')
```

---

## Ejercicio 1: Cargar e inspeccionar los datos

```python
# ============================================================
# SOLUCIÓN Ejercicio 1: Cargar e inspeccionar
# ============================================================

# 1. Cargar el dataset
df = pd.read_csv('../../datasets/universidad/rendimiento_academico.csv')

# 2. Dimensiones
print(f'📊 Dimensiones: {df.shape[0]:,} filas x {df.shape[1]} columnas')
print()

# 3. Tipos de datos
print('Tipos de datos:')
print(df.dtypes)
print()

# 4. Primeras filas
print('Primeras 5 filas:')
display(df.head())

# 5. Resumen estadístico
print('\nResumen estadístico:')
display(df.describe().round(2))

# Comentario interpretativo
print('\n📌 Observaciones iniciales:')
print(f'  - El dataset contiene {df.shape[0]:,} estudiantes de {df["carrera"].nunique()} carreras')
print(f'  - Variables numéricas: {len(df.select_dtypes(include=[np.number]).columns)}')
print(f'  - Variables categóricas: {len(df.select_dtypes(include="object").columns)}')
print(f'  - Variables booleanas: {len(df.select_dtypes(include="bool").columns)}')
```

---

## Ejercicio 2: Valores faltantes

```python
# ============================================================
# SOLUCIÓN Ejercicio 2: Valores faltantes
# ============================================================

# 1. Contar faltantes por columna
faltantes = df.isnull().sum()
print('❓ Valores faltantes por columna:')
print(faltantes)

# 2. Porcentaje de faltantes
pct_faltantes = (df.isnull().sum() / len(df) * 100).round(2)
print(f'\nPorcentaje de faltantes:')
print(pct_faltantes)

# 3. Visualizar o comentar
if faltantes.sum() == 0:
    print('\n✅ Este dataset NO tiene valores faltantes.')
    print('   En datos reales esto es poco común. Los datos de sistemas')
    print('   académicos suelen tener faltantes por registros incompletos,')
    print('   deserciones no registradas, o migraciones de sistema.')
else:
    fig, ax = plt.subplots(figsize=(12, 4))
    sns.heatmap(df.isnull().T, cmap='YlOrRd', yticklabels=True, ax=ax)
    ax.set_title('Mapa de Valores Faltantes')
    plt.tight_layout()
    plt.show()
```

---

## Ejercicio 3: Distribución del promedio

```python
# ============================================================
# SOLUCIÓN Ejercicio 3: Distribución del promedio
# ============================================================

# 1 y 2: Histograma + KDE + líneas de media/mediana
fig, ax = plt.subplots(figsize=(12, 6))

sns.histplot(df['promedio'], bins=30, kde=True, color='steelblue', ax=ax)

media = df['promedio'].mean()
mediana = df['promedio'].median()

ax.axvline(media, color='red', linestyle='--', linewidth=2, 
           label=f'Media: {media:.2f}')
ax.axvline(mediana, color='orange', linestyle='--', linewidth=2, 
           label=f'Mediana: {mediana:.2f}')
ax.axvline(10, color='gray', linestyle=':', linewidth=2, 
           label='Nota mínima aprobatoria: 10')

ax.set_xlabel('Promedio académico')
ax.set_ylabel('Frecuencia')
ax.set_title('Distribución del Promedio Académico — FACES')
ax.legend()

plt.tight_layout()
plt.show()

# 3. Estadísticas
print('\n📊 Estadísticas del promedio:')
print(f'  Media:               {media:.2f}')
print(f'  Mediana:             {mediana:.2f}')
print(f'  Desviación estándar: {df["promedio"].std():.2f}')
print(f'  Mínimo:              {df["promedio"].min():.1f}')
print(f'  Máximo:              {df["promedio"].max():.1f}')

# 4. Interpretación
pct_aprobados = (df['promedio'] >= 10).mean() * 100
print(f'\n📌 Interpretación:')
print(f'  - La distribución es aproximadamente simétrica (media ≈ mediana)')
print(f'  - El {pct_aprobados:.1f}% de los estudiantes tiene promedio >= 10 (aprobatorio)')
print(f'  - La mayoría de promedios se concentra entre {df["promedio"].quantile(0.25):.1f} y {df["promedio"].quantile(0.75):.1f}')
```

---

## Ejercicio 4: Boxplot del promedio por carrera

```python
# ============================================================
# SOLUCIÓN Ejercicio 4: Boxplot por carrera
# ============================================================

# 1. Boxplot
fig, ax = plt.subplots(figsize=(12, 6))
sns.boxplot(data=df, x='carrera', y='promedio', palette='Set2', ax=ax)

# 2. Línea de promedio general
promedio_general = df['promedio'].mean()
ax.axhline(promedio_general, color='red', linestyle='--', alpha=0.7,
           label=f'Promedio general: {promedio_general:.2f}')

ax.set_xlabel('Carrera')
ax.set_ylabel('Promedio académico')
ax.set_title('Distribución del Promedio por Carrera — FACES')
ax.legend()
ax.tick_params(axis='x', rotation=15)

plt.tight_layout()
plt.show()

# 3. Interpretación con datos
print('\n📌 Estadísticas por carrera:')
resumen = df.groupby('carrera')['promedio'].agg(['mean', 'median', 'std']).round(2)
resumen.columns = ['Media', 'Mediana', 'Desv.Est.']
print(resumen)

print(f'\n  - Mayor variabilidad: {resumen["Desv.Est."].idxmax()} (std={resumen["Desv.Est."].max():.2f})')
print(f'  - Mediana más alta: {resumen["Mediana"].idxmax()} ({resumen["Mediana"].max():.2f})')
print(f'  - Los outliers (puntos fuera de los bigotes) representan estudiantes')
print(f'    con rendimiento excepcionalmente alto o bajo para su carrera')
```

---

## Ejercicio 5: Tabla cruzada — Género x Beca

```python
# ============================================================
# SOLUCIÓN Ejercicio 5: Tabla cruzada género x beca
# ============================================================

# 1. Crosstab con conteos
ct_conteo = pd.crosstab(df['genero'], df['beca'], margins=True, margins_name='Total')
print('📊 Tabla cruzada: Género x Beca (conteos)')
print(ct_conteo)

# 2. Crosstab con porcentajes por fila
ct_pct = pd.crosstab(df['genero'], df['beca'], normalize='index').round(4) * 100
ct_pct.columns = ['Sin beca (%)', 'Con beca (%)']
print('\n📊 Tabla cruzada: Género x Beca (porcentajes)')
print(ct_pct)

# 3. Interpretación
pct_beca_f = ct_pct.loc['F', 'Con beca (%)']
pct_beca_m = ct_pct.loc['M', 'Con beca (%)']
diferencia = abs(pct_beca_f - pct_beca_m)

print(f'\n📌 Interpretación:')
print(f'  - Mujeres con beca: {pct_beca_f:.1f}%')
print(f'  - Hombres con beca: {pct_beca_m:.1f}%')
print(f'  - Diferencia: {diferencia:.1f} puntos porcentuales')
if diferencia < 3:
    print(f'  - La diferencia es pequeña, no parece haber brecha significativa de género en becas')
else:
    print(f'  - Existe una diferencia notable que merece investigación adicional')
```

---

## Ejercicio 6: Heatmap de correlaciones

```python
# ============================================================
# SOLUCIÓN Ejercicio 6: Heatmap de correlaciones
# ============================================================

# 1. Matriz de correlación
cols_num = df.select_dtypes(include=[np.number]).columns
correlaciones = df[cols_num].corr().round(3)

# 2. Heatmap con máscara triangular
mask = np.triu(np.ones_like(correlaciones, dtype=bool))

fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(correlaciones, mask=mask, annot=True, fmt='.2f',
            cmap='RdBu_r', center=0, vmin=-1, vmax=1,
            square=True, linewidths=0.5, ax=ax)
ax.set_title('Matriz de Correlación — Rendimiento Académico', fontsize=14)
plt.tight_layout()
plt.show()

# 3. Identificar correlaciones con promedio
print('\n🎯 Correlaciones con "promedio" (ordenadas por fuerza):')
corr_promedio = correlaciones['promedio'].drop('promedio').abs().sort_values(ascending=False)
for variable, valor in corr_promedio.items():
    signo = '+' if correlaciones.loc[variable, 'promedio'] > 0 else '-'
    fuerza = 'fuerte' if valor > 0.5 else 'moderada' if valor > 0.3 else 'débil'
    print(f'  {variable:25s} {signo}{valor:.3f}  ({fuerza})')

print('\n📌 Interpretación:')
print('  - materias_aprobadas tiene correlación positiva con promedio (lógico)')
print('  - materias_reprobadas tiene correlación negativa con promedio (lógico)')
print('  - Busca si asistencia_pct correlaciona con promedio (efecto de asistir a clases)')
```

---

## Ejercicio 7: Comparación — Estudiantes que trabajan vs no trabajan

```python
# ============================================================
# SOLUCIÓN Ejercicio 7: Trabajan vs no trabajan
# ============================================================

# 1. Filtrar grupos
trabajan = df[df['trabaja'] == True]
no_trabajan = df[df['trabaja'] == False]

print(f'Estudiantes que trabajan: {len(trabajan)}')
print(f'Estudiantes que NO trabajan: {len(no_trabajan)}')

# 2. Calcular promedios por grupo
comparacion = df.groupby('trabaja')[['promedio', 'asistencia_pct']].mean().round(2)
comparacion.index = ['No trabaja', 'Trabaja']
print(f'\n📊 Comparación:')
print(comparacion)

# 3. Gráfico de barras comparativo
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Promedio académico
sns.barplot(data=df, x='trabaja', y='promedio', palette=['seagreen', 'coral'], 
            ci=95, ax=axes[0])
axes[0].set_xticklabels(['No trabaja', 'Trabaja'])
axes[0].set_ylabel('Promedio académico')
axes[0].set_title('Promedio según condición laboral')

# Asistencia
sns.barplot(data=df, x='trabaja', y='asistencia_pct', palette=['seagreen', 'coral'],
            ci=95, ax=axes[1])
axes[1].set_xticklabels(['No trabaja', 'Trabaja'])
axes[1].set_ylabel('Asistencia (%)')
axes[1].set_title('Asistencia según condición laboral')

plt.suptitle('Impacto de la condición laboral en el rendimiento académico',
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# 4. Interpretación
dif_promedio = comparacion.loc['No trabaja', 'promedio'] - comparacion.loc['Trabaja', 'promedio']
dif_asistencia = comparacion.loc['No trabaja', 'asistencia_pct'] - comparacion.loc['Trabaja', 'asistencia_pct']

print(f'\n📌 Interpretación:')
print(f'  - Diferencia en promedio: {dif_promedio:+.2f} puntos')
print(f'  - Diferencia en asistencia: {dif_asistencia:+.1f} puntos porcentuales')
if dif_promedio > 0:
    print('  - Los estudiantes que NO trabajan tienen mejor rendimiento')
else:
    print('  - Los estudiantes que trabajan tienen igual o mejor rendimiento')
```
