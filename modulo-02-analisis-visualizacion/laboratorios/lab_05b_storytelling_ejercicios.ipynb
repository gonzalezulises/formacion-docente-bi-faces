## Ejercicio 3: Aplicar principios de storytelling

Ahora transforma el mismo grafico aplicando los **6 principios de Cole Nussbaumer Knaflic**:

### Lista de verificacion

| # | Principio | Que hacer | Ejemplo |
|---|-----------|-----------|--------|
| 1 | **Titulo descriptivo** | No "Grafico 1" sino una oracion que diga que pasa | "La matricula de FACES cayo 45% entre 2015 y 2022" |
| 2 | **Colores intencionales** | Resaltar el dato clave con un color contrastante | Rojo para el punto critico, gris para el resto |
| 3 | **Anotaciones** | Usar `plt.annotate()` para marcar el punto de inflexion | Flecha apuntando al anio de mayor caida |
| 4 | **Diseno limpio** | Eliminar cuadricula innecesaria, bordes superiores/derechos | `ax.spines['top'].set_visible(False)` |
| 5 | **Fuente de datos** | Agregar con `plt.figtext()` | "Fuente: Datos sinteticos basados en FACES-UC" |
| 6 | **Insight al pie** | Breve oracion que resume el hallazgo | `plt.figtext(0.1, -0.02, 'Insight: ...')` |

### Codigo de referencia para anotaciones

```python
ax.annotate('Punto critico',

            xytext=(x_texto, y_texto),      # donde va el texto
            fontsize=11, fontweight='bold',
            color='red',
            arrowprops=dict(arrowstyle='->', color='red', lw=1.5))
```

## Ejercicio 6: Secuencia narrativa de 3 graficos

Toda buena historia tiene 3 actos. Crea una figura con **3 subplots en una fila** que cuenten tu historia en 3 actos:

| Acto | Nombre | Que mostrar | Emocion que genera |
|------|--------|------------|--------------------|
| 1 | **Planteamiento** (Setup) | La linea base o contexto inicial | Contexto, tranquilidad |
| 2 | **Conflicto** (Conflict) | El problema, el cambio, la ruptura | Preocupacion, urgencia |
| 3 | **Resolucion** (Resolution) | Lo que esta pasando ahora o una proyeccion | Esperanza, accion |

### Ideas segun tu historia

| Historia | Acto 1 (Planteamiento) | Acto 2 (Conflicto) | Acto 3 (Resolucion) |
|----------|----------------------|-------------------|--------------------|
| (a) Matricula | Matricula total en el primer anio (barra unica o KPI) | Linea descendente completa | Barras por carrera del ultimo anio (cuales sobreviven) |
| (b) Inflacion | Inflacion mensual 2015-2016 (estable, un digito) | Explosion 2017-2019 (escalada dramatica) | Situacion reciente (se estabilizo o no?) |
| (c) Genero | Composicion de genero en el primer anio | Evolucion de la brecha a lo largo del tiempo | Composicion actual (se cerro la brecha?) |

### Estructura del codigo

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 5))

# Acto 1: Planteamiento
ax1.set_title('Acto 1: ...', fontweight='bold')
# ...

# Acto 2: Conflicto
ax2.set_title('Acto 2: ...', fontweight='bold', color=COLORES['rojo'])
# ...

# Acto 3: Resolucion
ax3.set_title('Acto 3: ...', fontweight='bold', color=COLORES['verde'])
# ...

fig.suptitle('Titulo que une los 3 actos', fontsize=16, fontweight='bold', y=1.05)
plt.tight_layout()
plt.show()
```
